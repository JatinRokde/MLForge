Anomaly Detection is one of the highest-value applications of Machine Learning in industry.

Unlike traditional classification:
```text
Fraud → Very Rare
Failures → Very Rare
Attacks → Very Rare
Defects → Very Rare
```

Most real-world systems have:
```text id="vcr0sn"
99.9% Normal Data
0.1% Abnormal Data
```

This creates a major challenge.

### What is Anomaly Detection?
An anomaly is:
> An observation that significantly deviates from normal behavior.

Example:

Customer transactions:

```text
100
120
95
110
130
```

Normal.

Suddenly:
```text
10000
```

Potential anomaly.

### Why Not Use Classification?
Fraud detection example:
```
99.95% Legitimate
0.05% Fraud
```

Problems:
- Very few fraud labels
- New fraud patterns appear constantly
- Labels arrive late

Thus:
```text
Unsupervised methods
```

become extremely important.

### Types of Anomalies
#### Point Anomaly
Single unusual observation.

Example:
```text 
Income:
40k
42k
38k
41k
500k
```
500k is anomalous.

#### Contextual Anomaly
Normal in one context.

Abnormal in another.

Example:
```
Temperature = 40°C
```
Normal:
```
Summer
```
Abnormal:
```
Winter
```

#### Collective Anomaly
Individual observations appear normal.

Group behavior abnormal.

Example:
```text 
100 login attempts
within 2 seconds
```

Possible attack.

### Core Principle
Most anomaly detection algorithms assume:
```text 
Normal observations
are common
```
and
```text 
Anomalies are rare
```

### Distance-Based Intuition
Suppose:
```text 
● ● ● ● ● ● ●
```
Dense region.

Another point:
```text 
                 ●
```
Far away.

Likely anomaly.

### Density-Based Intuition
Normal points:
```
Dense neighborhoods
```
Anomalies:
```
Sparse neighborhoods
```
This idea appears in LOF.

### Method 1: Isolation Forest
Most widely used production anomaly detector.

#### Core Idea
Instead of:
```text
Finding normal points
```
Isolation Forest tries to:
```text 
Isolate anomalies
```

Observation:
Anomalies are easier to isolate.

Example:
```text 
100
101
102
103
104
```

Need many splits.

But:
```text
1000
```
Can be isolated immediately.

#### Isolation Tree
Randomly:
* Choose feature
* Choose split value

Repeat recursively.

Example:
```text
Income < 50k?
```

Then:
```text 
Income < 20k?
```

Continue until isolated.

#### Path Length
Important concept.

Normal points:
```
Need many splits
```
Long path.

Anomalies:
```
Need few splits
```
Short path.

#### Isolation Score
Average path length:
$$
h(x)
$$

Short path:
```
High anomaly score
```
Long path:
```
Normal point
```

#### Python Example
```python 
from sklearn.ensemble import IsolationForest

model = IsolationForest(
    contamination=0.05,
    random_state=42
)

labels = model.fit_predict(X)
```

Output:
```text
1
```

Normal.
```text
-1
```

Anomaly.

#### Advantages
- Fast
- Scalable
- Works in high dimensions
- Production favorite

### Method 2: Local Outlier Factor (LOF)
#### Core Idea
Anomaly depends on local density.

Consider:
```
Cluster A
```
Dense city.

And:
```
Cluster B
```
Sparse rural area.

Same distance may mean different things.

LOF compares:

```text 
Point density
vs
Neighbor density
```

#### Local Reachability Density
LOF computes:
$$
LRD
$$

Local density around a point.

Then compares:
```
Neighbor Density
```
with
```
Own Density
```

#### LOF Score
Approximation:
$$
LOF = \frac{Neighbor \ Density}{Own \ Density}
$$

Normal:
$$
LOF ≈ 1
$$

Anomaly:
$$
LOF ≫ 1
$$

#### Python Example
```python 
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(
    n_neighbors=20
)

labels = lof.fit_predict(X)
```

Again:
```
1
```

Normal.
```text 
-1
```
Outlier.

#### Advantages
- Excellent local anomaly detection
- Detects density changes
- Strong for geographical data

#### Limitations
- Slower
- Poor scalability
- Sensitive to neighbor count

### Method 3: One-Class SVM
#### Core Idea
Train only on normal data.

Learn:
```text
Boundary around normal observations
```

Anything outside:
```text
Anomaly
```

#### Intuition
Imagine:
```
Normal region
```
inside a bubble.

New point:

Inside:
```
Normal
```
Outside:
```
Anomaly
```

#### Kernel Trick
Same kernel concepts as SVM.

Can learn:
- Nonlinear boundaries
- Complex regions

#### Python Example
```python 
from sklearn.svm import OneClassSVM

model = OneClassSVM(
    kernel="rbf",
    gamma="scale"
)

labels = model.fit_predict(X)
```

#### Advantages
- Powerful nonlinear detection
- Theoretically elegant

#### Limitations
- Slow
- Memory intensive
- Difficult on huge datasets

### Comparison

| Feature          | Isolation Forest | LOF       | One-Class SVM |
| ---------------- | ---------------- | --------- | ------------- |
| Scalability      | Excellent        | Moderate  | Poor          |
| High Dimensions  | Good             | Moderate  | Moderate      |
| Local Density    | Moderate         | Excellent | Moderate      |
| Speed            | Fast             | Medium    | Slow          |
| Production Usage | Very High        | Medium    | Medium        |
