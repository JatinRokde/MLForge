### What is Unsupervised Learning?
In Unsupervised Learning:
```
Input Features (X)
No Labels (Y)
```
Only raw data is available.

Goal:
```
Find hidden patterns
Find structure
Find relationships
Discover groups
Detect anomalies
Reduce dimensions
```

### Real-World Analogy
Imagine entering a library where:
- Thousands of books exist
- No categories exist
- No labels exist

The task is:
```
Automatically organize books
```
based on similarity.

This is unsupervised learning.

### Major Problem Types in Unsupervised Learning
#### 1. Clustering
Goal:
```
Group similar data points together
```

Examples:
- Customer segmentation
- User behavior analysis
- Market segmentation

Algorithms:
- K-Means
- Hierarchical Clustering
- DBSCAN
- Gaussian Mixture Models

#### 2. Dimensionality Reduction
Goal:
```
Compress information
while preserving patterns
```

Examples:
- PCA
- t-SNE
- UMAP

Used for:
- Visualization
- Noise reduction
- Faster training

#### 3. Anomaly Detection
Goal:
```
Find unusual observations
```

Examples:
- Fraud detection
- Intrusion detection
- Manufacturing defects

#### 4. Association Rule Mining
Goal:
```
Discover relationships
between items
```

Example:
```
People buying bread
often buy butter
```
Used in recommendation systems.

### How Unsupervised Learning Differs from Supervised Learning
| Aspect           | Supervised                 | Unsupervised    |
| ---------------- | -------------------------- | --------------- |
| Labels Available | Yes                        | No              |
| Target Variable  | Present                    | Absent          |
| Objective        | Predict                    | Discover        |
| Evaluation       | Straightforward            | Harder          |
| Examples         | Regression, Classification | Clustering, PCA |

### Challenges in Unsupervised Learning
Unlike supervised learning:

There is no:
```
Correct answer
```
during training.

This makes evaluation harder.

Evaluation often requires:
- Domain knowledge
- Silhouette score
- Cluster quality metrics
