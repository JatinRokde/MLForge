### What is Association Rule Mining?
Association Rule Mining discovers:
> Items that frequently occur together.

Example:

Suppose shopping transactions:

| Transaction | Items               |
| ----------- | ------------------- |
| T1          | Bread, Butter       |
| T2          | Bread, Milk         |
| T3          | Bread, Butter, Milk |
| T4          | Bread, Eggs         |

Possible hidden pattern:
```text
Bread → Butter
```

Meaning:

Customers purchasing bread often purchase butter.

### Market Basket Analysis
Classic use case.

Imagine supermarket receipts:
```
Transaction 1
Milk
Bread
Eggs
```
```
Transaction 2
Bread
Butter
```
```
Transaction 3
Milk
Bread
Butter
```
Goal:
```
Find hidden purchasing patterns
```

### Association Rule Structure
Rules look like:
$$
A ⇒ B
$$

Example:
$$
Bread ⇒ Butter
$$

Interpretation:
```
If Bread is purchased,
Butter is likely purchased.
```
Important:

Association ≠ Causation

**Incorrect** : Bread causes Butter purchase

**Correct** : Bread and Butter frequently occur together

### Key Metrics
Three metrics dominate:
1. Support
1. Confidence
1. Lift

#### Support
Most fundamental metric.

Support measures:
> How frequently an itemset appears.

Formula:
$$
Support(A) = \frac{Transactions \ containing \ A}{Total \ Transactions}
$$

Example:

Transactions:
```
5 transactions
```

Bread appears in:
```text 
4 transactions
```

Support:
$$
\frac{4}{5} = 0.8
$$

Meaning:
```
80% of transactions contain Bread
```

##### Support of Itemsets
Suppose:
```text
Bread + Butter
```

appears in:
```text 
3 transactions
```

out of:
```text
5 transactions
```

Support:
$$
\frac{3}{5} = 0.6
$$

#### Confidence
Measures rule reliability.

Formula:
$$
Confidence(A ⇒ B) = \frac{Support(A ∩ B)}{Support(A)}
$$

Interpretation:
> Probability of B given A.

Example:

Bread appears:
```text
4 times
```

Bread + Butter appears:
```text 
3 times
```

Confidence:
$$
\frac{3}{4} = 0.75
$$

Meaning:
```
75% of Bread buyers
also buy Butter
```

#### Lift
Most important business metric.

Problem:

Confidence can be misleading.

Suppose:
```
Milk purchased by everyone
```
Confidence becomes artificially high.

Lift corrects for this.

Formula:
$$
Lift(A ⇒ B) = \frac{Confidence(A ⇒ B)}{Support(B)}
$$

##### Lift Interpretation
**Lift = 1**

No relationship.

**Lift > 1**

Positive association.

**Lift < 1**

Negative association.

Example:
$$
Lift = 3
$$

Means:
```text 
Three times more likely
than random chance
```

### Example Calculation
Transactions:
| T | Items        |
| - | ------------ |
| 1 | Bread,Butter |
| 2 | Bread        |
| 3 | Bread,Butter |
| 4 | Milk         |
| 5 | Bread,Butter |

Support:
$$
Bread=4/5=0.8
$$

Support:
$$
Butter=3/5=0.6
$$

Support:
$$
Bread+Butter=3/5=0.6
$$

Confidence:
$$
0.6/0.8=0.75
$$

Lift:
$$
0.75/0.6=1.25
$$

Positive relationship.

### Apriori Algorithm
Most famous association mining algorithm.

### Apriori Principle
Core observation:
> If an itemset is frequent, all its subsets must also be frequent.

Example:

If:
```text 
Bread
Butter
Milk
```

is frequent.

Then:
```text
Bread + Butter
```

must also be frequent.

This allows massive pruning.

### Apriori Workflow
**Step 1:**

Find frequent 1-itemsets.

**Step 2:**

Generate candidate 2-itemsets.

**Step 3:**

Remove infrequent itemsets.

**Step 4:**

Generate 3-itemsets.

Continue.

### Example
Items:
```text 
Bread
Milk
Butter
Eggs
```

Frequent:
```text 
Bread
Milk
Butter
```

Generate:
```text 
Bread+Milk
Bread+Butter
Milk+Butter
```

Continue recursively.

### Apriori Limitation
Huge datasets create:
```text 
Millions of candidates
```

Very expensive.

This led to: **FP-Growth**

### FP-Growth
FP = Frequent Pattern

Goal:

Avoid candidate generation.

Uses:
```text 
FP Tree
```
Compressed transaction representation.

Example:

Transactions:

```text 
Bread Milk
Bread Butter
Bread Milk Butter
```

Shared paths stored efficiently.

Result:
- Faster
- Less memory
- Scalable

### FP Tree Intuition
Instead of storing:
```text 
Bread
Bread
Bread
```

multiple times,

store once with counts.

Compression achieved.

### Apriori vs FP-Growth
| Feature              | Apriori  | FP-Growth |
| -------------------- | -------- | --------- |
| Candidate Generation | Yes      | No        |
| Speed                | Slower   | Faster    |
| Memory               | Higher   | Lower     |
| Scalability          | Moderate | Excellent |

### Python Example (Apriori)
```python 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

freq_items = apriori(
    df,
    min_support=0.05,
    use_colnames=True
)

rules = association_rules(
    freq_items,
    metric="lift",
    min_threshold=1.0
)

print(rules.head())
```
