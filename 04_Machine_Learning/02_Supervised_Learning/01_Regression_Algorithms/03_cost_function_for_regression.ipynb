### What is a Cost Function?
A cost function measures:
```text
How wrong is the model?
```

In regression, the model predict:   
$
\hat{y} = wX + b 
$

Predictions are generally not perfect, so the error must be measured.

### Error vs Cost
For one sample:  
$
Error = y - \hat{y}
$
For all samples:    
Errors are aggregated to compute the **cost.**


### Most Important Cost Function: MSE
#### Mean Squared Error (MSE)
$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$

#### Why Square?
- Removes negative signs
- Penalizes large errors heavily
- Differentiable (important for gradient descent)

### Example Calculation
**Dataset:**
| Actual | Predicted |
| ------ | --------- |
| 10     | 8         |
| 20     | 18        |
| 30     | 35        |

**Errors:**
```text
(10 - 8) = 2
(20 - 18) = 2
(30 - 35) = -5
```

**Squared:**
```text
4, 4, 25
```

**MSE:**
```text
(4 + 4 + 25) / 3 = 11
```

### Python Implementation

````python
import numpy as np

y_true = np.array([10, 20, 30])
y_pred = np.array([8, 18, 35])

mse = np.mean((y_true - y_pred)**2)
print("MSE : ", mse)
````

### Geometric Interpretation
n simple linear regression, the model is:    
$
\hat{y} = w_0 + w_1 x
$    

where:
- $w_0$ = intercept
- $w_1$ = slope

The objective is to minimize the **Mean Squared Error (MSE)** cost function:
$
J(w_0, w_1) = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - (w_0 + w_1 x_i) \right)^2
$

#### Why the Cost Function Forms a Bowl Shape
The cost function contains **squared terms**. Squaring produces a quadratic expression in terms of $w_0$ and $w_1$.   

A quadratic function has the general form:   
$
ax^2 + bx + c
$

When $a \gt 0$, the graph is a **U-shaped parabola** (convex).

Since MSE consists only of positive squared terms, the curvature is always upward. That is the source of the bowl shape.

#### Cost vs One Parameter (Parabola)
If one parameter is fixed and the other varies:
- The cost behaves as a quadratic function
- The graph froms a smooth U-shape
- The lowest point corresponds to minimum error

This minimum is unique.

### Matrix Form of MSE
In vector form:   
$
J(\mathbf{w}) = \frac{1}{n} (X\mathbf{w} - \mathbf{y})^{T}(X\mathbf{w} - \mathbf{y})
$

Where:
- X = feature matrix
- w = weight vector
- y = target vector

### Derivative of MSE
To minimize MSE:

Take derivative wrt weights.

Result:    
$
\frac{\partial J}{\partial \mathbf{w}} = -\frac{2}{n} X^{T} (\mathbf{y} - X\mathbf{w})
$

This gradient is used in:
```
w = w - learning_rate * gradient
```
This is how regression learns.

### Gradient Descent Example

````python
import numpy as np

# Data
X = np.array([[1], [2], [3], [4]])
y = np.array([30, 40, 55, 70])

# Add bias term
X = np.c_[np.ones(X.shape[0]), X]

w = np.zeros(X.shape[1])
lr = 0.01
n = len(y)

for i in range(1000):
    y_pred = X.dot(w)
    gradient = (-2/n) * X.T.dot(y - y_pred)
    w -= lr * gradient

print("Weights: ", w)
````

### When to Use What?
| Cost Function | Use Case           |
| ------------- | ------------------ |
| MSE           | Default regression |
| MAE           | Outliers present   |
| RMSE          | Interpretability   |


#### Outlier Effect
One large error -> MSE increases drastically.
