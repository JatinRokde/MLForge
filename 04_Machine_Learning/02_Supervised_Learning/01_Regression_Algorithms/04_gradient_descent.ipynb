### Problem Setup
The linear regression model in matrix form:   
$
\hat{y} = Xw
$   

Where:   
- X -> (n × m) feature matrix
- 𝑤 -> (m × 1) weight vector
- 𝑦 -> (n × 1) target 

Bias is included in X by adding column of 1s.

### Cost Function (MSE in Matrix Form)
$
J(w) = \frac{1}{n} (y - Xw)^{T} (y - Xw)
$

This is the compact matrix representation of Mean Squared Error.
- (y − Xw) -> error vector
- Squaring is achieved via transpose multiplication
- Dividing by n gives the average error

### Gradient Derivation
Taking derivative of $J(w)$ with respect to $w$:   
$
\nabla J(w) = -\frac{2}{n} X^{T}(y - Xw)
$

This vector represents the direction of steepest increase.

To minimize the cost, update in the opposite direction:   
$
w := w - \alpha \nabla J(w)
$

Substituting the gradient:   
$
w := w + \frac{2\alpha}{n} X^{T}(y - Xw)
$

Where:
- $\alpha$ -> learning rate

### Algorithm Steps
1. Initialize weights (zeros or small random values)
2. Predict:   
$
\hat{y} = Xw
$
3. Compute error:  
$
e = y - \hat{y}
$
4. Compute gradient:  
$
-\frac{2}{n} X^{T}e
$
5. Update weights:   
$
w := w + \frac{2\alpha}{n} X^{T}e
$
6. Repeat until convergence (cost stabilizes or maximum iterations reached)


### Implementation from Scratch
**Example Dataset:**
| Area | Bedrooms | Price |
| ---- | -------- | ----- |
| 1000 | 2        | 60    |
| 1500 | 3        | 90    |
| 2000 | 4        | 130   |
| 1200 | 2        | 70    |


````python
import numpy as np

# Prepare Data
X = np.array([
    [1000, 2],
    [1500, 3],
    [2000, 4],
    [1200, 2]
])

y = np.array([60, 90, 130, 70])

# Add bias column (1s)
X = np.c_[np.ones(X.shape[0]), X]

n_samples, n_features = X.shape

# Initialize Weights
w = np.zeros(n_features)

learning_rate = 1e-7
epochs = 1000

# Gradient Descent
for i in range(epochs):
    # Prediction
    y_pred = X.dot(w)

    # Error
    error = y - y_pred

    # Gradient
    gradient = (-2/n_samples) * X.T.dot(error)

    # Update
    w -= learning_rate * gradient

    # Print loss
    if i % 100 == 0:
        loss = np.mean(error**2)
        print(f"Epoch {i}, Loss {loss}")

print("Final Weights: ", w)
````

### Important Observations
#### Learning Rate Matters
If too large:
```
Loss explodes
```

If too small:
```
Very slow convergence
```

#### Feature Scaling is Critical
Area = 1000  
Bedrooms = 2    

Huge scale difference -> unstable gradients

### Why Gradient Descent Instead of Closed Form?
Linear regression also has **Normal Equation:**  
$
w = (X^T X)^{-1} X^T y
$

But:
| Method           | Pros        | Cons                |
| ---------------- | ----------- | ------------------- |
| Normal Equation  | Exact       | Slow for large data |
| Gradient Descent | Scales well | Iterative           |

For large datasets -> Gradient Descent preferred.

### Computational Complexity
Per iteration:
$
O(nm)
$  

Where:
- n = samples
- m = featues

Scales well to large data.
