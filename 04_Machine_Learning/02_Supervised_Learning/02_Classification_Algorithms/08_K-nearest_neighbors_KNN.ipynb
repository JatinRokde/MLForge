**K-Nearest Neighbors (KNN)** is a **supervised machine learning algorithm** used for:
- **Classification**
- **Regression**

KNN is called a **lazy learning algorithm** because:

- No explicit training phase exists
- The algorithm stores the dataset
- Computation happens during prediction time

KNN predicts output based on the **closest data points** in feature space.

### Core Intuition
#### Fundamental Assumption

> Similar data points exist close to each other.

Examples:
| Scenario               | Intuition                                     |
| ---------------------- | --------------------------------------------- |
| House pricing          | Nearby houses have similar prices             |
| Spam detection         | Similar email structures belong to same class |
| Recommendation systems | Similar users like similar products           |
| Fraud detection        | Similar transaction patterns cluster together |

### KNN Classification Example

Suppose a dataset contains:
| Height | Weight | Class |
| ------ | ------ | ----- |
| 170    | 65     | Fit   |
| 172    | 68     | Fit   |
| 160    | 85     | Unfit |
| 158    | 88     | Unfit |

New data point:
```
Height = 171
Weight = 66
```

KNN process:

1. Calculate distance from all points
1. Select nearest neighbors
1. Perform majority voting
1. Predict dominant class

If nearest neighbors are mostly `"Fit"`:
```
Prediction = Fit
```

### Mathematical Foundation
#### Euclidean Distance

Most commonly used distance metric.

##### Formula
$
d(x, y) = \sqrt{(x_1 - y_1)^2 + (x_2 - y_2)^2 + \cdots + (x_n - y_n)^2}
$

Where:
- $x$ = existing data point
- $y$ = query point

##### Example

Point A:
```
(2, 3)
```
Point B:
```
(5, 7)
```
Distance:  

$
d = \sqrt{(5 - 2)^2 + (7 - 3)^2} \\
d = \sqrt{9 + 16} \\
d = \sqrt{25} \\
d = 5
$

### Complete KNN Workflow
#### Step 1 — Choose K

K = number of nearest neighbors.

Example:
```
K = 3
```

Means:
- Select 3 nearest points

#### Step 2 — Compute Distances

Distance between query point and every training point.

#### Step 3 — Sort Distances

Nearest points come first.

#### Step 4 — Select Top K

Take first K neighbors.

#### Step 5 — Voting / Averaging
##### Classification

Use majority voting.

##### Regression

Use averaging.

### KNN for Classification
#### Majority Voting
Suppose:
```
K = 5
```
Nearest classes:
```
A, A, B, A, B
```
Prediction:
```
A
```
Because:
```
A occurs 3 times
B occurs 2 times
```

### KNN for Regression

Instead of voting:
- Average target values

Example:

Nearest values:
```
20, 25, 30
```

Prediction:
```
20 + 25 + 30 / 3 = 25
```

### Choosing Optimal K
#### Small K

Example:
```
K = 1
```
Characteristics:
- Sensitive to noise
- Overfitting risk
- High variance

#### Large K

Example:
```
K = 50
```
Characteristics:
- Too smooth
- Underfitting risk
- High bias

### Bias-Variance Tradeoff in KNN
| K Value | Bias | Variance |
| ------- | ---- | -------- |
| Small K | Low  | High     |
| Large K | High | Low      |

### Distance Metrics
#### A. Euclidean Distance
Most common.

$
d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$

Used for:
- Continuous features

#### B. Manhattan Distance
$
\sum |x_i - y_i|
$

Used when:
- Grid-like movement exists

#### C. Minkowski Distance
Generalized form. 

$
(\sum |x_i - y_i|^p)^{1/p}
$

#### D. Cosine Similarity

Used for:
- Text vectors
- NLP
- Recommendation systems

$
\cos(\theta) = \frac{A \cdot B}{\|A\| \, \|B\|}
$

### Feature Scaling in KNN
#### Extremely Important

KNN is **distance-based.**

Large-scale features dominate distance calculations.

##### Problem Example
| Feature | Range            |
| ------- | ---------------- |
| Age     | 18–60            |
| Salary  | 10,000–1,000,000 |

Salary dominates distance.

Result:
- Wrong neighbors selected

### Feature Scaling Techniques
#### Standardization
$
z = \frac{x - \mu}{\sigma}
$

Mean = 0  
Standard deviation = 1

#### Normalization

$
x' = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
$

Range:
```
0 to 1
```

### Why KNN is Called Lazy Learning

Traditional models:
```
Train first → Predict later
```

KNN:
```
Store data → Predict directly
```
No mathematical model gets built.

### Time Complexity
#### Training Complexity
$
O(1)
$ 

Very fast training.

#### Prediction Complexity
$
O(n)
$

Because distance to all points is calculated.

Large datasets become expensive.

### Space Complexity

KNN stores entire dataset.

$
O(n)
$

Memory-heavy for large systems.

### Curse of Dimensionality
#### Major Production Problem

As dimensions increase:

- Distances become less meaningful
- Points appear equally distant
- Performance degrades

##### Example

2D:
```
Neighbors are meaningful
```
1000D:
```
Everything becomes sparse
```

### Solutions to Curse of Dimensionality
#### A. Feature Selection

Remove irrelevant features.

#### B. PCA

Use dimensionality reduction.

#### C. Use Alternative Algorithms

Examples:
- Random Forest
- XGBoost
- Neural Networks

### Weighted KNN

Closer neighbors receive higher importance.

##### Formula
$
Weight = 1 / Distance
$

##### Example
| Neighbor | Distance | Weight |
| -------- | -------- | ------ |
| A        | 1        | 1      |
| B        | 5        | 0.2    |

Closer point influences prediction more.

### Decision Boundary

KNN produces:
- Non-linear decision boundaries

This makes KNN powerful for complex distributions.

##### Solution

Use:

- KD Trees
- Ball Trees

KNN unsuitable for ultra-low latency systems.

Example:

- High-frequency trading
- Real-time fraud prevention

#### C. Dynamic Data

Frequent updates require re-indexing.


### KD Tree Optimization

KD Tree reduces nearest neighbor search complexity.

Instead of:
```
Checking every point
```
Use:
```
Spatial partitioning
```
Average complexity:

$
O(logn)
$

for low-dimensional data.

### Problem Statement
A large e-commerce company wants to build a **real-time product recommendation engine.**

When a customer opens a product page, the platform should instantly recommend:

- Similar products
- Frequently co-viewed products

Requirements:

- Recommendations should adapt quickly to new products

- Similarity search
- Neighbor retrieval

Core idea:
> Similar users behave similarly
> Similar products are purchased/viewed together

````python
import pandas as pd
import numpy as np
import joblib
import os

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
````

````python
# Load Dataset
DATA_PATH = "./data/knn_recommendation_dataset.csv"

df = pd.read_csv(DATA_PATH)
df.head()
````

````python
df.shape
````

````python
# Dataset Overview
print(df.info())
print(df.describe())
````

````python
# Feature Selection
FEATURE_COLUMNS = [
    "price",
    "avg_rating",
    "popularity_score",
    "session_duration_minutes",
    "pages_viewed",
    "purchase_frequency",
    "click_through_rate",
    "discount_percentage",
    "wishlist_count",
    "embedding_1",
    "embedding_2",
    "embedding_3",
    "embedding_4",
    "embedding_5"
]

X = df[FEATURE_COLUMNS]
````

````python
# Feature Scaling
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
print(X_scaled.shape)
````

````python
# Train KNN Model

knn_model = NearestNeighbors(
    n_neighbors=10,
    metric="cosine",
    algorithm="brute"
)

knn_model.fit(X_scaled)
````

````python
# Save Artifacts
model_dir = "project-knn_classification"

os.makedirs(model_dir, exist_ok=True)

joblib.dump(knn_model, os.path.join(model_dir, "knn_model.pkl"))
joblib.dump(scaler, os.path.join(model_dir, "scaler.pkl"))
joblib.dump(FEATURE_COLUMNS, os.path.join(model_dir, "feature_columns.pkl"))
````

````python
# Recommendation Function
product_lookup = df.reset_index(drop=True)

def get_similar_products(product_index, top_k=5):
    # converts into 2D vector
    query_vector = X_scaled[product_index].reshape(1, -1)

    distances, indices = knn_model.kneighbors(
        query_vector,
        n_neighbors=top_k+1
    )

    recommendations = []

    for idx, dist in zip(indices[0][1:], distances[0][1:]):
        row = product_lookup.iloc[idx]

        recommendations.append({
            "product_id": row["product_id"],
            "category": row["category"],
            "brand": row["brand"],
            "similarity_score": round(1 - dist, 4)
        })

    return recommendations
````

````python
# Test Recommendation Engine
sample_product_index = 100

recommendations = get_similar_products(
    product_index=sample_product_index,
    top_k=5
)

recommendations
````

````python
batch_results = []

for idx in range(10):

    recs = get_similar_products(idx, top_k=3)

    batch_results.append({
        "source_product": df.iloc[idx]["product_id"],
        "recommendations": recs
    })

batch_results
````
