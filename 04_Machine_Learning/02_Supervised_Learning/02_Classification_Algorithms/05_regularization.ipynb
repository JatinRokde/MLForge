### Regularization in Logistic Regression

Regularization is a technique used to **prevent overfitting** in Logistic Regression by adding a **penalty term** to the cost function.

Without regularization:

- Model may memorize training data
- Coefficients can become very large
- Poor generalization on unseen data

Regularization helps the model:

- Stay simpler
- Reduce variance
- Improve generalization
- Handle multicollinearity

### Why Logistic Regression Needs Regularization

Logistic Regression tries to minimize this cost function:  

$
J(\theta) = -\frac{1}{m} \sum_{i=1}^{m}
\left[y^{(i)} \log\left(h_\theta(x^{(i)})\right) + (1-y^{(i)}) \log\left(1-h_\theta(x^{(i)})\right) \right]
$

Where:

$
h_\theta(x) = \sigma(\theta^T x)
$

If features are noisy or too many:

- coefficients may explode
- decision boundary becomes overly complex

Regularization controls coefficient size.

### Intuition

Think of regularization as:

> “Don’t trust any single feature too much.”

It discourages large weights.

Example:

Without regularization:
```
θ=[0.2,50,−80,120]
```
This means the model depends heavily on a few features.

With regularization:
```
θ=[0.2,2.1,−1.8,3.5]
```
Much more stable.

### Types of Regularization
| Type       | Name    | Effect                       |
| ---------- | ------- | ---------------------------- |
| L1         | Lasso   | Makes some weights exactly 0 |
| L2         | Ridge   | Shrinks weights smoothly     |
| ElasticNet | L1 + L2 | Combination of both          |


#### L2 Regularization (Ridge)
Most common in Logistic Regression.

##### Modified Cost Function
$
J(\theta) = LogLoss + \frac{\lambda}{2m}\sum_{j=1}^{n} \theta_j^2
$

Notice:
- We add squared weights
- Bias term $\theta_0 $ is usually excluded

##### Intuition of L2
Large weights are penalized heavily because of squaring.

Example:
| Weight | Squared |
| ------ | ------- |
| 2      | 4       |
| 10     | 100     |

So the optimizer avoids huge coefficients.

##### Effect of L2
- Reduces overfitting
- Keeps all features
- Makes weights smaller
- Smooth decision boundary
- Works well with correlated features

##### Geometric Interpretation

L2 constrains weights inside a circle/sphere.

The optimizer searches:

- minimize loss
- while staying inside small weight region

#### L1 Regularization (Lasso)
##### Cost Function
$
J(\theta) = LogLoss + \frac{\lambda}{m}\sum_{j=1}^{n} |\theta_j|
$

Instead of squares:  
$
|\theta|
$  
is used.

##### Key Property of L1

L1 can force coefficients to become exactly:
$
0
$
Thus:
- performs feature selection automatically
- removes unimportant features

##### Example
Before:
```
[2.3,−1.2,0.8,4.5]
```
After L1:
```
[2.1,0,0,4.2]
```
Some features disappear completely.

##### Effect of L1
- Sparse models
- Automatic feature selection
- Useful in high-dimensional datasets
- Can improve interpretability

#### ElasticNet Regularization

Combination of L1 and L2.

##### Cost Function
$
J(\theta) = LogLoss + {\lambda_1}\sum_{j=1}^{n} |\theta_j| + {\lambda_2}\sum_{j=1}^{n} \theta_j^2
$

It combines:
- feature selection (L1)
- stability (L2)

### Comparison
| Property                        | L1   | L2     |
| ------------------------------- | ---- | ------ |
| Feature Selection               | Yes  | No     |
| Sparse Model                    | Yes  | No     |
| Stable with correlated features | Less | More   |
| Coefficients become 0           | Yes  | Rarely |
| Most commonly used              | Less | Yes    |

Role of Lambda (λ)
$
λ
$
controls regularization strength.

#### Small Lambda
$
λ≈0
$

- almost no regularization
- may overfit

#### Large Lambda
$
λ→∞
$

- weights become tiny
- model underfits

### Bias-Variance Tradeoff
| Lambda | Bias | Variance |
| ------ | ---- | -------- |
| Small  | Low  | High     |
| Large  | High | Low      |
Regularization increases bias slightly to reduce variance significantly.

### Effect on Decision Boundary

Without regularization:

- highly irregular boundary
- sensitive to noise

With regularization:

- smoother boundary
- better generalization

### Regularization in Scikit-Learn
#### L2 (default)
```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    penalty='l2'
)
```

#### L1
```python
model = LogisticRegression(
    penalty='l1',
    solver='liblinear'
)
```

#### ElasticNet
```python
model = LogisticRegression(
    penalty='elasticnet',
    solver='saga',
    l1_ratio=0.5
)
```

### Important Parameter: C
$
C = 1 / \lambda
$

| C       | Regularization        |
| ------- | --------------------- |
| Large C | Weak regularization   |
| Small C | Strong regularization |
```python
model = LogisticRegression(C=0.01)
```

### When to Use Which?
#### Use L2 When
- normal datasets
- correlated features
- want stable model
- default choice

#### Use L1 When
- many irrelevant features
- high-dimensional data
- need feature selection

Examples:
- NLP
- text classification
- genomics

#### Use ElasticNet When
- many correlated features
- need both sparsity + stability


### Summary

| Concept        | Meaning                  |
| -------------- | ------------------------ |
| Regularization | Penalizing large weights |
| Goal           | Prevent overfitting      |
| L1             | Feature selection        |
| L2             | Weight shrinkage         |
| ElasticNet     | Combination              |
| Lambda ↑       | Simpler model            |
| C ↑            | Weaker regularization    |
