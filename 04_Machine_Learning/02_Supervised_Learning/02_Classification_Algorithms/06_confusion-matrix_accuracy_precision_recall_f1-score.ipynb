### Why Accuracy Alone is Dangerous

Suppose fraud detection dataset:

- 990 normal
- 10 fraud

Model predicts:
```
All = Normal
```
Accuracy:
```
990 / 1000 = 99% 
```

BUT:  
**It failed to detect ALL frauds**

This is why better metrics is needed.

### Confusion Matrix
A **confusion matrix** is a table that compares **actual vs predicted** values.
|                 | Predicted Positive  | Predicted Negative  |
| --------------- | ------------------- | ------------------- |
| Actual Positive | True Positive (TP)  | False Negative (FN) |
| Actual Negative | False Positive (FP) | True Negative (TN)  |

- **TP**: Correctly predicted positive
- **TN**: Correctly predicted negative
- **FP**: Incorrectly predicted positive (false alarm)
- **FN**: Missed positive case

Example

Fraud detection:
|               | Pred Fraud | Pred Normal |
| ------------- | ---------- | ----------- |
| Actual Fraud  | 8 (TP)     | 2 (FN)      |
| Actual Normal | 5 (FP)     | 985 (TN)    |


### Accuracy
Accuracy measures overall correctness:

Accuracy = TP + TN / TP + TN + FP + FN
	​
- Good when classes are balanced
- Can be misleading with imbalanced data

Example
(8+985)/1000=99.3%

Looks good — but still missed frauds.

### Precision
Precision answers: *Out of predicted positives, how many were correct?*

Precision = TP / TP + FP

- High precision -> few false positives
- Important when false alarms are costly

Example

8/(8+5) = 0.615

Use Case
- When false positives are costly

Example:
- Email spam filter
- Showing wrong ads

### Recall (Sensitivity)

Recall answers: *Out of actual positives, how many did we catch?*

Recall = TP / TP + FN
	
- High recall -> few missed cases
- Important in critical detection tasks (e.g., disease detection)

Example

8/(8+2) = 0.8

Use Case
- When missing positives is dangerous

Example:
- Fraud detection
- Disease detection

### F1 Score
F1 balances precision and recall:

F1=2 [(Precision . Recall) / (Precision + Recall)]

- Harmonic mean (not average)
- Useful when you need a trade-off between precision and recall

Example

F1 ≈ 0.69

When to Use
- Imbalanced datasets
- Need balance

### Trade-off
- Increase precision -> recall may drop
- Increase recall -> precision may drop

### Quick Intuition
- Accuracy -> overall performance
- Precision -> correctness of positive predictions
- Recall -> coverage of actual positives
- F1 -> balance between precision & recall

### When to Use What
| Metric    | Use Case              |
| --------- | --------------------- |
| Accuracy  | Balanced dataset      |
| Precision | Avoid false positives |
| Recall    | Avoid false negatives |
| F1        | Balance both          |

### Real Production Scenarios
##### Fraud Detection
Focus: **Recall**
Missing fraud = loss

##### Spam Detection
Focus: **Precision**
Wrong spam = bad UX

##### Medical Diagnosis
Focus: **Recall**
Missing disease = dangerous
