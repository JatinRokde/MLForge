Support Vector Machine (SVM) is a supervised machine learning algorithm used for:

1. Classification
1. Regression
1. Outlier/Anomaly Detection

SVM is one of the most mathematically elegant machine learning algorithms because it transforms a prediction problem into an optimization problem.

The primary objective is:
> Find the optimal decision boundary that maximizes separation between classes.

Unlike Logistic Regression, which focuses on probability estimation, SVM focuses on finding the best possible boundary with the largest safety margin.

### Inutition
Consider a binary classification problem:
| Age | Income | Purchased |
| --- | ------ | --------- |
| 25  | 30000  | No        |
| 45  | 80000  | Yes       |
| 35  | 50000  | No        |
| 55  | 120000 | Yes       |

A line can separate "Yes" and "No" customers.

Many separating lines may exist.

SVM asks:
> Which separating line is the best?

Answer:
> The line that leaves the maximum distance from both classes.

This distance is called the **Margin**.

### Hyperplane
The decision boundary in SVM is called a Hyperplane.
#### 2D
A line

$
w_1x_1 + w_2x_2 + b = 0
$

#### 3D
A plane

##### n-Dimensions
A hyperplane

$
w^Tx + b = 0
$

Where:
- $w$ = weights
- $x$ = features
- $b$ = bias

### Classification Rule
For a new sample:

$
f(x) = w^Tx + b
$

If 

$
f(x) > 0
$

Class = +1

If 

$
f(x) < 0
$

Class = -1

### Margin
The margin is the distance between:
- Hyperplane
- Closest positive samples
- Closest negative samples

Example:
```
Class A      Margin      Hyperplane      Margin      Class B

   ● ● ●       |             |              |          ▲ ▲ ▲
```

Small Margin:
- Tiny change in new data may cause misclassification.

Larger margin:
- Better generalization
- Less overfitting
- Better robustness

### Support Vectors
The points closest to the decision boundary are called Support Vectors.

Example:
```
● ● ● ●

    ●      <- support vector

------------

         ▲ <- support vector

▲ ▲ ▲ ▲
```

Only these points determine the hyperplane.

Removing far-away points often does not change the model.

This is why the algorithm is called: **Support Vector Machine**

### Why Support Vectors Matter

Assume 10,000 points exist.

Only 30 points lie near the boundary.

The optimal hyperplane is determined entirely by those 30 points.

This makes SVM mathematically efficient.

### Maximum Margin Principle
Among all possible separating hyperplanes:

Choose:
$$
Maximum Margin Hyperplane
$$

Mathematically:  
Margin

$$
\frac{2}{\lVert w \rVert}
$$

Maximizing the margin means:

$$
\max \frac{2}{\lVert w \rVert}
$$

Since the factor 2 is constant, this is equivalent to:

$$
\max \frac{1}{\lVert w \rVert}
$$

which is equivalent to:

$$
\min \lVert w \rVert
$$

and, for mathematical convenience:

$$
\min \frac{1}{2}\lVert w \rVert^2
$$

This becomes the optimization objective in a hard-margin SVM.

### Hard-Margin SVM
Used when the data is perfectly separable.

Constraint
For every training sample:

$$
y_i \left( w^T x_i + b \right) \geq 1,
\quad \forall i
$$

Optimization:

$$
\min_{w,b} \frac{1}{2}\lVert w \rVert^2
$$

subject to:

$$
y_i \left( w^T x_i + b \right) \geq 1,
\quad \forall i
$$

### Problem

Real-world datasets rarely have perfect separation.

Examples:
- Fraud Detection
- Customer Churn
- Medical Diagnosis

Usually classes overlap.

Hard Margin fails.

### Soft Margin SVM

Most production systems use Soft Margin SVM.

It allows some classification errors by introducing **slack variables**:

$$
\xi_i
$$

The constraint becomes:

$$
y_i \left( w^T x_i + b \right) \geq 1 - \xi_i
$$

where

$$
\xi_i \geq 0
$$

#### Optimization Objective

$$
\min_{w,b,\xi}
\frac{1}{2}\lVert w \rVert^2
+
C \sum_{i=1}^{n} \xi_i
$$

subject to:

$$
y_i \left( w^T x_i + b \right) \geq 1 - \xi_i,
\quad \forall i
$$

$$
\xi_i \geq 0,
\quad \forall i
$$

### Regularization Parameter (C)
One of the most important hyperparameters.
$$
C
$$
controls:

##### Margin vs Error Tradeoff

**Large C**
```
C = 100
```
Punishes errors heavily.

Characteristics:

- Narrow margin
- Lower bias
- Higher variance
- Overfitting risk

**Small C**
```
C = 0.01
```
Allows more violations.

Characteristics:

- Wider margin
- Higher bias
- Lower variance
- Better generalization

Example:
| C Value | Behavior              |
| ------- | --------------------- |
| 0.01    | Very soft margin      |
| 0.1     | Soft margin           |
| 1       | Default               |
| 10      | Strict classification |
| 100     | Very strict           |

#### Geometric Interpretation
Large C:
```
Trying to classify every point correctly
```
Small C:
```
Allows few errors
Focuses on wider margin
```

Consider:
```
Inner Circle = Class A
Outer Ring = Class B
```
No straight line can separate them.

Linear SVM fails.

### Feature Transformation
Map data into a higher dimension.

Example:  
Original feature:
$$
x
$$
Transform:
$$
(x,x^2)
$$
A non-linear problem can become linearly separable.

### Kernel Trick
The **Kernel Trick** is a technique used in **Support Vector Machines (SVMs)** to handle data that is **not linearly separable** in its original feature space.

Instead of explicitly transforming the data into a higher-dimensional space, the kernel trick computes the **inner products** in that higher-dimensional space directly, making the computation much more efficient.

#### The need for Kernel Trick
Consider a dataset like this:
```
○ ○ ○
○ ● ○
○ ○ ○
```
The classes cannot be separated by a straight line in 2D.

However, if the data is transformed into a higher-dimensional space, the classes may become linearly separable.

For example:
$$
(x_1, x_2) → (x_1, x_2, x_1^2, x_2^2)
$$

In the new space, a hyperplane can separate the classes.

#### The Key Idea
Instead of computing the transformation
$$
ϕ(x)
$$
explicitly, SVM uses a **kernel function**:
$$
K(x_i, x_j) = ϕ(x_i)^T ϕ(x_j)
$$
which directly computes the dot product in the transformed feature space.

This avoids the expensive computation of creating high-dimensional features.

#### Common Kernels
##### 1. Linear Kernel
No Transformation is applied.
$$
K(x_i, x_j) = x_i^Tx_j
$$
Used when the data is approximately linearly separable.

##### 2. Polynomial Kernel
Creates polynomial decision boundaries.
$$
K(x_i, x_j) = (x_i^Tx_j + c)^d
$$

where:
- $d$ = degree of the polynomial
- $c$ = constant term

##### 3. RBF (Gaussian) Kernel
The most commonly used kernel in practice.
$$
k(x_i, x_j) = exp(-γ||x_i - x_j||^2)
$$

where:
- $γ$ controls the influence of each training point
- This kernel can model highly non-linear relationships.

##### 4. Sigmoid Kernel
Inspired by neural networks.
$$
k(x_i, x_j) = tanh(αx_i^Tx_j + c)
$$

##### Mathematical Intuition
Without kernels:
$$
SVM works on ϕ(x)
$$

and computes
$$
ϕ(x_i)^Tϕ(x_j) 
$$	​
directly.

With the kernel trick:
$$
ϕ(x_i)^Tϕ(x_j) = K(x_i, x_j)
$$

so we never need to compute $ϕ(x)$ explicitly.

### Gamma Parameter (γ)

Used in RBF kernel.

Controls influence of a point.

**Small Gamma**
- Far points influence each other.
- Smooth boundary.
- Higher bias.

Smooth Curve

**Large Gamma**
- Only nearby points matter.
- Highly complex boundary.
- Overfitting risk.

Highly Wiggly Curve

### Understanding C and Gamma Together
Most tuning effort focuses here.
| Gamma | C    | Result       |
| ----- | ---- | ------------ |
| Low   | Low  | Underfit     |
| Low   | High | Moderate     |
| High  | Low  | Complex      |
| High  | High | Overfit Risk |

### Choosing Kernels
| Dataset                      | Kernel     |
| ---------------------------- | ---------- |
| Text                         | Linear     |
| High Dimensional Sparse Data | Linear     |
| Nonlinear Medium Dataset     | RBF        |
| Feature Interactions Needed  | Polynomial |
| Complex Boundaries           | RBF        |

````python
import os
import joblib
import pandas as pd

from scipy.sparse import hstack
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score
)
````

````python
ARTIFACT_DIR = "project-support_vector_machine_classifier"

os.makedirs(
    ARTIFACT_DIR,
    exist_ok=True
)
````

````python
df = pd.read_csv(
    "./data/incident_severity_dataset.csv"
)

print(f"Dataset Shape: {df.shape}")
df.head()
````

````python
# Features
TEXT_COLUMN = "incident_title"

NUMERIC_COLUMNS = [
    "affected_users",
    "cpu_usage",
    "error_rate"
]

TARGET_COLUMN = "severity"
````

````python
# Label Encoding
label_encoder = LabelEncoder()

y = label_encoder.fit_transform(
    df[TARGET_COLUMN]
)
````

````python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(
    df,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
````

````python
# TF-IDF
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1, 2),
    stop_words="english"
)

X_train_text = tfidf.fit_transform(
    X_train[TEXT_COLUMN]
)

X_test_text = tfidf.transform(
    X_test[TEXT_COLUMN]
)
````

````python
# Numerical Features
scaler = StandardScaler()

X_train_num = scaler.fit_transform(
    X_train[NUMERIC_COLUMNS]
)

X_test_num = scaler.transform(
    X_test[NUMERIC_COLUMNS]
)
````

````python
# Combine Features
X_train_final = hstack(
    [
        X_train_text,
        X_train_num
    ]
)

X_test_final = hstack(
    [
        X_test_text,
        X_test_num
    ]
)

print(
    f"Final Feature Size: {X_train_final.shape}"
)
````

````python
# Model
model = LinearSVC(
    C=1.0,
    random_state=42,
    max_iter=10000
)
````

````python
# Training
model.fit(
    X_train_final,
    y_train
)
````

````python
# Prediction
predictions = model.predict(
    X_test_final
)
````

````python
# Evaluation
accuracy = accuracy_score(
    y_test,
    predictions
)

print("\nAccuracy")
print(accuracy)

print("\nClassification Report")

print(
    classification_report(
        y_test,
        predictions,
        target_names=label_encoder.classes_
    )
)

print("\nConfusion Matrix")

print(
    confusion_matrix(
        y_test,
        predictions
    )
)
````

````python
joblib.dump(
    model,
    f"{ARTIFACT_DIR}/svm_model.pkl"
)

joblib.dump(
    tfidf,
    f"{ARTIFACT_DIR}/tfidf_vectorizer.pkl"
)

joblib.dump(
    scaler,
    f"{ARTIFACT_DIR}/scaler.pkl"
)

joblib.dump(
    label_encoder,
    f"{ARTIFACT_DIR}/label_encoder.pkl"
)
````
