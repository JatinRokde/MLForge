### Why Ensembles Exist
Single models often suffer from:
| Problem       | Example       |
| ------------- | ------------- |
| High variance | Decision Tree |
| High bias     | Weak learner  |

Two major strategies emerged:

##### Bagging → Reduce Variance
##### Boosting → Reduce Bias

### Bagging (Bootstrap Aggregating)
#### Core Idea
Train multiple models:
- Independently
- In parallel
- On random bootstrap samples

Then aggregate outputs.

#### Bagging Workflow
```
Dataset
 → Bootstrap Sampling
 → Train Multiple Models Independently
 → Voting / Averaging
 → Final Prediction
```

### Bootstrap Sampling
Sampling with replacement.

Example:

Original:
```
[A,B,C,D]
```
Bootstrap sample:
```
[A,C,C,D]
```
Different models see different data.

### Why Bagging Works
Decision trees are high variance.

Small data change:
- very different tree.

Bagging averages predictions:
- Reduces variance
- Stabilizes model

### Random Forest = Bagging + Random Features
Random Forest extends bagging by adding:
- Random feature selection.

This decorrelates trees further.

### Bias–Variance Effect of Bagging
| Property | Effect                |
| -------- | --------------------- |
| Bias     | Similar               |
| Variance | Reduced significantly |

Advantages of Bagging
- Parallelizable
- Stable
- Lower overfitting
- Robust to noise

### Limitations of Bagging
- Large models
- Doesn’t reduce bias much
- Less interpretable

### Boosting
#### Core Idea
Train models:
- Sequentially
- Each new model focuses on previous mistakes

#### Boosting Workflow
```
Model1
 → Errors identified
 → Model2 focuses on errors
 → Model3 focuses further
 → Combine all models
```

#### Intuition
Suppose teacher corrects student repeatedly.

Each iteration improves weaknesses.

That’s boosting.

#### AdaBoost Intuition
Misclassified points get higher weight.

Future learners focus more on hard samples.

#### Gradient Boosting Intuition
New model learns residual errors.

Example:
```
Residual = Actual - Prediction
```
Each stage corrects previous mistakes.

### Why Boosting Works
Weak learners individually poor.

Sequential correction builds strong learner.

### Bias–Variance Effect of Boosting
| Property | Effect                |
| -------- | --------------------- |
| Bias     | Reduced significantly |
| Variance | Can increase          |

Boosting can overfit if uncontrolled.

### Bagging vs Boosting - Core Difference
#### Bagging
```
Independent Learners
```

#### Boosting
```
Sequential Learners
```

### Parallelization
| Method   | Parallelizable?        |
| -------- | ---------------------- |
| Bagging  | Yes                    |
| Boosting | No (mostly sequential) |

### Overfitting Behavior
#### Bagging
Usually robust.

#### Boosting
Can overfit if:
- Too many estimators
- Deep trees
- High learning rate

### Performance Comparison
| Feature            | Bagging  | Boosting   |
| ------------------ | -------- | ---------- |
| Variance Reduction | Strong   | Moderate   |
| Bias Reduction     | Weak     | Strong     |
| Training Style     | Parallel | Sequential |
| Speed              | Faster   | Slower     |
| Overfitting Risk   | Lower    | Higher     |
| Interpretability   | Moderate | Lower      |

#### Use Bagging / Random Forest When
- Need stable baseline
- Fast training
- Low tuning effort
- Medium complexity data

#### Use Boosting When
- Need highest accuracy
- Complex tabular data
- Competition-grade performance

### Why Gradient Boosting Became So Dominant
Because it:
- Handles non-linearity
- Handles interactions
- Optimizes residuals aggressively
- Achieves state-of-the-art tabular performance
