### Why Gradient Boosting Was Not Enough
Traditional Gradient Boosting works well.

But it has limitations:
- Slow training
- High memory usage
- Overfitting risk
- Weak handling of missing values
- Limited parallelization

XGBoost was designed to solve these issues.

### What is XGBoost?
XGBoost stands for: **Extreme Gradient Boosting**

It is:
> A highly optimized implementation of Gradient Boosting with regularization, parallelization, and system-level optimizations.

Important:
```
Gradient Boosting = Concept
XGBoost = Optimized Engineering Implementation
```

High-Level Architecture

Traditional Gradient Boosting:
```
Tree1
 → Residuals
Tree2
 → Residuals
Tree3
```
XGBoost:
```
Tree1
 → Residuals
 + Regularization
 + Parallelized split finding
 + Missing value handling
 + Pruning
Tree2
```
Same core idea, much better implementation.

### Core Learning Formula
Gradient Boosting:
$$
F(x) = \sum_{m=1}^{M} f_m(x)
$$

XGBoost also follows:
$$
\hat{y}_i = \sum_{k=1}^{K} f_k(x_i)
$$

Where:
- $f_k$ = decision tree
- $K$ = number of trees

### The Big Innovation — Regularized Objective
Traditional Gradient Boosting:
$$
Loss
$$
only.

XGBoost:
$$
Objective = Loss + Regularization
$$

Full Objective:
$$
\text{Objective} = \sum_{i=1}^{n} L(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)
$$

where:
$$
\Omega(f) = \gamma T + \frac{1}{2}\lambda \|w\|^2
$$

### Understanding Regularization Terms
#### T
Number of leaves.

More leaves:
```
More complexity
```
Penalty applied.

#### λ (lambda)
L2 regularization on leaf weights.

Reduces overfitting.

#### γ (gamma)
Minimum gain required to split.

Small gain?
```
No split
```
Acts like pruning.

### Why XGBoost Generalizes Better
Unlike basic Gradient Boosting:

XGBoost explicitly controls:
- Tree complexity
- Leaf weights
- Split quality

### Second-Order Optimization
Traditional Gradient Boosting uses:
$$
Gradient
$$
only.

XGBoost uses:
$$
Gradient + Hessian
$$
Where:
- Hessian = second derivative.

This gives:
- Better optimization
- Faster convergence

### Tree Growth Strategy
For each possible split:

XGBoost computes:
$$
Gain
$$
Gain measures improvement.

Split only happens if:
$$
Gain>γ
$$
This prevents useless branches.

### Missing Value Handling
Huge production advantage.

Example:
```
Salary = Missing
```
Many algorithms require imputation.

XGBoost learns:
```
Missing → Left
```
or
```
Missing → Right
```
automatically.

### Sparsity Awareness
Works very well with:
- Sparse features
- One-hot encoding
- Text-derived features

### Parallelization
Boosting is sequential.

But split evaluation can be parallelized.

XGBoost parallelizes:
```
Best split search
```
Result:

Much faster training.

Contribution of each tree.
```
learning_rate=0.05
```

#### max_depth
Tree depth.
```
max_depth=6
```

Minimum samples required.

Controls overfitting.

#### subsample
Row sampling.
```
subsample=0.8
```

#### colsample_bytree
Feature sampling.
```
colsample_bytree=0.8
```

#### gamma
Minimum split gain.

#### reg_alpha
L1 regularization.

L2 regularization.

### XGBoost vs Random Forest
| Feature            | Random Forest | XGBoost        |
| ------------------ | ------------- | -------------- |
| Training           | Parallel      | Sequential     |
| Bias Reduction     | Moderate      | Strong         |
| Variance Reduction | Strong        | Strong         |
| Accuracy           | High          | Usually Higher |
| Tuning Complexity  | Moderate      | Higher         |

### XGBoost vs Gradient Boosting
| Feature        | Gradient Boosting | XGBoost              |
| -------------- | ----------------- | -------------------- |
| Regularization | Basic             | Advanced             |
| Speed          | Moderate          | Faster               |
| Missing Values | Manual            | Native               |
| Optimization   | First-order       | First + Second-order |
| Scalability    | Moderate          | High                 |
