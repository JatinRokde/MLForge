Random Forest became popular because it reduced variance.

Gradient Boosting became popular because it aggressively reduces bias.

For tabular datasets, Gradient Boosting and its descendants have dominated production ML for years.

### The Main Idea
AdaBoost says:
> Focus more on difficult samples.

Gradient Boosting says:
> Learn the mistakes (residuals) made by previous models.

This is a fundamentally different approach.

### Intuition
Suppose actual values:
| Customer | Actual Revenue |
| -------- | -------------- |
| A        | 100            |
| B        | 200            |
| C        | 300            |

First Model

Predictions:
| Customer | Prediction |
| -------- | ---------- |
| A        | 120        |
| B        | 180        |
| C        | 250        |

Errors:
| Customer | Residual |
| -------- | -------- |
| A        | -20      |
| B        | +20      |
| C        | +50      |

Where:
$$
Residual = Actual - Prediction
$$

Instead of predicting revenue again,

the next model predicts:
```
[-20, +20, +50]
```
the errors.

### Core Principle
Each new model learns:
$$
Residuals
$$
of the previous model.

### Mathematical View
Final prediction:
$$
F(x)=f_1(x) = f_2(x) = f_3(x) + ...
$$	​

where:
- $f_1$ = first tree
- $f_2$ = residual correction tree
- $f_3$ = next correction tree  
etc.

### Step-by-Step Algorithm
#### Step 1
Train first tree.

Prediction:
$$
\hat{y}_1
$$

#### Step 2
Compute Residuals
$$
r_1 = y - \hat{y}_1
$$

#### Step 3
Train second tree on residuals.
$$
f_2(x)
$$

#### Step 4
Update prediction.
$$
\hat{y}_2 = \hat{y}_1 + f_2(x)
$$

#### Step 5
Compute new residuals.

Repeat.

### Why It Is Called "Gradient" Boosting
The residuals are actually:
$$
r_m = -\nabla L
$$
More precisely, for the $m$-th iteration:
$$
r_{im} = -\left[\frac{\partial L(y_i, F(x_i))}{\partial F(x_i)}\right]_{F(x)=F_{m-1}(x)}
$$

the negative gradient of the loss function.

Thus:

Gradient Boosting performs gradient descent in function space.

### Function Space vs Parameter Space
Linear Regression:

Gradient descent updates weights.
$$
w:=w−η∇L
$$

Gradient Boosting:

Adds new functions (trees).
$$
F_m(x) = F_{m-1}(x) + \eta h_m(x)
$$
where:
- $h_m$ = new tree
- $\eta$ = learning rate

### Learning Rate
Very important.
$$
F_m(x) = F_{m-1}(x) + \eta h_m(x)
$$

#### Small Learning Rate
Example:
```
0.01
```
- Slower learning
- Better generalization

#### Large Learning Rate
Example:
```
1.0
```
- Fast learning
- Overfitting risk

### Why Small Trees Are Used
Gradient Boosting usually uses:
```
max_depth = 3–8
```
Not huge trees.

Reason:
- Each tree is only correcting errors.

### Example Walkthrough
Actual:
```
[100, 200, 300]
```
Tree 1:
```
[120, 180, 250]
```
Residuals:
```
[-20, 20, 50]
```
Tree 2 learns:
```
[-15, 18, 40]
```
Updated prediction:
```
[105, 198, 290]
```
Much closer.

### Loss Function Flexibility
Unlike AdaBoost,

Gradient Boosting can optimize many losses.

##### Regression
MSE
$$
L(y, \hat{y}) = (y - \hat{y})^2
$$
For N samples:
$$
L = \frac{1}{N}\sum_{i=1}^{N}(y_i - \hat{y}_i)^2
$$

##### Classification
Log loss
$$
L(y, p) = -y \log(p) - (1-y)\log(1-p)
$$

A simplified form often shown for the positive class is:
$$
L(y, p) = -y \log(p)
$$

This flexibility is a huge advantage.

#### max_depth
Complexity of individual trees.

#### subsample
Fraction of training data. Used for regularization.

### Gradient Boosting vs Random Forest
| Feature  | Random Forest   | Gradient Boosting |
| -------- | --------------- | ----------------- |
| Training | Parallel        | Sequential        |
| Goal     | Reduce variance | Reduce bias       |
| Trees    | Independent     | Error-correcting  |
| Speed    | Faster          | Slower            |
| Accuracy | Strong          | Often higher      |

### Why Gradient Boosting Often Wins
Each tree learns:
```
What previous trees missed
```
Instead of averaging independent trees.

This makes learning much more targeted.

### Overfitting Risks
Gradient Boosting can overfit.

Especially when:
```
Too many trees
Deep trees
High learning rate
```

### Regularization Techniques
#### Shrinkage
Small learning rate.

#### Early Stopping
Stop when validation score stops improving.

#### Subsampling
Use only subset of rows.

#### Tree Constraints
Limit depth.
