### Real ML Pipeline
Overall picture:
```
Dataset
 ↓
Model
 ↓
Prediction
```
Reality:
```
Business Problem
 ↓
Data Collection
 ↓
EDA
 ↓
Feature Engineering
 ↓
Baseline Model
 ↓
Evaluation
 ↓
Model Selection
 ↓
Hyperparameter Tuning
 ↓
Deployment
 ↓
Monitoring
 ↓
Retraining
```
The model itself is only a small part.

### Step 1 — Understand the Business Problem
Never start with:
```python
from sklearn...
```
Start with:
```
What decision is being automated?
```

#### Example
Problem:
```
Predict customer churn
```
Business question:
```
Which customers are likely to leave?
```

#### Example
Problem:
```
Predict house prices
```
Business question:
```
What should be the expected listing price?
```

### Step 2 — Identify Problem Type
#### Regression
Output:
```
Continuous value
```
Examples:
- House price
- Revenue
- Temperature

#### Classification
Output:
```
Discrete class
```
Examples:
- Fraud / Not Fraud
- Churn / No Churn

#### Ranking
Examples:
- Search results
- Recommendations

#### Clustering
Examples:
- Customer segmentation

### Step 3 — Understand Dataset Characteristics
Before choosing model:

Ask:
#### Dataset Size
```
1000 rows?
100k rows?
10M rows?
```

#### Feature Types
```
Numerical?
Categorical?
Text?
Images?
```

#### Missing Values
#### Class Imbalance
#### Feature Count
```
10?
100?
10,000?
```

### Step 4 — Build Baseline First
One of the most important production principles.

Never start with:
```
CatBoost
XGBoost
Neural Networks
```
Start simple.

#### Classification Baseline
```
LogisticRegression()
```

#### Regression Baseline
```
LinearRegression()
```
Why?

Baseline provides:
```
Minimum acceptable benchmark
```

### Model Selection Strategy
#### Small Dataset (<10k rows)
Classification:
```
Logistic Regression
SVM
Random Forest
```

Regression:
```
Linear Regression
Ridge
Random Forest
```

#### Medium Dataset (10k–1M rows)
Classification:
```
Random Forest
XGBoost
CatBoost
```

Regression:
```
Gradient Boosting
XGBoost
LightGBM
```

#### Large Dataset (>1M rows)
Often:
```
LightGBM
Distributed XGBoost
Deep Learning
```
depending on problem.

### Choosing Models by Feature Type
#### Mostly Numerical Features
Preferred:
```
XGBoost
LightGBM
```

#### Many Categorical Features
Preferred:
```
CatBoost
```

#### High-Dimensional Sparse Features
Examples:
- TF-IDF
- Bag-of-Words

Preferred:
```
Linear SVM
Logistic Regression
```

#### Images
Preferred:
```
CNNs
Vision Transformers
```

### Evaluation Strategy
#### Classification
Evaluate:
```
Accuracy
Precision
Recall
F1
ROC-AUC
```

#### Imbalanced Classification
Focus on:
```
Recall
Precision
PR-AUC
```

#### Regression
Evaluate:
```
MAE
RMSE
R²
```

### Cross Validation Strategy
#### Standard Data
```
KFold
```
or
```
StratifiedKFold
```

#### Time Series
Never use random split.

Use:
```
TimeSeriesSplit
```

### Hyperparameter Tuning Strategy
#### Random Search
Preferred starting point.
```
RandomizedSearchCV
```

#### Grid Search
Smaller search space.
```
GridSearchCV
```

#### Bayesian Optimization
Advanced production tuning.

Examples:
- Optuna
- Hyperopt

### Model Comparison Framework
| Model    | CV Score | Latency  | Memory |
| -------- | -------- | -------- | ------ |
| Logistic | 0.82     | Very Low | Low    |
| RF       | 0.88     | Medium   | Medium |
| XGB      | 0.91     | Medium   | Medium |
| CatBoost | 0.92     | Medium   | Medium |

### Accuracy Is Not Everything
Suppose:

Model A:
```
Accuracy = 95%
Latency = 500ms
```
Model B:
```
Accuracy = 94%
Latency = 10ms
```
For real-time systems:
```
Model B may be better
```

### Model Selection Cheat Sheet
| Scenario                  | Preferred Model              |
| ------------------------- | ---------------------------- |
| Quick baseline            | Logistic / Linear Regression |
| Small tabular dataset     | Random Forest                |
| General tabular data      | XGBoost                      |
| Large tabular data        | LightGBM                     |
| Many categorical features | CatBoost                     |
| Sparse text features      | Linear SVM                   |
| Maximum interpretability  | Logistic Regression          |
| Fast inference            | Logistic Regression          |
| Highest tabular accuracy  | XGBoost / CatBoost           |
