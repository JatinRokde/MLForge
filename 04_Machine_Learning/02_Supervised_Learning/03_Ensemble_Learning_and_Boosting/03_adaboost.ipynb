AdaBoost introduced the revolutionary idea:
> Focus more on the mistakes.

This concept became the foundation for modern boosting systems.

### What is AdaBoost?
AdaBoost = Adaptive Boosting.

It combines:
- Multiple weak learners
- Sequential learning
- Weighted correction of mistakes

Usually uses:

#### Decision Stumps
A stump = tree with depth = 1.

Very weak learner.

### Core Intuition
Suppose model misclassifies some points.

AdaBoost says:
```
Increase importance of difficult points
```
Next learner focuses more on them.

### High-Level Workflow
```
Train weak learner
 → Identify mistakes
 → Increase weights of mistakes
 → Train next learner
 → Combine learners
```

#### Why “Adaptive”?
Because each learner adapts based on previous errors.

### Initial Sample Weights
Initially:

All samples equally important.

For N samples:
$$
w_i = \frac{1}{N}
$$​

### Step-by-Step AdaBoost Algorithm
#### Step 1 — Train Weak Learner
Train stump on weighted data.

#### Step 2 — Compute Error
Weighted error:
$$
\epsilon = \sum_{i=1}^{N} w_i \cdot I\!\left(y_i \neq \hat{y}_i\right)
$$
where:
- I = indicator function

#### Step 3 — Compute Learner Weight
$$
\alpha = \frac{1}{2}\ln\!\left(\frac{1-\epsilon}{\epsilon}\right)
$$

##### Interpretation
Small error:
- Large alpha
- Learner trusted more

Large error:
- Small alpha
- Learner trusted less

#### Step 4 — Update Sample Weights
Misclassified samples:
- Weight increased

Correct samples:
- Weight decreased

Formula:
$$
w_i = w_i \exp\!\left(-\alpha\, y_i\, h_i(x)\right)
$$

Then normalize weights.

#### Step 5 — Repeat
Train next learner on updated weights.

### Final Prediction
Weighted voting:
$$
F(x) = \operatorname{sign}\!\left(\sum_{m=1}^{M} \alpha_m h_m(x)\right)
$$

where:
- $h_m(x)$ = prediction of the $m$-th weak learner
- $\alpha_m$ = importance (weight) of the $m$-th learner

### Example Intuition
Suppose:

#### First Stump
Misclassifies:
```
Points A and B
```
Weights of A and B increase.

#### Second Stump
Focuses more on A and B.

#### Third Stump
Corrects remaining mistakes.

### Why AdaBoost Works
Weak learners individually:
```
Slightly better than random
```
Sequential correction creates:
```
Strong classifier
```

### AdaBoost and Bias–Variance
AdaBoost primarily:
- Reduces bias

Can also reduce variance somewhat.

### Decision Boundary Behavior
AdaBoost creates:

- Non-linear boundaries
- Sharp corrections around difficult regions

````python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X = [[2,3],
     [3,4],
     [7,8],
     [8,9]]

y = [0,0,1,1]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25
)

base_model = DecisionTreeClassifier(max_depth=1)

model = AdaBoostClassifier(
    estimator=base_model,
    n_estimators=50,
    learning_rate=1.0,
    random_state=42
)

model.fit(X_train, y_train)

pred = model.predict(X_test)

print(pred)
````

### Important Hyperparameters
#### n_estimators
Number of weak learners.

More learners:
- Lower bias
- More overfitting risk

#### learning_rate
Controls contribution of each learner.

#### estimator
Base weak learner.

Usually:
```
Decision stump
```

### AdaBoost vs Random Forest
| Feature              | Random Forest   | AdaBoost       |
| -------------------- | --------------- | -------------- |
| Training             | Parallel        | Sequential     |
| Goal                 | Reduce variance | Reduce bias    |
| Trees                | Deep-ish        | Usually stumps |
| Overfitting          | Lower           | Higher risk    |
| Sensitivity to noise | Lower           | Higher         |

### Advantages
- Simple boosting concept
- Works well on structured data
- Less tuning than advanced boosting

### Limitations
- Sensitive to outliers
- Sensitive to noisy labels
- Sequential training slower

Because difficult points get increasing weight.

Outliers may dominate learning.

### Why AdaBoost Historically Important
AdaBoost introduced:
- Sequential boosting
- Weighted error correction
- Ensemble reweighting

Modern boosting evolved from these ideas.

### AdaBoost vs Gradient Boosting
AdaBoost:
- Reweights samples

Gradient Boosting:
- Fits residual errors directly

Gradient boosting is generally more powerful.
