### The Fundamental Learning Problem
Suppose the real relationship is:
$$
y=f(x)
$$
But:
- True function is unknown.
- Only samples are available.

The ML model tries to learn:
$$
\hat f(x)
$$
from finite data.

### Sources of Prediction Error
Prediction error comes from three components:
$$
Error = Bias^2 + Variance + Irreducible \ Error
$$
This is one of the most important equations in ML.

### Irreducible Error
Some noise cannot be learned.

Example:

House price influenced by:

- Seller emotions
- Negotiation skill
- Future market conditions

Not present in data.

No model can learn this perfectly.

### What is Bias?
Bias measures:
> How far the average model prediction is from reality.

#### High Bias
Model assumptions too simple.

Example:

Trying to fit:
$$
y=x^2
$$
using:
$$
y=mx+b
$$
Impossible.

Result
```
Underfitting
```

### Visual Intuition of High Bias
True relationship:
```
Curved
```
Model:
```
Straight line
```
Systematic error everywhere.

### Examples of High-Bias Models
- Linear Regression on complex data
- Logistic Regression on highly nonlinear data
- Decision Tree with depth=1

### What is Variance?
Variance measures:
> How much the model changes when training data changes.

Example:

Train on Dataset A:
```
Accuracy = 95%
```
Train on Dataset B:
```
Accuracy = 70%
```
Huge variation.

High variance.

### Visual Intuition of High Variance
Model follows:
- Signal

and

- Noise

Result:
```
Excellent training performance
Poor test performance
```

### Examples of High Variance Models
- Deep Decision Trees
- Large Neural Networks
- High-degree Polynomial Regression

### Underfitting vs Overfitting
#### Underfitting

High bias.

Low variance.
```
Train Error = High
Test Error = High
```

#### Overfitting

Low bias.

High variance.
```
Train Error = Very Low
Test Error = High
```

Goal:
```
Low Bias
Low Variance
```
This is where ensembles help.

### Bias–Variance Decomposition
For regression:
$$
E[(y - \hat y)^2] = Bias^2 + Variance + Noise
$$

Interpretation:

Prediction error comes from:
- Wrong assumptions
- Instability
- Random noise

### Why Decision Trees Overfit
Decision Trees:
- Very flexible

Can create:
```
One leaf per sample
```
Result:
```
Bias → Very Low
Variance → Very High
```

### Why Random Forest Works
Recall:

Random Forest:
```
Many trees
↓
Average predictions
```

#### Mathematical Intuition
Suppose:
$$
M
$$
trees.

Prediction:
$$
\hat y = \frac{1}{M}\sum_{i=1}^{M} T_i
$$

Averaging reduces variance.

#### Approximation
If trees independent:
$$
Variance_{RF} = \frac{\sigma ^2}{M}	​
$$

Result:
- Similar bias
- Much lower variance

### Why Random Feature Sampling Matters
If all trees identical:
$$
Variance
$$
not reduced much.

Random features create diversity.

This is why:
$$
Bagging alone < Random Forest
$$

### Why Boosting Works
Boosting attacks: **Bias**

instead of variance.

#### Idea
Each tree learns:
```
What previous trees missed
```
Prediction:
$$
F(x) = f_1(x) + f_2(x) + f_3(x)
$$

Bias decreases continuously.

### Boosting Bias Reduction
Imagine:

Tree 1 captures:
```
70% pattern
```
Tree 2 captures:
```
15% remaining pattern
```
Tree 3 captures:
```
10% remaining pattern
```
Total model becomes highly accurate.

### Why Boosting Can Overfit
As more trees added:

Bias decreases.

But eventually:

Variance starts increasing.

Result:
```
Training Error ↓
Validation Error ↑
```
Classic overfitting.

### Why XGBoost Works So Well
XGBoost balances:
#### Bias Reduction
via boosting.

#### Variance Control
via regularization.

Objective:
$$
Loss + Regularization
$$
This simultaneously controls:
- Bias
- Variance

### LightGBM Perspective
LightGBM:

Aggressive leaf-wise growth.

Advantage:
```
Lower bias
```
Risk:
```
Higher variance
```
Hence:
```python
num_leaves
max_depth
```
must be tuned carefully.

### CatBoost Perspective
CatBoost improves:
```
Bias reduction
```
while reducing:
```
Prediction shift
```
and leakage-induced variance.

### Learning Curves and Bias–Variance
#### High Bias
Training Error:
```
High
```
Validation Error:
```
High
```
Both close together.

#### High Variance
Training Error:
```
Very Low
```
Validation Error:
```
High
```
Large gap.

### How to Fix High Bias
Increase model capacity.

Examples:
- More trees
- Greater depth
- Better features
- Nonlinear models
- Boosting

### How to Fix High Variance
Examples:
- More training data
- Regularization
- Bagging
- Feature selection
- Ea
