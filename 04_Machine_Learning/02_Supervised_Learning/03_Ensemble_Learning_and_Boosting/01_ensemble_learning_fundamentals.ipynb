### What is Ensemble Learning?
Ensemble Learning means:
> Combining multiple models to create a stronger model.

Core idea:
```
Wisdom of crowds
```
One weak model may fail.

Many models together become robust.

### Why Ensembles Work
Single model:
- May overfit
- May underfit
- Sensitive to noise

Combining models:
- Reduces variance
- Reduces bias
- Improves generalization

### Types of Ensemble Learning
#### 1. Bagging (Bootstrap Aggregating)
**Idea**: Train multiple models independently on different random subsets of the training data and combine their predictions.

**Process**
1. Create multiple bootstrap samples from the dataset.
1. Train a separate model on each sample.
1. Aggregate predictions:
   - Classification: Majority voting
   - Regression: Averaging

**Mathematical Representation**  
For regression:

$
\hat{y} = \frac{1}{n}\sum_{i=1}^{n} T_i
$

For classification:

$
\hat{y} = \operatorname{Mode}(T_1, T_2, \ldots, T_n)
$

where $T_i$ is the prediction of the $ith$ model.

Examples
- Random Forest
- Bagging Classifier
- Bagging Regressor

Advantages
- Reduces variance
- Less prone to overfitting
- Easy to parallelize

#### 2. Boosting
**Idea**: Train models sequentially, where each new model focuses on correcting the errors of previous models.

**Process**
1. Train a weak learner.
1. Identify mistakes.
1. Increase focus on misclassified samples.
1. Train the next learner.
1. Combine all learners into a strong model.

**Mathematical Representation**  
Boosting combines weak learners as:

$
F(x) = \sum_{m=1}^{M} \alpha_m h_m(x)
$

where:
- $(h_m(x)) = (m)$-th weak learner
- $(\alpha_m)$ = weight assigned to that learner

Examples
- AdaBoost
- Gradient Boosting
- XGBoost
- LightGBM
- CatBoost

Advantages
- High predictive accuracy
- Handles complex patterns
- Often wins machine learning competitions

#### 3. Stacking (Stacked Generalization)
**Idea**: Combine predictions from multiple different models using a meta-model.

**Process**
1. Train several base models.
1. Collect their predictions.
1. Use those predictions as inputs to another model.
1. The meta-model makes the final prediction.

**Mathematical Representation**  
Base models:

$
h_1(x), h_2(x), \ldots, h_n(x)
$

Meta-model:

$
\hat{y} = g\bigl(h_1(x), h_2(x), \ldots, h_n(x)\bigr)
$

where:
- $h_i$ = base learners
- $g$ = meta-learner

Examples
- Random Forest + SVM + XGBoost → Logistic Regression
- Multiple neural networks → Meta-classifier

Advantages
- Leverages strengths of different algorithms
- Often achieves higher accuracy than a single model

##### Comparison
| Technique    | Training Style | Goal              | Example          |
| ------------ | -------------- | ----------------- | ---------------- |
| **Bagging**  | Parallel       | Reduce Variance   | Random Forest    |
| **Boosting** | Sequential     | Reduce Bias       | XGBoost          |
| **Stacking** | Layered        | Combine Strengths | Stacked Ensemble |

**Quick Summary** 
- **Bagging** -> Many independent models trained in parallel.
- **Boosting** -> Models trained sequentially to fix previous mistakes.
- **Stacking** -> Multiple models whose predictions are combined by a meta-model.

### Weak Learners vs Strong Learners
Weak learner:

Slightly better than random.

Example:
```
Decision stump
(depth=1 tree)
```

Strong learner:

High-performing model.

Boosting converts weak -> strong.

### Ensemble Intuition Using Variance
Suppose models have noisy predictions:
| Model | Prediction |
| ----- | ---------- |
| 1     | 90         |
| 2     | 110        |
| 3     | 100        |

Average:
```
100
```
Noise cancels out.

### Bias vs Variance Perspective
#### Bagging
- Reduces variance
- Parallel training

Best for:
```
High-variance models
```
Example:

Decision Trees.

#### Boosting
- Reduces bias
- Sequential training

Best for:
```
Weak learners
```

### Ensemble Architecture
```
Input Data
 → Multiple Models
 → Aggregation
 → Final Prediction
```

### Aggregation Methods
#### Classification
##### Majority Voting
Example:
```
[A, A, B, A]
→ A
```

#### Regression
##### Averaging
Example:
```
[100,110,90]
→ 100
```

### Bias-Variance Tradeoff
| Method        | Bias     | Variance |
| ------------- | -------- | -------- |
| Single Tree   | Low      | High     |
| Random Forest | Low      | Lower    |
| Boosting      | Very Low | Moderate |

### Why Trees Dominate Ensembles
Decision trees are:
- Flexible
- Non-linear
- High variance

Perfect candidates for ensembling.

### Advantages of Ensemble Learning
- Higher accuracy
- Better generalization
- Robustness
- Handles complex relationships

### Limitations

- Larger models
- Slower inference
- Harder interpretability
- More tuning required
