### Machine Learning
Machine Learning is a field of AI where computers **learn patterns from data and make predictions or decisions** without being explicitly programmed with rules.

Instead of writing:
```
IF age > 60 AND BP > 140 THEN disease = yes
```

Provide:
- Data
- Desired Output
- Algorithm

The model **learns the rule itself.**

### Formal Definition
Machine Learning is a **method of building models** that **learn patterns from historical data** and use those patterns to **make predictions or decisions on unseen data**.

### Key Components of ML
1. Data
2. Features (X)
3. Target (y)
4. Algorithm
5. Loss Function
6. Optimizer
7. Model

#### 1. Data
Raw information.

Examples:
- CSV files
- Databases
- Logs
- Images
- Text

#### 2. Features (X)
Input variables.

Example:
```python
X = [size, bedrooms, location_score]
```

#### 3. Target (y)
What to predict.
```python
y = house_price
```

#### 4. Algorithm
Learning method.

Examples:
- Linear Regression
- Decision Tree
- KNN

#### 5. Loss Function
Measures error.

Example:
```python
Actual Price = 100
Predicted Price = 90

Error = (100 - 90)^2 = 100
```
Lower loss -> better model

#### 6. Optimizer
Method to minimize loss.   
Most common: Gradient Descent

#### 7. Model
Final learned mathematical function.
```python
y = f(X)
```

### Example

````python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
X = np.array([[800], [1200], [1500], [2000]])
y = np.array([50, 75, 95, 130])

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
pred = model.predict([[1800]])
print(pred)
````

The model learned:
```python
price = w * size + b
```

Where:
- w = slope
- b = intercept

### Types of Machine Learning
1. Supervised Learning
2. Unsupervised Learning
3. Semi-Supervised Learning
4. Reinforcement Learning

#### 1. Supervised Learning
**Definition**: Learning from **labeled data.**   
That means:
```python 
Input (X) -> Output (y) provided
```

Example:
| Size | Bedrooms | Price |
| ---: | -------: | ----: |
| 1000 |        2 |    60 |
| 1500 |        3 |    90 |

Here, **Price is known** -> labeled.

**Goal:** Learning mapping:  
```python
X -> y
```
So model can predict **y for new X.**

##### Two Types of Supervised Learning
##### a. Regression (continuous output)
Predict a number.   
Examples:
- House price
- Salary
- Temperature

Output:
```python
y = 78.6
```

Algorithms:
- Linear Regression
- Ridge/Lasso
- Random Forest Regressor

##### b. Classification (categorical output)
Predict a class.   
Examples:
- Spam / Not spam
- Fraud / Not fraud
- Disease / No disease

Output:
```python
y = "Spam"
```

Algorithms:
- Logistic Regression
- KNN
- Decision Tree
- Random Forest

### Example (Supervised)

````python
from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4]]
y = [30, 40, 55, 70]

model = LinearRegression()
model.fit(X, y)

print(model.predict([[5]]))
````

#### 2. Unsupervised Learning
**Definition**: Learning from **unlabeled data.**   
No target column.
```python
Only X available
```

**Goal:** Discover:
- Groups 
- Structure
- Patterns

##### Common Tasks
##### a. Clustering
Group similar data.   
Example:   
Customers grouped by behavior.

##### b. Dimensionality Reduction
Reduce features.   
Example:    
100 columns -> 10 important columns.

### Example

````python
from sklearn.cluster import KMeans

X = [[1, 2], [1, 3], [8, 9], [9, 10]]

model = KMeans(n_clusters=2)
model.fit(X)

print(model.labels_)
````

**When to Use**  
- No labels
- Want hidden patterns

#### 3. Semi-Supervised Learning
**Definition**: Small amount of labeled data + large unlabeled data    
Example:    
- 1000 images
- Only 100 labeled   

Use 100 labels + 900 unlabeled

Why Useful?  
Labeling is expensive.

Used In   
- Image classification
- Speech recognition

Simple Idea   
1. Train model on labeled data
2. Predict labels for unlabeled
3. Retrain

#### 4. Reinforcement Learning (RL)
**Defintion**: An **agent** learns by interacting with environment and receiving **rewards or penalties.**

**Core Elements**
- Agent
- Environment
- Action
- Reward
- State

**Example**
Game playing AI:
- Move left -> -1
- Move right -> +10

Goal: Maximize total reward.

Used In
- Robotics
- Games
- Self-driving cars


### Quick Comparison

| Type            | Labels?   | Output              |
| --------------- | --------- | ------------------- |
| Supervised      | Yes       | Predict value/class |
| Unsupervised    | No        | Groups / structure  |
| Semi-supervised | Few       | Hybrid              |
| Reinforcement   | No labels | Actions             |


### ML Problem Framing
Converting a **business question** into a **Machine Learning task**.  
> *Regession vs Classification vs Clustering*

#### Three Most Common ML Problem Types
1. Regression
2. Classification
3. Clustering

##### 1. Regression Problems
**Defintion**: Predict a **continuous numerical value.**   
Ouptut is a number.
```python
y = 72.5
```

Examples:
- House price
- Salary
- Sales amount
- Temperature

How to Recognize   
Ask:
> Is the output a number on a scale?   

If yes -> Regression.

Example Dataset:
| Area | Rooms | Price |
| ---: | ----: | ----: |
|  900 |     2 |    55 |
| 1200 |     3 |    75 |

Target = Price -> continuous

**Typical Algortithms**
- Linear Regression
- Ridge / Lasso
- Random Forest Regressor
- Gradient Boosting Regressor

### Example

````python
from sklearn.linear_model import LinearRegression

X = [[900], [1200], [1500]]
y = [55, 75, 95]

model = LinearRegression()
model.fit(X, y)

print(model.predict([[1300]]))
````

##### 2. Classification Problems
**Definition**: Predict a **category / class.**
```python
y ∈ {Spam, Not Spam}
```

##### Types of Classification
##### a. Binary Classification
Two classes.
- Yes / No
- Fraud / Not Fraud

##### b. Multiclass Classification
More than two classes.
- Digits 0–9
- Disease types

##### c. Multi-label Classification
More than one label per sample.
- Movie genres: Action + Comedy

How to Recognize  
Ask:
> Is the output a label or category?

If yes -> Classification.

Example Dataset:
| Age | Income | Purchased |
| --: | -----: | --------: |
|  25 |    30k |        No |
|  40 |    80k |       Yes |

Target = Purchased -> category

**Typical Algorithms**
- Logistic Regression
- KNN
- Naive Bayes
- Decision Tree
- Random Forest

### Example

````python
from sklearn.linear_model import LogisticRegression

X = [[25, 30], [40, 80]]
y = [0, 1]

model = LogisticRegression()
model.fit(X, y)

print(model.predict([[30, 50]]))
````

##### 3. Clustering 
**Definition**: Group similar data points **without labels.**

Examples:
- Customer segmentation
- Group similar articles
- Image grouping

How to Recognize  
Ask:
> Do we have labels?  

If no -> Clustering

Example   
Customers with similar purchase behavior

**Typical Algorithms**
- K-Means
- Hierarchical Clustering
- DBSCAN

### Example

````python
from sklearn.cluster import KMeans

X = [[1, 2], [1, 3], [8, 9], [9, 10]]

model = KMeans(n_clusters=2)
model.fit(X)

print(model.labels_)
````

### ML Pipeline
An ordered sequence of steps that transforms:   
**Raw Data -> Deployed Model -> Business Value**

**Steps:**
- Business Problem
- Data Collection
- EDA
- Data Cleaning
- Feature Engineering
- Train / Validation / Test Split
- Model Training
- Evaluation
- Hyperparameter Tuning
- Final Model
- Deployment
- Monitoring and Retraining

### Train / Validation / Test Split
If the model is trained and tested in the **same data**, it can memorize.  
Then:
- Accuracy looks high
- Real-world performance is poor

This is called **Overfitting**

#### Three Sets
##### Training Set
Used to learn patterns

##### Validation Set
Used to tune model and hyperparamters

##### Test Set
Used only once for final evaluation

#### Typical Splits
| Dataset Size | Train | Val | Test |
| ------------ | ----- | --- | ---- |
| Small        | 60%   | 20% | 20%  |
| Medium       | 70%   | 15% | 15%  |
| Large        | 80%   | 10% | 10%  |


#### Why not just Train & Test?
Without validation:   
It tune hyperparameters using test set.  
This leaks information.   
Test accuracy becomes optimistic.    

### Overfitting & Underfitting
#### Underfitting
**Definition**: Model is **too simple** to learn patterns.

**Symptoms**
- Low training accuracy
- Low validation accuracy

**Example**
Trying to fit straight line to curved data.

**Causes**
- Too simple model
- Important features missing
- Too much regularization

**Fix**
- More complex model
- Better features
- Reduce regularization

#### Overfitting
**Definition**: Model learns **training data too well**, including noise.

**Symptoms**
- High training accuracy
- Low validation accuracy

**Example**
Polynomial curve passing through every point.

**Causes**
- Too complex model
- Too many features
- Small dataset

**Fix**
- More data
- Regularization
- Simpler model
- Early stopping

#### Summary
```python
Underfitting -> both bad
Good fit -> both good
Overfitting -> train good, val bad
```


### Bias-Variance Tradeoff
Almost every ML decision is a tradeoff between:
- Simplicity
- Complexity

Bias–Variance explains **why models fail.**

#### Bias
**Definition**: Error due to **oversimplified assumptions.**

**High Bias Model**
- Too simple
- Misses patterns

**Symptoms**
- Low train accuracy
- Low test accuracy

**Example**
Fitting straight line to curved data.

#### Variance
**Definition**: Error due to **model sensitivity to training data.**

**High Variance Model**
- Too complex
- Learns noise

**Symptoms**
- High train accuracy
- Low test accuracy

**Example**
Very high-degree polynomial.


### Tradeoff
- Reducing bias usually increases variance.  
- Reducing variance usually increases bias.

Goal:    
Balance both.

#### Relationship Summary
| Problem      | Bias | Variance |
| ------------ | ---- | -------- |
| Underfitting | High | Low      |
| Good Fit     | Low  | Low      |
| Overfitting  | Low  | High     |

#### Reduce Bias
- More features
- Complex model
- Less regularization

#### Reduce Variance
- More data
- Simpler model
- Regularization
- Cross-validation
