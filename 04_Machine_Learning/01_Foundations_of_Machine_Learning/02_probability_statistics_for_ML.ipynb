### 1. Mean (Average)
The center of the data.

$
  \mu = \frac{1}{n} \sum_{i=1}^{n} x_i
$

#### Example:
Data:
```
[10, 20, 30]
```

Mean:
```
(10 + 20 + 30) / 3 = 20
```

#### Used in
- loss functions
- normalization
- Gaussian models

````python
import numpy as np
data = np.array([10,20,30])
print(np.mean(data))
````

### 2. Variance & Standard Deviation
Mean tells center.   
Variance tells **spread.**

$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2
$

#### Example:
Data:
```
[10, 20, 30]
Mean = 20
```

Deviations:
```
(-10, 0, 10)
Squares:
(100, 0, 100)
```

Variance:
```
(100 + 0 + 100)/3 = 66.67
```

#### Standard Deviation
$
\sigma = \sqrt{variance}
$

#### Uses:
- Standardization uses mean & std
- Gaussian distribution uses variance
- Bias–Variance concept comes from this

### 3. Probability Basics
Probability measures chance.

$
P(A) = \frac{\text{favorable outcomes}}{\text{total outcomes}}
$

#### Example:
Coin flip:
```
P(Heads) = 0.5
```


### 4. Important Distributions
#### Uniform Distribution
All values equally likely.

#### Normal (Gaussian) Distribution
Bell curve.   
Defined by:
- Mean $(\mu)$
- Variance $(\sigma^2)$

##### Uses:
- Many natural datasets follow normal distribution
- Linear regression assumes normally distributed errors
- Naive Bayes uses Gaussian distribution

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(0,1,1000)
plt.hist(data, bins=30)
plt.show()
````

### 5. Conditional Probability
Probability of A given B.

$
P(A|B)
$

#### Example:
Probability of disease given positive test.

### 6. Bayes Theorem
$
P(A \mid B) = \frac{P(B \mid A)\,P(A)}{P(B)}
$

#### Interpretation:
Posterior = Likelihood * Prior / Evidence

#### Used in:
- Naive Bayes classifier
- Bayesian inference

#### Example:
Spam detection:
- P(Spam) = 0.4
- P(Word="Offer"|Spam) = 0.8
- P(Word="Offer") = 0.5

$
P(Spam|Offer) = \frac{0.8 * 0.4}{0.5} = 0.64
$
