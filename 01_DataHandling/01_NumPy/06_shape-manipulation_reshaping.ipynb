### reshape()
Change Shape without changing Data

````python
import numpy as np
````

````python
a = np.arange(6) # (6,)
b = a.reshape(2, 3) # (2, 3)
b
````

### View vs Copy
`reshape()` usually returns a view

````python
a = np.arange(6)
b = a.reshape(2, 3)
b[0, 0] = 999
print(a)
````

For safety, use `copy()`

````python
a = np.arange(6)
b = a.reshape(2, 3).copy()
b[0, 0] = 999
print(a)
````

### Transpose for 3D and Higher

````python
A = np.zeros((2, 3, 4))
A
````

````python
A.shape
````

(axis 0, axis 1, axis 2) = (2, 3, 4) = (batch, rows, columns)

````python
B = np.transpose(A, axes=(1, 0, 2))
B
````

````python
B.shape
````

- Old axis 1 → new axis 0
- Old axis 0 → new axis 1
- Old axis 2 → stays axis 2

np.transpose(array, axes=(new_order_of_axes))

### swapaxes() & moveaxis()

#### swapaxes()
np.swapaxes(array, axis1, axis2)

````python
A = np.zeros((2, 3, 4))
B = np.swapaxes(A, 0, 1)

print(A.shape)
print(B.shape)

# Equivalent to:
np.transpose(A, axes=(1, 0, 2))
````

#### moveaxis()
np.moveaxis(array, source, destination)

````python
A = np.zeros((2, 3, 4))
B = np.moveaxis(A, 0, -1)
print(A.shape)
print(B.shape)
````

**NOTE:** `transpose`, `swapaxes`, `moveaxis` -> creates views

### Adding and Removing Dimensions

#### Add Dimension

````python
a = np.array([1, 2, 3])
print(a.shape)

print(a[:, np.newaxis]) # (3, 1)
print(a[np.newaxis, :]) # (1, 3)

np.expand_dims(a, axis=1)
````

#### Remove size-1 dimensions

````python
np.squeeze(a)
````

### Stacking Arrays

#### stack (new dimension)

````python
np.stack([a, a, a], axis=0)
````

#### Vertical/Horizontal

````python
print(np.vstack([a, a]))
print(np.hstack([a, a]))
````

### Concatenation
- `stack` -> adds dimension
- `concatenate` -> extends existing one

````python
np.concatenate([a, a], axis=0)
````

````python
print(np.split(a, 3))
print(np.array_split(a, 4))

````
