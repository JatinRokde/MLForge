```python
import numpy as np

a = np.array([1, 2, 3])
print(a.dtype)
```

**Why dtype matters:**
- Memory: `float32` = 4 bytes, `float64` = 8 bytes
- Speed: smaller types → better cache usage
- Correctness: integers cannot represent `NaN`
- ML compatibility: GPUs prefer float32

| Category | Examples                          |
| -------- | --------------------------------- |
| Integers | `int8`, `int16`, `int32`, `int64` |
| Unsigned | `uint8`, `uint16`, `uint32`       |
| Floats   | `float16`, `float32`, `float64`   |
| Boolean  | `bool`                            |
| Complex  | `complex64`, `complex128`         |


```python
np.dtype(np.float64).itemsize   # 8 bytes
```

```python
np.dtype(np.float32).itemsize   # 4 bytes
```

### Explicitly Setting dtype

#### At creation time:

```python
np.array([1, 2, 3], dtype=np.float32)
```

```python
np.zeros((2, 3), dtype=np.int32)
```

#### Casting later (astype)

```python
a = np.array([1, 2, 3])
b = a.astype(np.float32)
```

`astype()` **always creates a copy.**

### Automatic Type Promotion (Upcasting)
NumPy chooses the smallest common type that can represent all values.

```python
np.array([1, 2, 3.5]).dtype
```

#### Common upcasting rules:
int -> float -> complex

```python
np.array([1, 2, True]).dtype
```

```python
np.array([1, 2.0]).dtype
```

```python
np.array([1, 2 + 3j]).dtype
```

#### Arithmetic upcasts

```python
a = np.array([1, 2, 3], dtype=np.int32)
b = a + 0.5
b.dtype
```

### NaN, inf, -inf
- Means 'Not a Number'
- Exists **only in floating types**

**Why can’t NumPy store NaN in integer arrays?**

Because:
- NaN is a **floating-point concept (IEEE 754)**
- Integers have no bit-pattern for NaN

This is fundamental to numerical computing.

```python
a = np.array([1, 2, np.nan])
a.dtype
```

```python
print(np.isnan(a))
print(np.isinf(a))
print(np.isfinite(a))
```

### NaN-Safe Aggregations

```python
print(np.mean([1, 2, np.nan]))
```

```python
print(np.nanmean([1, 2, np.nan]))  
print(np.nansum([1, 2, np.nan]))    
print(np.nanmax([1, np.nan, 4]))    
```

### Replacing/Handling NaN Values

#### Replace NaN with values:

```python
a = np.array([1, np.nan, 3])
a[np.isnan(a)] = 0
a
```

#### Using np.nan_to_num:

```python
np.nan_to_num(a, nan=0, posinf=1e9, neginf=-1e9)
```

Used heavily in:
- feature engineering
- loss stabilization
- numerical pipelines

### Silent Precision Bugs

```python
a = np.array([1e20, 1, -1e20])
np.sum(a)

# Expected: 1
```

This is **floating-point precision loss.**
