### Precision Loss

````python
a = np.array([1e20, 1, -1e20])
np.sum(a)
````

**Because:**
- `1e20` and `-1e20` cancel
- `1` is too small to survive rounding

### Log-Sum-Exp

#### Unstable
```python
np.log(np.sum(np.exp(x)))
```

#### Stable
```python
m = np.max(x)
np.log(np.sum(np.exp(x - m))) + m
```

### Floating-Point Comparisons

````python
0.1 + 0.2 == 0.3
````

````python
np.isclose(0.1 + 0.2, 0.3)
````

### Exercise 1
Why does this return `False`?
```python
0.1 + 0.2 == 0.3
```

Floating-point numbers cannot represent `0.1` and `0.2` exactly in binary, the result is stored as a slightly different value than `0.3`, so exact equality fails.

### Exercise 2
Fix this unstable softmax:
```python
softmax = np.exp(x) / np.sum(np.exp(x))
```

#### Stable
```python
x_shifted = x - np.max(x)
softmax = np.exp(x_shifted) / np.sum(np.exp(x_shifted))
```

### Exercise 3
Fix this loss computation:
```python
loss = -np.log(p)
```
Assume `p` may contain zeros.

#### Stable
```python
eps = 1e-9
loss = -np.log(np.clip(p, eps, 1 - eps))
```
