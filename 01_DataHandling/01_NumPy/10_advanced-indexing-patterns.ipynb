### `np.where()` - Conditional Logic without Loops
```python
np.where(condition, x, y)
```

````python
import numpy as np
````

````python
a = np.array([1, -2, 3, -4])

np.where(a > 0, a, 0)
````

#### `np.where()` as Index Finder

````python
idx = np.where(a < 0)
idx
````

### Logical Opoerations

````python
mask = np.logical_and(a > 0, a < 5)
a[mask]
````

#### Equivalent Shorthand:

````python
a[(a > 0) & (a < 5)]
````

### `np.take()` - Indexed Reads

````python
a = np.array([10, 20, 30, 40])
np.take(a, [0, 3])
````

### `np.put()` - Indexed Writes

````python
a = np.array([1, 2, 3, 4])
np.put(a, [0, 2], 100)
a
````

### `np.ix_()` - Cartesian Indexing

````python
rows = [0, 2]
cols = [1, 3]

np.ix_(rows, cols)
````

### One-Hot Encoding (NumPy)

````python
labels = np.array([0, 2, 1])
num_classes = 3

one_hot = np.zeros((len(labels), num_classes))
one_hot[np.arange(len(labels)), labels] = 1

one_hot
````

### Binning & Bucketing

````python
values = np.array([10, 25, 35, 60])
bins = np.array([0, 20, 40, 60])

np.digitize(values, bins)
````

### Masked Updates

````python
X = np.array([[1, -1],
              [3, -2]])

X[X < 0] = 0
X
````

### Exercise 1
Replace negative values with `-1` using `np.where`

````python
a = np.array([3, -2, 5, -7, 0])
result = np.where(a < 0, -1, a)
print(result)
````

### Exercise 2
Select rows `[1, 3]` and columns `[0, 2]` using `np.ix_`

````python
X = np.arange(20).reshape(5, 4)

rows = [1, 3]
cols = [0, 2]

submatrix = X[np.ix_(rows, cols)]
print(submatrix)
````

### Exercise 3
Create one-hot encoding for:
```python
labels = np.array([1, 0, 2, 1])
num_classes = 3
```

````python
labels = np.array([1, 0, 2, 1])
num_classes = 3

one_hot = np.zeros((len(labels), num_classes), dtype=int)
one_hot[np.arange(len(labels)), labels] = 1

print(one_hot)
````
