### NumPy (Numerical Python) is:
- a **high-performance numerical computing library**
- built around a multi-dimensional array type called `ndarray`
- written in C and optimized for vectorized computing

Everything in NumPy revolves around ndarray:
A **fixed-size, homogeneous** (same dtype), **N-dimensional** array.

Internally it stores:
- Pointer to memory
- Shape (dimensions)
- Strides (how to move to next element)
- dtype (data type)

```python
import numpy as np
```

```python
a = np.array([1, 2, 3])

print(a)
print(type(a))
print(a.dtype)
print(a.shape)
print(a.ndim) 
```

#### Array Attributes

```python
arr = np.array([[1, 2, 3],
               [4, 5, 6]])

print(arr.ndim) # number of dimensions
print(arr.shape) # rows, columns
print(arr.size) # total elements
print(arr.dtype) # data type
print(arr.itemsize) # bytes per element
print(arr.nbytes) # total bytes = itemsize * size
```

### Array Creation Methods

#### 1. np.array()
Convert Python list -> ndarray

```python
np.array([1, 2, 3])
np.array([[1, 2], [3, 4]])
```

#### 2. np.arange(start, stop, step)
Like range(), but returns ndarray

```python
np.arange(0, 10, 2)
```

#### 3. np.linspace(start, end, num)
Evenly spaced values

```python
np.linspace(0, 2, 5)
```

#### 4. Zeros/Ones/Full

```python
print(np.zeros((2, 3)))
print(np.ones((2, 3)))
print(np.full((2, 3), 5))
```

#### 5. Identity Matrix

```python
np.eye(3)
```

#### 6. Empty Array
Creates array **without initializing memory** → very fast

Contains garbage values

```python
np.empty((3, 2))
```
