### Vector-Matrix & Matrix-Vector Products

````python
A = np.array([[1, 2],
              [3, 4]])

x = np.array([10, 20])
````

````python
A @ x
````

````python
x @ A
````

**x is a 1D vector (shape (2,))**, NumPy handles it very differently than a 2D matrix.

It acts like a **"shapeshifter."** NumPy will automatically treat it as either a row or a column vector depending on where it sits in the multiplication, and the result will always be returned as a 1D array.

### Norms (Magnitude, Length, Size)
A **norm** measures the size of a vector or matrix.

| Norm | Meaning   |
| ---- | --------- |
| L2   | Euclidean |
| L1   | Manhattan |
| ∞    | Max       |


````python
x = np.array([3, 4])
````

#### L2 norm (Euclidean distance)

````python
np.linalg.norm(x)
````

#### L1 norm (Manhattan distance)

````python
np.linalg.norm(x, ord=1)
````

#### Infinity norm (Max absolute value)

````python
np.linalg.norm(x, ord=np.inf)
````

### Solving Linear Systems (Ax = b)

#### Bad Approach 
```python
x = np.linalg.inv(A) @ b
```

- expensive
- unstable

#### Correct Approach
```python
x = np.linalg.solve(A, b)
```

````python
A = np.array([[2, 1],
              [1, 3]])

b = np.array([5, 10])

np.linalg.solve(A, b)
````

### Matrix Inverse, Determinant, Rank

````python
print(np.linalg.inv(A))
print(np.linalg.det(A))
print(np.linalg.matrix_rank(A))
````

### Eigenvalues & Eigenvectors
**Intuition:**

An **eigenvector** of a matrix `A`:
- does **not change direction**
- only scales

```python
A v = λ v
```

Where:
- `v` -> eigenvector 
- `λ` -> eigenvalue


````python
A = np.array([[2, 0],
              [0, 3]])

eigvals, eigvecs = np.linalg.eig(A)

print(eigvals)
print(eigvecs)
````

### Singular Value Decomposition(SVD)

Used in:
- dimensionality reduction
- recommender systems
- low-rank approximations

````python
U, S, Vt = np.linalg.svd(A)
print(U)
print(S)
print(Vt)
````
