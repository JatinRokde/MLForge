### Vectorization
Applying an operation to an entire array **without explicit Python loops.**

````python
import numpy as np

a = np.array([1, 2, 3, 4])
b = a ** 2
print(b)
````

### Universal Functions (UFuncs)
UFuncs are **vectorized functions** that operate element-wise.

````python
print(np.add(a, 10))
print(np.multiply(a, 2))
print(np.sqrt(a))
print(np.exp(a))
print(np.log(a))
````

### Broadcasting
NumPy does not create expanded copies of arrays; it reuses the same memory and adjusts strides to simulate expansion.

#### Broadcasting Rules:
When operating on two arrays:
1. Compare shapes from right to left
2. Dimensions are compatible if:
    - they are equal, or
    - one of them is 1
3. Result shape = max of each dimension

````python
a = np.array([1, 2, 3])
a + 10
````

````python
a = np.arange(3).reshape(3,1) # (3, 1)
print(a)
b = np.arange(4) # (1, 4)
print(b)
a + b # (3, 4)
````

````python
X = np.array([[10, 20],
              [30, 40],
              [50, 60]])

mean = X.mean(axis=0) # (2,)
std = X.std(axis=0) # (2,)

X_norm = (X - mean) / std
print(X_norm)
````
