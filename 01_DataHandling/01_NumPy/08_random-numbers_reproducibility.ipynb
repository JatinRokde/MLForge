### Creating a Random Generator

````python
import numpy as np
````

````python
rng = np.random.default_rng(42)
````

- `42` is the **seed**
- Same seed -> same results (reproducibility)

### Uniform Random Numbers
#### Random floats in [0, 1)

````python
rng.random(5)
````

#### Random floats in a shape

````python
rng.random((2, 3))
````

### Random Integers

````python
rng.integers(low=0, high=10, size=5)
````

- `low` inclusive
- `high` exclusive

### Normal (Gaussian) Distribution

````python
rng.normal(loc=0.0, scale=1.0, size=5)
````

Parameters:
- `loc` -> mean (μ)
- `scale` -> standard deviation (σ)

Used in:
- weight initialization
- noise injection
- probabilistic models

### Other Useful Distributions

````python
rng.uniform(-1, 1, size=5)
````

````python
rng.exponential(scale=1.0, size=5)
````

````python
rng.binomial(n=10, p=0.5, size=5)
````

| Scenario          | Distribution |
| ----------------- | ------------ |
| Random guess      | Uniform      |
| Measurement noise | Normal       |
| Waiting time      | Exponential  |
| Binary outcomes   | Binomial     |
| Event counts      | Poisson      |
| Sampling data     | Choice       |


### Reproducibility

````python
rng1 = np.random.default_rng(42)
rng2 = np.random.default_rng(42)

print(rng1.random(5))
print(rng2.random(5))
````

### Shuffling Data
#### Shuffle in-place
Modifies the original array

````python
arr = np.array([1, 2, 3, 4, 5])
rng.shuffle(arr)
arr
````

#### Permutation (safe)
- Returns a shuffled copy
- Original unchanged

````python
rng.permutation(arr)
````

### Train-Test Split

````python
X = np.arange(20).reshape(10, 2)

indices = rng.permutation(len(X))
train_idx = indices[:7]
test_idx = indices[7:]
````

````python
X_train = X[train_idx]
X_train
````

````python
X_test = X[test_idx]
X_test
````

### Random Choice/Sampling

````python
rng.choice([10, 20, 30], size=5)
````

````python
rng.choice(10, size=5, replace=False)
````

### Testing with Randomness
#### Deterministic test example:

````python
def test_random_output():
    rng = np.random.default_rng(0)
    x = rng.random(3)
    assert np.allclose(x, [0.63696169, 0.26978671, 0.04097352])
````
