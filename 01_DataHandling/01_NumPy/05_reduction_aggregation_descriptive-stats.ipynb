**Reductions** are the operations that **collapse dimensions:**

`sum`, `mean`, `min`, `max`, etc.

### Basic Reductions (1D)

```python
import numpy as np

a = np.array([1, 2, 3, 4])

np.sum(a)
np.mean(a)    
np.min(a)     
np.max(a)     
np.prod(a) 
```

### Reductions on 2D Arrays

```python
X = np.array([[1, 2, 3],
              [4, 5, 6]])
```

#### Without axis:

```python
np.sum(X)
```

#### axis:
axis tells NumPy which dimension to collapse

**Visual model:**
- `axis=0` → collapse rows → **operate column-wise**
- `axis=1` → collapse columns → **operate row-wise**

```python
X = [[1, 2, 3],
     [4, 5, 6]]
```

#### `axis=0` (column-wise)

```python
np.sum(X, axis=0) # column-wise
```

#### `axis=1` (row-wise)

```python
np.sum(X, axis=1)
```

```python
np.sum(X, axis=0).shape   # (3,)
```

```python

np.sum(X, axis=1).shape   # (2,)
```

#### Keep dimensions

```python
np.sum(X, axis=0, keepdims=True)
```

#### argmin/argmax (Indices, not values)
Returns **indices**, not values

```python
a = np.array([10, 50, 20])

print(np.argmax(a))
print(np.argmin(a))
```

### Statistical Aggregations

```python
print(np.mean(X, axis=0))
print(np.std(X, axis=0))
print(np.var(X, axis=0))
print(np.median(X, axis=0))
print(np.percentile(X, 50, axis=0))
```

```python
a = np.array([1, 2, 3, 4])

print(np.cumsum(a))    # [1  3  6 10]
print(np.cumprod(a))   # [1  2  6 24]
```

### NaN-Safe Reductions

```python
arr = np.array([1, np.nan, 3])

print(np.sum(arr))       # nan
print(np.nansum(arr))   # 4
print(np.nanmean(arr))   # 2
print(np.nanmax(arr))   # 3
```

### Chaining Reductions (ML Pattern)
Example: Mean squared value per row

```python
X = np.array([[1, 2],
              [3, 4]])
np.mean(X**2, axis=1)
```
