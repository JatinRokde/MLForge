### Pandas
**Pandas is a high-level data manipulation library built on top of NumPy**, designed for:
- Tabular data (rows & columns)
- Labeled data (named columns + index)
- Fast vectorized operations
- Data cleaning, transformation, aggregation

### Core Building Blocks
Pandas has two fundamental data structures:
1. **Series** → 1D labeled array (column)
2. **DataFrame** → 2D labeled table (collection of Series)

````python
import pandas as pd
````

````python
scores = pd.Series([85, 90, 78])
scores
````

#### Custom Index

````python
scores = pd.Series([85, 90, 78], index=["Math", "Science", "English"])
scores
````

**Important:**
Index is *not* just position - it's a **first-class concept** in Pandas.

Pandas treats the index as first-class because it enables **automatic data alignment, joins, time-based operations, and safe transformations without relying on position.**

**Real-world example:**
- User IDs
- Timestamps
- Transaction IDs

#### Seres behave like NumPy + Dict

````python
scores.max()
````

````python
scores["Math"]
````

### Pandas DataFrame
A DataFrame is:
- A table of rows & columns
- Each column is a Series
- All columns share the same index

````python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "salary": [50000, 60000, 70000]
}

df = pd.DataFrame(data)
df
````

````python
df["age"] # returns Series
````

````python
df[["name", "salary"]] # returns DataFrame
````

### DataFrame vs Series
| Feature    | Series          | DataFrame       |
| ---------- | --------------- | --------------- |
| Dimensions | 1D              | 2D              |
| Labels     | Index           | Index + Columns |
| Use case   | Single variable | Dataset / table |


**Pandas is Powerful because:**
- Vectorization
- Label-based alignment
- Rich ecosystem

````python
s1 = pd.Series([1, 2, 3], index=["a", "b", "c"])
s2 = pd.Series([10, 20, 30], index=["b", "c", "d"])

s1 + s2
````
