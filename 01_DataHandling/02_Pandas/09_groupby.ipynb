**GroupBy** as three steps:
```
SPLIT  →  APPLY  →  COMBINE
```

- **Split** data into groups
- **Apply** a function to each group
- **Combine** results back together

### Basic GroupBy

````python
import pandas as pd

df = pd.DataFrame({
    "city": ["Pune", "Pune", "Mumbai", "Mumbai", "Delhi", "Delhi", "Delhi"],
    "user_id": [101, 102, 103, 104, 105, 106, 107],
    "age": [25, 35, 40, 60, 30, 45, 30],
    "salary": [50000, 70000, 80000, 90000, 60000, 75000, 65000]
})
````

#### Average salary per city

````python
df.groupby("city")["salary"].mean()
````

Flow:
- Split by `city`
- Applied `mean`
- Combined into a Series

#### Multiple aggregations

````python
df.groupby("city")["salary"].agg(["mean", "min", "max"])
````

### GroupBy Multiple Columns


#### Group by city and age

````python
df.groupby(["city", "age"])["salary"].mean()
````

### Aggregation Multiple Columns

````python
df.groupby("city").agg({
    "salary": ["mean", "max"],
    "age": "mean",
    "user_id": "count"
})
````

### Renaming Aggregated Columns

````python
df.groupby("city").agg(
    avg_salary=("salary", "mean"),
    max_salary=("salary", "max"),
    user_count=("user_id", "count")
)
````

### `reset_index()`
GroupBy puts group keys into the **index**.

````python
result = df.groupby("city").agg(
    avg_salary=("salary", "mean")
)
result
````

To convert back to normal DataFrame:

````python
result.reset_index()
````

### `transform()` - Same Shape as Original

#### City-wise average salary per row

````python
df["city_avg_salary"] = (
    df.groupby("city")["salary"].transform("mean")
)
df
````

| Method        | Output size      |
| ------------- | ---------------- |
| `agg()`       | Reduced          |
| `transform()` | Same as original |


### GroupBy + Filtering

#### Cities with average salary > 70000

````python
city_avg = df.groupby("city")["salary"].mean()
city_avg[city_avg > 70000]
````

Or inline:

````python
df.groupby("city").filter(lambda x: x["salary"].mean() > 70000)
````

### ML Feature Engineering Example

````python
df.assign(
    city_salary_mean = lambda x: x.groupby("city")["salary"].transform("mean"),
    city_salary_std = lambda x: x.groupby("city")["salary"].transform("std"),
    salary_vs_city_avg = lambda x: x["salary"] - x.groupby("city")["salary"].transform("mean")
)
````

### Summary Table

| Goal                  | Use                     |
| --------------------- | ----------------------- |
| Aggregate             | `groupby().agg()`       |
| Row-level group stats | `groupby().transform()` |
| Complex logic         | `groupby().apply()`     |
| Filter groups         | `groupby().filter()`    |
| Clean output          | `reset_index()`         |

````python
import pandas as pd

df = pd.DataFrame({
    "city": ["Pune", "Pune", "Mumbai", "Mumbai", "Delhi", "Delhi", "Delhi"],
    "user_id": [101, 102, 103, 104, 105, 106, 107],
    "age": [25, 35, 40, 60, 30, 45, 30],
    "salary": [50000, 70000, 80000, 90000, 60000, 75000, 65000]
})
````

````python
df.groupby("city")["salary"].mean()
````

````python
df.groupby("city")["salary"].agg(["mean", "max", "min"])
````

### Exercise 2
Create a grouped DataFrame with columns:
- `avg_salary`
- `max_salary`
- `user_count`
- `avg_age`


````python
grouped_df = (
    df.groupby("city")
    .agg(
        avg_salary=("salary", "mean"),
        max_salary=("salary", "max"),
        user_count=("user_id", "count"),
        avg_age=("age", "mean")
    )
)

grouped_df
````

### Exercise 3
