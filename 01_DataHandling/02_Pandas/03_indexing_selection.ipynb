### Quick Recap
| Concept      | Meaning                     |
| ------------ | --------------------------- |
| **Index**    | Labels (not positions)      |
| **Position** | Integer location (0, 1, 2…) |
| Pandas       | Prefers **labels first**    |


````python
import pandas as pd
````

````python
df = pd.DataFrame({
    "name": ["Asha", "Ravi", "Neha"],
    "age": [21, 23, 22],
    "city": ["Pune", "Mumbai", "Delhi"]
}, index=["u1", "u2", "u3"])

df
````

````python
df["age"] # Series
````

````python
df[["name", "city"]] # DataFrame
````

### `.loc` - Label-Based Indexing
Syntax: 
```python
df.loc[row_label, column_label]
```

````python
df.loc["u1"]
````

````python
df.loc["u1", "age"]
````

````python
df.loc[["u1", "u3"], ["name", "city"]]
````

### Label-Based Slicing
**Important: It is inclusive of both ends**

````python
df.loc["u1":"u3"] 
````

### `.iloc` - Position-Based Indexing
Syntax
```python
df.iloc[row_position, column_position]
```

````python
df.iloc[0]
````

````python
df.iloc[0, 1] # age of first row
````

````python
df.iloc[:, 1:] # all rows, columns from position 1
````

#### Position-Based Slicing
**Important: End is exclusive**

````python
df.iloc[0:2]
````

### `.at` & `.iat` - Fast Scalar Access
Used when accessing **single values.**

|Method|Based On|Speed|
|----|-----|--------|
|`at`|label|very fast|
|`iat`|position|very fast|

````python
df.at["u2", "city"]
````

````python
df.iat[1, 2]
````

### Boolean Indexing

````python
df[df["age"] > 21]
````

#### Equivalent (preferred):

````python
df.loc[df["age"] > 21]
````

#### Multiple Conditions

````python
df.loc[(df["age"] > 21) & (df["city"] == "Mumbai")]
````

**Important**

Always use:
- `&`, `|`, Parantheses

### `.query()` - Clean, SQL - like Filters

````python
df.query("age > 21 and city == 'Mumbai'")
````

### Chained Indexing
Avoid this - may fail

````python
df[df["age"] > 21]["city"] = "Bangalore"
````

#### Correct Way

````python
df.loc[df["age"] > 21, "city"] = "Bangalore"
df
````

````python
# ML feature engineering
df.loc[df["age"] < 18, "is_minor"] = 1
df.loc[df["age"] >= 18, "is_minor"] = 0
df
````

### Summary Table
| Task                    | Use               |
| ----------------------- | ----------------- |
| Select by label         | `loc`             |
| Select by position      | `iloc`            |
| Single value (label)    | `at`              |
| Single value (position) | `iat`             |
| Filtering rows          | `loc` + condition |
| Assignment              | `loc`             |
