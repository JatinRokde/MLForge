### Creating a Series

#### 1. From a Python List

````python
import pandas as pd
````

````python
s = pd.Series([10, 20, 30])
s
````

#### 2. With Explicit `dtype`

````python
s = pd.Series([10, 20, 30], dtype="int32")
s
````

#### 3. From a Dictionary

````python
s = pd.Series({
    "a": 100,
    "b": 200,
    "c": 300
})
s
````

#### 4. From NumPy Array

````python
import numpy as np

arr = np.array([1.5, 2.5, 3])
s = pd.Series(arr)
s
````

### Creating a DataFrame

#### 1. From a Dictionary of Lists

````python
data = {
    "name": ["Asha", "Ravi", "Neha"],
    "age": [21, 23, 22],
    "salary": [40000, 50000, 45000]
}

df = pd.DataFrame(data)
df
````

#### 2. Explicit Column Order

````python
df = pd.DataFrame(data, columns=["name", "salary", "age"])
df
````

#### 3. From List of Dictionaries

````python
records = [
    {"name": "Asha", "age": 21, "city": "Pune"},
    {"name": "Ravi", "age": 23, "city": "Mumbai"},
    {"name": "Neha", "age": 22, "city": "Delhi"}
]

df = pd.DataFrame(records)
df
````

#### 4. Missing Keys in Dicts

````python
records = [
    {"name": "Asha", "age": 21},
    {"name": "Ravi", "city": "Mumbai"},
]

df = pd.DataFrame(records)
df
````

### Creating DataFrame From NumPy Arrays

#### 1 . Simple Example

````python
arr = np.array([[1, 2],
                [3, 4]])

df = pd.DataFrame(arr, columns=["A", "B"])
df
````

#### 2. Control Index & Columns

````python
df = pd.DataFrame(
    arr, 
    index=["row1", "row2"],
    columns=["col1", "col2"]
)

df
````

### `dtype` Control

#### Automatic Inference Pitfall

````python
df = pd.DataFrame({
    "id": [1, 2, 3],
    "score": [90, None, 85]
})

df.dtypes
````

Integer column became float due to `NaN`.

#### Correct Way

````python
df = pd.DataFrame({
    "id": pd.Series([1, 2, 3], dtype="int64"),
    "score": pd.Series([90, None, 85], dtype="Int64")
})

df.dtypes
````

`Int64` supports missing values.
