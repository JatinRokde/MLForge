### What is Matplotlib?
Matplotlib is Python’s foundational plotting library.
Almost every visualization library (including Seaborn, Pandas plotting) sits on top of it.

Key idea:
> Matplotlib draws **figures** containing **axes**, on which data is plotted.

#### Figure - *the whole canvas*
A **Figure** is the **entire drawing area.**
Think of it as:
- A blank window
- A page in a report
- The full image saved to disk (.png, .pdf, etc.)

**Key points**
- A figure can contain one or more Axes
- Controls global things like:
   - Size (figsize)
   - DPI
   - Saving (fig.savefig())

#### Axes — *where data is actually plotted*
An **Axes** is the **actual plot area** with:
- X and Y axis
- Ticks and labels
- Grid
- Title
- Lines, bars, scatter points, etc.

All plotting happens on an Axes, not on the Figure.

```python
import matplotlib.pylab as plt
```

### Two Ways to Use Matplotlib
#### A. Pyplot (State-Based API) - Beginner Friendly
- Implicit
- Easy
- Less control

```python
plt.plot([1, 2, 3], [10, 20, 30])
plt.show()
```

Matplotlib:
- Creates figure
- Creates axes
- Plots data Automatically

#### B. Object-Oriented (OO) API - *Professional Way*
- Explicit
- Scalable
- Production-grade

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [10, 20, 30])
plt.show()
```

| Component | Meaning                                  |
| --------- | ---------------------------------------- |
| Figure    | Entire canvas/window                     |
| Axes      | Actual plot area                         |
| Axis      | X-axis or Y-axis                         |
| Artist    | Everything you see (lines, text, legend) |

Think:
- **Figure** = paper
- **Axes** = graph drawn on paper

### Simple Line Plot

```python
x = [1, 2, 3, 4]
y = [10, 15, 25, 30]

fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_title("Simple Line Plot")
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")

plt.show()
```

### Common Plot Types

#### Line Plot (Trends, Time-Series)

```python
fig, ax = plt.subplots()
ax.plot(x, y, color="blue", linestyle="--", marker="o")
plt.show()
```

Use when:
- Trend matters
- Time-series
- Continuous data

#### Scatter (Relationships)

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

Use when:
- Relationship between variables
- Feature vs target

#### Bar Chart (Category Comparison)

```python
categories = ["A", "B", "C"]
values = [10, 20, 15]

fig, ax = plt.subplots()
ax.bar(categories, values)
plt.show()
```

#### Histogram (Distribution)

```python
import numpy as np

data = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(data, bins=30)
plt.show()
```

### Styling Basics (Colors, Markers, Lines)

```python
fig, ax = plt.subplots()

ax.plot(
    x, y, 
    color="green",
    linewidth=2,
    linestyle="-.",
    marker="s"
)

plt.show()
```

### Figure Size & DPI

```python
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)
```

- `figsize`: inches (width=8, height=5)
- `dpi`: resolution

### `plt.show()` 
- Renders the plot
- Clears the figure after display
