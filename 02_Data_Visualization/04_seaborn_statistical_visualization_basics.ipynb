### What is Seaborn
**Seaborn** is a **high-level statistical visualization library** built on **Matplotlib.**

Key problems Seaborn solves:
- Sensible statistical defaults
- Beautiful plots with minimal code
- Direct integration with **Pandas DataFrames**
- Automatic handling of:
    - Confidence intervals
    - Aggregations
    - Categorical grouping

In industry:
- **Seaborn -> EDA**
- **Matplotlib -> final / production plots**


### Seaborn vs Matplotlib
| Aspect          | Matplotlib        | Seaborn     |
| --------------- | ----------------- | ----------- |
| Abstraction     | Low-level         | High-level  |
| Code length     | Verbose           | Concise     |
| Defaults        | Plain             | Statistical |
| Pandas-friendly | Manual            | Native      |
| Best use        | Control & reports | EDA         |


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df.head()
```

Common Datasets:
- `tips`
- `iris`
- `penguins`
- `titanic`

### Distribution Plots
#### Histogram - `histplot`

```python
sns.histplot(df["total_bill"], bins=20)
plt.show()
```

Used when:
- Checking skewness
- Understanding spread

### KDE Plot (Smooth Distribution)
A **KDE plot** (short for **Kernel Density Estimation plot**) is a smooth, continuous way to visualize the **distribution of numerical data**. It’s often used as a refined alternative to a histogram.

#### What it shows
- The **probability density** of a variable (how values are distributed)
- A **smoothed curve** instead of bars
- Peaks indicate where values are more concentrated

```python
sns.kdeplot(df["total_bill"], fill=True)
plt.show()
```

- Better for continuous data
- Misleading for small datasets

#### Combined Distribution - `displot`

```python
sns.displot(
    df,
    x="total_bill",
    kde=True
)
plt.show()
```

### Categorical Plots
#### Count Plot

```python
sns.countplot(
    data=df,
    x="day"
)
plt.show()
```

Use when:
- Counting occurrences
- Class balance checks

#### Bar Plot (With Aggregation)

```python
sns.barplot(
    data=df,
    x="day",
    y="total_bill",
    estimator="mean",
    capsize=.1
)
plt.show()
```

- Aggregates automatically
- Shows confidence intervals

#### Box Plot (Outliers & Spread)

```python
sns.boxplot(
    data=df,
    x="day",
    y="total_bill"
)
plt.show()
```

#### Violin Plot (Distribution + Density)

```python
sns.violinplot(
    data=df,
    x="day",
    y="total_bill"
)
plt.show()
```

### Scatter & Line Plots
#### Scatter Plot

```python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="day",
    style="sex"
)
plt.show()
```

#### Line Plot 

```python
sns.lineplot(
    data=df,
    x="size",
    y="total_bill"
)
plt.show()
```

### Semantic Mappings
| Parameter | Meaning           |
| --------- | ----------------- |
| `hue`     | Color grouping    |
| `style`   | Marker/line style |
| `size`    | Size encoding     |


```python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="day",
    size="size"
)
plt.show()
```
