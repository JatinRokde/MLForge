### FacetGrid - Full Control Over Faceting
`FacetGrid` is the engine behind `relplot` and `catplot`

```python
df = sns.load_dataset("tips")

g = sns.FacetGrid(
    data=df,
    col="day",
    hue="sex",
    height=4
)

g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()

plt.savefig(
    "06_facetGrid_scatter_plot.png",
    dpi=500,
    bbox_inches="tight"
)
plt.show()
```

- Maximum control
- Slightly more verbose
- Used in advanced dashboards

### Multiple Semantic Mappings (Advanced)

```python
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="day",
    style="sex",
    size="size"
)
plt.show()
```

In one visualization, you can see:
- How tips relate to total bill
- Differences by `day`
- Differences by `gender`
- Impact of `group size`

This is a **multivariate visualization** (more than two variables).

### Combining Seaborn + Matplotlib
Seaborn creates plots -> Matplotlib customizes them.

```python
fig, ax = plt.subplots(figsize=(8, 5))

sns.boxplot(
    data=df,
    x="day",
    y="total_bill",
    ax=ax
)

ax.set_title("Total Bill Distribution by Day")
ax.set_xlabel("Day")
ax.set_ylabel("Total Bill ($)")

plt.show()
```

### Advanced Annotations in Seaborn Plots

```python
ax = sns.boxplot(
    data=df,
    x="day",
    y="total_bill"
)

ax.annotate(
    "High variance",
    xy=(2, 40),
    xytext=(1.5, 45),
    arrowprops=dict(arrowstyle="->")
)

plt.show()
```

### Controlling Confidence Intervals

```python
sns.lineplot(
    data=df,
    x="size",
    y="total_bill",
    errorbar=None
)
plt.show()
```
