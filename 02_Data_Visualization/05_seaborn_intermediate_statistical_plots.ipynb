### Relationship Plots - `relplot` (High-Level API)
`relplot` is a **figure-level** function used to visualize relationships between variables, especially numerical relationships.

Think of `relplot()` as a **high-level wrapper** around `scatterplot()` and `lineplot()`.

`relplot()` is used to:
Show relationships between **x and y**
Encode extra variables using color (`hue`), **size**, and **style**
Create **multiple subplots (facets)** easily

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")

sns.relplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="day",
    col="sex",
    kind="scatter"
)
plt.show()
```

- Automaticallt creates multiple subplots
- Ideal for segmented analysis

### Categorical Grids - `catplot`
`catplot` is the **categorical counterpart** of `replot`.

```python
sns.catplot(
    data=df,
    x="day",
    y="total_bill",
    hue="sex",
    kind="box"
)
plt.show()
```

Supported `kind` values:
- `bar`
- `box`
- `violin`
- `strip`
- `swarm`

Use `catplot` when you need **faceting + categorical comparison**

### Regression Plots - `regplot` & `lmplot`
#### `regplot` (Axes-level)

```python
sns.regplot(
    data=df,
    x="total_bill",
    y="tip"
)
plt.show()
```

- Adds regression line
- Confidence interval by default

#### `lmplot` (Figure-level)

```python
sns.lmplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex"
)
plt.show()
```

- `regplot` -> single plot
- `lmplot` -> multiple groups / facets

### Pair Plot - `pairplot`
Used to visualize *all pairwise relationships*.

```python
sns.pairplot(
    df,
    hue="sex"
)
plt.show()
```

- Feature correlation
- Outlier detection
- Multivariate patterns

### Join Plot - `joinplot` (Relationship + Distribution)

```python
sns.jointplot(
    data=df,
    x="total_bill",
    y="tip",
    kind="scatter"
)
plt.show()
```

Other `kind` options:
- `kde`
- `hex`
- `reg`


> Relationship in center    
> Marginal distributions on sides

### Confidence Intervals
By default, Seaborn:
- Aggregates values
- Shows **95% CI**

Disable CI if needed:

```python
sns.barplot(
    data=df,
    x="day",
    y="total_bill",
    errorbar=None
)
plt.show()
```

### Working Natively with Pandas DataFrames
Tidy data:
- One variable per column
- One observation per row

```python
sns.boxplot(
    data=df,
    x="day",
    y="tip",
    hue="sex"
)
plt.show()
```

### Axes-level vs Figure-level (Critical Distinction)
| Type         | Examples                                    | Use Case     |
| ------------ | ------------------------------------------- | ------------ |
| Axes-level   | `scatterplot`, `boxplot`, `regplot`         | Fine control |
| Figure-level | `relplot`, `catplot`, `lmplot`, `jointplot` | Faceting     |
