### Subplots - Multiple Plots in One Figure
#### Creating Subplots 

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(2, 2, figsize=(8, 6))

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sine")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("Cosine")

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title("Tanget")

axes[1, 1].plot(x, x)
axes[1, 1].set_title("Linear")

plt.tight_layout()
plt.show()
````

#### `tight_layout()` vs `constrained_layout`
`tight_layout()`
- Adjusts spacing automatically
- Most commonly used
```python
plt.tight_layout()
```

`constrained_layout=True`
- Better for complex figures
```python
fig, ax = plt.subplots(contrained_layout=True)
```

Use one, **not both**

### Axis Control
#### Setting Axis Limits
```python
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)
```

#### Custom Ticks
```python
ax.set_xticks([0, 2, 4, 6, 8, 10])
ax.set_yticks([-1, 0, 1])
```

#### Custom Tick Labels
```python
ax.set_xticklabels(["A", "B", "C", "D", "E", "F"])
```

### Grid Control
```python
ax.grid(True)
```

**Advanced:**
```python
ax.grid(
    True,
    linestyle="--",
    alpha=0.6
)
```

### Annotations & Text
#### Adding Text
```python
ax.text(5, 0.5, "Peak Value", fontsize=10)
```

#### Annotations with Arrows
```python
ax.annotate(
    "Maximum",
    xy=(np.pi/2, 1),
    xytext(4, 0.5),
    arrowprops=dict(arrowstyle="->")
)
```

### Legends - Control
```python
ax.plot(x, np.sin(x), label="sin(x)")
ax.plot(x, np.cos(x), label="cos(x)")
ax.legend()
```

**Advanced placement:**
```python
ax.legend(loc="upper right",frameon=False)
```

### Handling Multiple Lines Cleanly
```python
for power in range(1, 4):
    ax.plot(x, x**power, label=f"x^{power}")

ax.legend()
```


### Real Example

````python
# Data
months = np.arange(1, 13)
sales = [120, 135, 150, 145, 160, 170, 180, 175, 165, 155, 150, 160]

# Style Sheet
plt.style.use("seaborn-v0_8")

# Figure & Axes
fig, ax = plt.subplots(figsize=(14, 5), dpi=120)

# Main Plot
ax.plot(
    months,
    sales,
    color="steelblue",
    linewidth=2,
    marker="o",
    label="Monthly Sales"
)

# Axis Labels & Title
ax.set_title("Monthly Sales Trend - 2024", fontsize=14, fontweight="bold")
ax.set_xlabel("Month")
ax.set_ylabel("Sales (in $K)")

# Axis Limits
ax.set_xlim(1, 12)
ax.set_ylim(110, 190)

# Custom Ticks
ax.set_xticks(months)
ax.set_yticks([120, 140, 160, 180])

# Custom Tick Labels
ax.set_xticklabels(
    ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
     "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
)

# Grid Styling
ax.grid(
    True,
    linestyle="--",
    linewidth=0.6,
    alpha=0.6
)

# Legend Control
ax.legend(
    loc="upper left",
    frameon=False
)

# Annotations
max_sales = max(sales)
max_month = months[sales.index(max_sales)]

ax.annotate(
    f"Peak Sales: {max_sales}",
    xy=(max_month, max_sales),
    xytext=(max_month - 3, max_sales - 20),
    arrowprops=dict(arrowstyle="->", color="black"),
    fontsize=10
)

# Extra Text on Plot
ax.text(
    1.2,
    183,
    "Sales steadily increased mid-year",
    fontsize=9,
    color="dimgray"
)

# Layout Adjustment
plt.tight_layout()

# Save Figure
plt.savefig(
    "02_monthly_sales_trend.png",
    dpi=300,
    bbox_inches="tight"
)


plt.show()
````
