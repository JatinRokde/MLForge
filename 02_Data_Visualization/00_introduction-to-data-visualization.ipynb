### Why Data Visualization Matters
Visualization is not decoration. It is a **thinking tool**.

In production and ML workflows, visualization helps you:
- Understand data distributions
- Detect outliers and anomalies
- Validate assumptions
- Communicate insights to non-technical stakeholders
- Debug models (bias, variance, overfitting)

### Types of Data (Visualization Perspective)
#### A. Numerical Data
**Examples:**
- Age
- Salary
- Temperature
- Model Loss

**Characteristics:**
- Continuous or discrete
- Ordered
- Can be aggregated

**Best Plots:**
- Histogram
- Line Plot
- Box Plot
- KDE Plot
***

#### B. Categorical Data
**Examples:**
- Gender
- Department
- Product Category
- Model class labels

**Characteristics:**
- Finite values
- No inherent numerical meaning

**Best Plots:**
- Bar Chart
- Count Plot
- Pie Chart (rare)
***

#### C. Time-Series Data
**Examples:**
- Stock prices
- Website Traffic
- Training loss over epochs

**Characteristics:**
- Ordered by time
- Trend + seasonality + noise

**Best Plots:**
- Line Plot
- Area Plot
***

#### D. Relationship Data
**Examples:**
- Height vs Weight
- Feature vs Target
- Prediction vs Actual

**Best Plots:**
- Scatter Plot
- Pair Plot
- Heatmap (correlation)

### Choosing the Right Plot
| Goal                           | Best Plot      |
| ------------------------------ | -------------- |
| Distribution                   | Histogram, KDE |
| Compare categories             | Bar plot       |
| Trend over time                | Line plot      |
| Relationship between variables | Scatter plot   |
| Spread & outliers              | Box plot       |
| Correlation                    | Heatmap        |


### Matplotlib vs Seaborn
#### Matplotlib
Low-level
- Extremely flexible
- Full control
- Verbose but powerful

Use when:
- You need custom layouts
- Production reports
- Fine-grained control

#### Seaborn
High-level
- Statistical defaults
- Beautiful with less code
- Built on top of Matplotlib

Use when:
- EDA
- Statistical relationships
- Faster insights
