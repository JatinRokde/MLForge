### Why Were Transformers Needed?
Recall the limitations of RNNs/LSTMs:
```
Sequential Processing

Slow Training

Difficulty With Very Long Context

Limited Parallelization
```

Diagram:
```
Word1
  ↓
Word2
  ↓
Word3
  ↓
Word4
```

Each word depends on the previous one.

Cannot process all words simultaneously.

### The Big Idea
Researchers asked:
```
Do We Really Need RNNs?
```
What if we use:
```
Only Attention
```

The answer became: **Transformers**

##### *The Famous Paper*
*In 2017, researchers at Google published the paper:*

***Attention Is All You Need***

*The title itself reveals the idea:*
> Attention
> Is All You Need

*No RNNs.*

*No LSTMs.*

### What is a Transformer?
#### Definition
A Transformer is a deep learning architecture that uses attention mechanisms to process sequences and learn relationships between tokens.

### High-Level Architecture
Diagram:
```
Input Tokens
      ↓
Embedding
      ↓
Self-Attention
      ↓
Feed Forward Network
      ↓
Output
```

### Core Components
For your roadmap, remember:
```
Embeddings

Self-Attention

Multi-Head Attention

Feed Forward Network

Positional Encoding
```

These are the essential building blocks.

#### 1. Embeddings
Neural networks cannot understand words directly.

Example:
```
I love NLP
```

Convert words into vectors:
```
I      → [0.2, 0.8, ...]

love   → [0.5, 0.1, ...]

NLP    → [0.9, 0.3, ...]
```

Diagram:
```
Words
  ↓
Embeddings
  ↓
Vectors
```

#### 2. Self-Attention
This is the heart of Transformers.

Recall normal attention:
```
Decoder
   ↓
Looks At Encoder
```

Transformers introduce: **Self-Attention**

Each word looks at every other word.

##### Example
Sentence:
```
I love Deep Learning
```
When processing:
```
Learning
```
the model can attend to:
```
I

love

Deep

Learning
```
simultaneously.

Diagram:
```
I  ↔  love
↕      ↕
Deep ↔ Learning
```
Every word can interact with every other word.

##### Why Self-Attention Is Powerful
Consider:
```
The animal didn't cross the street
because it was tired.
```
Question:
```
What is "it"?
```
Self-attention can directly connect:
```
it
↓
animal
```
even if many words separate them.

##### RNN vs Self-Attention
**RNN**
```
Word1
 ↓
Word2
 ↓
Word3
 ↓
Word4
```
Sequential.

**Self-Attention**
```
Word1 ↔ Word2

Word1 ↔ Word3

Word1 ↔ Word4

Word2 ↔ Word4
```
Direct connections.

#### 3. Multi-Head Attention
One attention mechanism may focus on:
```
Grammar
```
Another may focus on:
```
Meaning
```
Another may focus on:
```
Relationships
```
Transformers use multiple attention heads.

Diagram:
```

Input
 ↓

Head 1
Head 2
Head 3
Head 4

 ↓

Combine
```

**Intuition**

Different heads learn different perspectives.

Like multiple experts reading the same sentence.

#### 4. Positional Encoding
Problem:

Attention alone doesn't know word order.

Sentence:
```
Dog bites man
```
vs
```
Man bites dog
```
Same words.

Different meaning.

Solution: **Positional Encoding**

Adds position information.

Diagram:
```
Word
  +
Position
  ↓
Input To Transformer
```

Now the model knows:
```
First Word

Second Word

Third Word
```

#### 5. Feed Forward Network
After attention:
```
Relationships Found
```
The transformer passes information through a normal neural network.

Diagram:
```
Attention
     ↓
Feed Forward Network
     ↓
Better Representation
```

##### Transformer Block
The fundamental unit:
```
Input
  ↓
Self-Attention
  ↓
Feed Forward Network
  ↓
Output
```

##### Stacking Transformer Blocks
Real models use many layers.
```
Input
 ↓

Transformer Block
 ↓

Transformer Block
 ↓

Transformer Block
 ↓

Output
```
This enables deep understanding.

### Why Transformers Won
#### 1. Parallel Processing
RNN:
```
Word1
 ↓
Word2
 ↓
Word3
```
Sequential.

Transformer:
```
Word1

Word2

Word3

Word4
```
Processed simultaneously.

Much faster.

#### 2. Long-Range Dependencies
Can directly connect:
```
Word1
```
and
```
Word100
```
through attention.

#### 3. Better Scaling
More data + more compute:
```
Better Performance
```
This scaling property led to LLMs.

### Transformer Family

#### Encoder-Based
Examples:

Used for:
```
Classification

Search

Understanding
```

#### Decoder-Based
Examples:

#### Encoder-Decoder
Examples:

Used for:
