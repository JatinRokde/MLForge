### Why Was Attention Needed?
Recall the Encoder-Decoder architecture:
```
Input Sentence
        ↓
     Encoder
        ↓
 Context Vector
        ↓
     Decoder
        ↓
Output Sentence
```

Problem:
```
Entire Sentence
       ↓
Single Context Vector
```

For long sentences:
```
Information Lost
```

This became known as the: **Bottleneck Problem**

##### Example
Input:
```
The movie that I watched
last month with my friends
was absolutely amazing.
```
The encoder must compress:
```
All Words
```
into:
```
One Vector
```
Difficult.

### Intuition Behind Attention
Instead of forcing the decoder to use:
```
One Context Vector
```
why not allow it to:
```
Look Back
At Important Words
```
whenever needed?

This idea became: **Attention**

### Core Idea
Instead of:
```
One Fixed Context Vector
```
the decoder dynamically decides:
```
Which Input Words
Are Important Right Now
```

### Visual Diagram
Without Attention:
```
Input Sentence
       ↓

   Encoder
       ↓

[ One Vector ]
       ↓

   Decoder
```

With Attention:
```
Input Words

h1  h2  h3  h4  h5

 \   |   |   |  /
  \  |   |   | /
   \ |   |   |/

    Attention

        ↓

     Decoder
```



Decoder can access all encoder states.

##### Example: Translation
Input:
```
I love NLP
```
Suppose decoder is generating:
```
NLP
```
Attention may focus heavily on:
```
NLP
```
and less on:
```
I
love
```

### Attention Scores
The model assigns importance scores.

Example:
```
I       → 0.1

love    → 0.2

NLP     → 0.7
```

Diagram:
```
Input Words

I      love      NLP
|        |        |
0.1      0.2      0.7
```

Higher score: **More Attention**

### Attention Weights
These scores become: **Attention Weights**

Example:
```
I       → 10%

love    → 20%

NLP     → 70%
```

The model focuses mostly on:
```
NLP
```

### Context Vector with Attention
Old Seq2Seq:
```
One Context Vector
For Entire Sentence
```

Attention:
```
New Context Vector
For Every Output Word
```

Diagram:
```
Output Word 1
       ↓
Context Vector 1

Output Word 2
       ↓
Context Vector 2

Output Word 3
       ↓
Context Vector 3
```
Much more flexible.

### Why Attention Solves the Bottleneck
Old approach:
```
Input
 ↓

Single Vector
 ↓

Output
```

Attention:
```
Input
 ↓

All Encoder States
 ↓

Choose Important Parts
 ↓

Output
```
Information is no longer forced through a single bottleneck.

##### Example
Input:
```
The cat sat on the mat
```
When generating:
```
cat
```
Attention focuses on:
```
cat
```
When generating:
```
mat
```
Attention shifts to:
```
mat
```

### Visual Attention Map
```
Output Word

      ↓

cat  sat  mat

0.8  0.1  0.1
```

Focus on:
```
cat
```

Later:
```
cat  sat  mat

0.1  0.1  0.8
```
Focus shifts.

### Why This Was Revolutionary
Before Attention:
```
Long Sentences
      ↓
Poor Performance
```

After Attention:
```
Long Sentences
      ↓
Much Better Performance
```

Machine translation quality improved dramatically.

### Key Benefits
#### 1. Better Long-Range Dependencies
Instead of remembering information through many time steps:
```
Word1
 ↓
Word2
 ↓
Word3
 ↓
...
 ↓
Word100
```
the model can directly attend to:
```
Word1
```
when needed.

#### 2. Solves Memory Bottleneck
No need for:
```
Single Context Vector
```

#### 3. Better Interpretability
Attention weights show:
```
What The Model Focused On
```

### Attention in Encoder-Decoder
Updated architecture:
```
Input
  ↓

Encoder
  ↓

Attention
  ↓

Decoder
  ↓

Output
```

Historical Evolution
```
RNN
 ↓
LSTM
 ↓
GRU
 ↓
Seq2Seq
 ↓
Attention
```
This was the breakthrough that changed NLP.

### The Famous Paper
In 2014, researchers introduced attention for machine translation.

This became one of the most influential ideas in Deep Learning.
