So far:
```
RNN
 ↓
LSTM
 ↓
GRU
```
These models can process sequences.

But if want to solve problems like:
```
English Sentence
        ↓
French Sentence
```
or
```
Long Article
      ↓
Summary
```
Notice:
```
Input = Sequence

Output = Sequence
```

This led to: **Sequence-to-Sequence (Seq2Seq) Models**

### Why Do We Need Seq2Seq?
Traditional RNN classification:
```
Sentence
    ↓
Positive
```
Input:
```
Sequence
```
Output:
```
Single Value
```
Many NLP tasks require:
```
Sequence
     ↓
Sequence
```
instead.

#### Examples
##### Machine Translation
```
English
 ↓
French
```
Example:
```
I love NLP
```
to
```
J'aime le NLP
```

##### Summarization
```
Long Article
      ↓
Short Summary
```

##### Chatbots
```
Question
    ↓
Answer
```

##### Text Generation
```
Prompt
   ↓
Generated Text
```

### Core Idea
Seq2Seq converts:
```
Input Sequence
        ↓
Output Sequence
```
Diagram:
```
Input Words
      ↓
Seq2Seq
      ↓
Output Words
```

### Encoder-Decoder Architecture
The key innovation.

A Seq2Seq model has two parts:
```
Encoder

Decoder
```

Diagram:
```
Input Sentence
        ↓
     Encoder
        ↓
 Context Vector
        ↓
     Decoder
        ↓
Output Sentence
```

### Encoder
##### Purpose
Read and understand the input sequence.

Example:
```
I love NLP
```

Encoder processes:
```
I
↓
love
↓
NLP
```

one token at a time.

### Encoder Memory
During processing:
```
Word1
 ↓
Memory

Word2
 ↓
Memory

Word3
 ↓
Memory
```
Finally produces:
```
Context Vector
```

### Context Vector
This is the most important concept.

Think of it as:
```
Compressed Meaning
Of The Entire Sentence
```

Diagram:
```
I love NLP
      ↓
[Context Vector]
```

### Decoder
The decoder uses the context vector to generate the output sequence.

Diagram:
```
Context Vector
        ↓
      Decoder
        ↓
Output Words
```

### Example Translation
Input:
```
I love NLP
```
Encoder:
```
Understand Sentence
```
Context:
```
Sentence Meaning
```
Decoder:
```
Generate French
```
Output:
```
J'aime le NLP
```

### Visual Diagram
Input
```
I → love → NLP

        ↓

      Encoder

        ↓

   Context Vector

        ↓

      Decoder

        ↓

J' → aime → NLP
```

### Why This Was Revolutionary
Before Seq2Seq:
```
Input Length
=
Output Length
```
was often required.

Seq2Seq allows:
```
Short Input
Long Output
```
or
```
Long Input
Short Output
```

#### Example
Translation:
```
3 Words
     ↓
4 Words
```

Summarization:
```
100 Words
      ↓
10 Words
```

Chatbot:
```
Question
      ↓
Answer
```
Different lengths.

### Training Idea
During training:

Input:
```
I love NLP
```

Target:
```
J'aime le NLP
```

The decoder learns:
```
Next Word Prediction
```
repeatedly.

### Major Limitation
The encoder must compress:
```
Entire Sentence
```
into:
```
Single Context Vector
```

Diagram:
```
Long Sentence
       ↓
Single Vector
```

### Why Is This Problematic?
Consider:
```
Very Long Paragraph
```
The context vector must store:
```
Everything
```
inside one fixed-size representation.

Information gets lost.

#### Example
Sentence:
```
100 Words Long
```
Encoder:
```
Compress
```
into:
```
One Vector
```
Difficult.

### The Bottleneck Problem
Diagram:
```
Long Input
      ↓

   Encoder
      ↓

[ One Vector ]
      ↓

   Decoder
      ↓

Output
```
The single vector becomes a bottleneck.

### This Led To...
Researchers asked:
```
Why Force Decoder
To Use Only One Vector?
```

Instead:
```
Let Decoder Look Back
At Important Words
```

This idea became: **Attention**

### Evolution
```
RNN
 ↓
LSTM
 ↓
GRU
 ↓
Seq2Seq
 ↓
Attention
 ↓
Transformer
 ↓
GPT / BERT
```
Seq2Seq is a critical step in this journey.
