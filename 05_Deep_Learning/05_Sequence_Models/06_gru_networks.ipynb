LSTMs solved many issues, but they introduced:
```
More Parameters

More Gates

More Computation
```

Researchers then asked:
```
Can We Get
LSTM-Like Performance
With Simpler Architecture?
```

The answer was: **GRU (Gated Recurrent Unit)**

### What is a GRU?
#### Definition
A GRU is a simplified version of an LSTM that uses fewer gates while still handling long-term dependencies effectively.

Diagram:
```
RNN
 ↓
LSTM
 ↓
GRU
```

Think of GRU as:
```
Lightweight LSTM
```

### Why GRU Was Created
LSTM contains:
- Forget Gate
- Input Gate
- Output Gate
- Cell State
- Hidden State

This works well but increases complexity.

GRU simplifies this.

Diagram:
```
LSTM
 ↓
Many Components

GRU
 ↓
Fewer Components
```



### Main Idea
LSTM:
```
Long-Term Memory
(Cell State)

+

Hidden State
```

Two states.

GRU:
```
Single Memory State
```

Only one state.

Simpler.

### LSTM vs GRU
#### LSTM
```
Cell State (Ct)

Hidden State (ht)
```

#### GRU
```
Hidden State Only
```

No separate cell state.

### GRU Gates
Instead of three gates:
- Forget Gate
- Input Gate
- Output Gate

GRU uses only:
- Update Gate
- Reset Gate

### Visual Comparison
##### LSTM
```
Forget Gate

Input Gate

Output Gate
```

##### GRU
```
Update Gate

Reset Gate
```

Much simpler.

#### 1. Update Gate
##### Purpose
Decides:
```
How Much Old Information
Should Be Kept
```

and
```
How Much New Information
Should Be Added
```

Think of it as a combination of:
```
Forget Gate

+

Input Gate
```

from LSTM.

Diagram:

```text
Old Memory
      ↓
Update Gate
      ↓
New Memory
```

#### 2. Reset Gate
##### Purpose
Decides:
```
How Much Past Information
Should Be Ignored
```

when processing current input.

**Example**
```
Previous Context
```
may not be useful.

The reset gate can reduce its influence.

Diagram:
```
Past Memory
      ↓
Reset Gate
      ↓
Relevant Information
```

### Intuition
Imagine you're reading a document.

#### Update Gate
Asks:
```
Should I Remember This?
```

#### Reset Gate

Asks:
```
Should I Ignore
Some Old Information?
```

### Memory Flow
GRU memory:
```
Previous Memory
       ↓
Update Gate

Reset Gate
       ↓
New Memory
```



### Why GRU Works
Even though it has:
```
Fewer Gates
```

it still:

- Controls Information Flow
- Preserves Important Context
- Reduces Vanishing Gradients

similar to LSTM.

### Example
Sentence:
```
I grew up in France.

...

I speak fluent French.
```

GRU can preserve:

```
France
```
across many time steps, just like LSTM.


### Architecture Comparison
#### Vanilla RNN
```
Input
 ↓
Memory
 ↓
Output
```

Simple but weak memory.

#### LSTM
```
Input
 ↓

Forget Gate

Input Gate

Output Gate

 ↓

Cell State
```

Strong memory.

#### GRU
```
Input
 ↓
Update Gate

Reset Gate
 ↓
Memory
```

Simpler memory.

### Parameter Comparison
Approximate complexity:
```
RNN
 ↓
Fewest Parameters

GRU
 ↓
Moderate

LSTM
 ↓
Most Parameters
```

### Training Speed
Generally:
```
RNN
 ↓
Fastest

GRU
 ↓
Faster

LSTM
 ↓
Slightly Slower
```

because GRU has fewer computations.


### Performance Comparison
In many tasks:
```
GRU ≈ LSTM
```

Performance is often very similar.

This made GRUs popular.

### LSTM vs GRU
| LSTM | GRU |
|--------|--------|
| 3 Gates | 2 Gates |
| Cell State + Hidden State | Hidden State Only |
| More Parameters | Fewer Parameters |
| Slightly Slower | Faster |
| More Complex | Simpler |

### When to Use?
Historically:

#### LSTM
```
Need Maximum Capacity
```

#### GRU
```
Need Simplicity

Need Faster Training
```

Today:
```
Transformers
```

have largely replaced both.


### PyTorch Example
```python
import torch.nn as nn

gru = nn.GRU(
    input_size=100,
    hidden_size=256,
    num_layers=1
)
```

Very similar to:

```python
nn.LSTM(...)
```

### Historical Importance
Sequence model evolution:
```
RNN
 ↓
Vanishing Gradient
 ↓
LSTM
 ↓
GRU
 ↓
Attention
 ↓
Transformers
```

GRU was an intermediate step toward better sequence modeling.

### Why GRU Didn't End the Story
Even GRUs still process sequences:
```
One Step At A Time
```

Diagram:
```
Word1
 ↓
Word2
 ↓
Word3
 ↓
Word4
```

This causes:
```
Slow Training

Limited Long-Range Context
```

Researchers wanted something better.

That led to: **Attention**
