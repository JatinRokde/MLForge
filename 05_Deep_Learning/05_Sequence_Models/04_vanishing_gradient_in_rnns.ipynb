### Quick Recap
Consider an RNN:
```
x1 → h1 → h2 → h3 → h4 → Output
```

During training:
```
Output
  ↑
 h4
  ↑
 h3
  ↑
 h2
  ↑
 h1
```

Gradient flows backward.

### What is a Gradient?
Recall:
```
Gradient
```

tells us:
```
How Much To Update Weights
```

Large gradient:
```
Learn Quickly
```

Small gradient:
```
Learn Slowly
```

Zero gradient:
```
No Learning
```

### What is the Vanishing Gradient Problem?
#### Definition
During backpropagation, gradients become extremely small as they travel backward through many layers or time steps.

Diagram:
```
Output
 ↑

0.8
 ↑

0.4
 ↑

0.2
 ↑

0.1
 ↑

0.01
 ↑

0.0001
```

Eventually:
```
Almost Zero
```

### Why Does It Happen?
During BPTT:
```
Gradient
    ×
Gradient
    ×
Gradient
    ×
Gradient
```

repeated many times.

#### Example
Suppose:
```
0.5
```

is multiplied repeatedly.
```
0.5 × 0.5
= 0.25

0.25 × 0.5
= 0.125

0.125 × 0.5
= 0.0625
```

Continue:
```
0.000001
```

Eventually almost zero.

### Visual Intuition
```
Step 1 → 1.0

Step 2 → 0.5

Step 3 → 0.25

Step 4 → 0.125

Step 5 → 0.0625

Step 6 → 0.03125
```

Gradient rapidly disappears.

### In RNNs
Suppose:
```
x1 → x2 → x3 → x4 → x5 → x6
```

Prediction depends on:
```
x1
```

and
```
x6
```

During training:
```
Gradient
 ↓
x6
 ↓
x5
 ↓
x4
 ↓
x3
 ↓
x2
 ↓
x1
```

By the time it reaches:
```
x1
```

the gradient may be:
```
~0
```

### What Does This Mean?
The model cannot learn:
```
Long-Term Dependencies
```

because earlier words receive almost no learning signal.

#### Example: Long-Term Dependency
Sentence:
```
I grew up in France.

... many words later ...

I speak fluent French.
```

To understand:
```
French
```

the model should remember:
```
France
```

Vanilla RNN often forgets.
Diagram:
```
France
  ↓
Memory
  ↓
Memory
  ↓
Memory
  ↓
French

Information Lost
```

#### Another Example
Sentence:
```
The movie that I watched
three weeks ago
was amazing.
```

To understand:
```
amazing
```

the model should remember:
```
movie
```

Vanishing gradients make this difficult.

### Short-Term vs Long-Term Memory
Vanilla RNNs handle **Short-Term Dependencies**

Example:
```
I love NLP
```

Easy.

Vanilla RNNs Struggle With **Long-Term Dependencies**

Many words between related concepts.
```
Difficult.
```

### Why Activation Functions Contribute
Recall:
- Sigmoid
- Tanh

Outputs are bounded.

##### Sigmoid
```
0 → 1
```

##### Tanh
```
-1 → 1
```

Their derivatives are often:
```
Less Than 1
```

Repeated multiplication:
```
Small × Small × Small
```

causes gradients to vanish.

### Practical Consequence
The RNN starts behaving like:
```
Recent Words Matter

Older Words Ignored
```

Diagram:
```
Old Information
      ↓
Forgotten

Recent Information
      ↓
Remembered
```

### Why This Was a Big Problem
Early NLP systems needed:
- Context
- Memory
- Long-Term Relationships

Vanilla RNNs struggled.

This limited performance on:
- Translation
- Language Modeling
- Speech Recognition

# How Researchers Solved It

The solution: **LSTM (Long Short-Term Memory)**

Idea:
```
Keep Important Information

Forget Unimportant Information
```

using specialized memory mechanisms.

Diagram:
```
RNN
 ↓
Memory Problems
 ↓
LSTM
 ↓
Better Memory
```

### Historical Evolution
```
Feedforward Networks
         ↓
RNN
         ↓
Vanishing Gradient
         ↓
LSTM
         ↓
GRU
         ↓
Attention
         ↓
Transformers
```

The entire evolution of NLP was driven by solving this problem.
