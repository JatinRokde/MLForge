### What is a Sequence?
#### Definition
A sequence is data where: **Order Matters**

**Examples**

#### Text
```
I love dogs
```

vs
```
Dogs love I
```

Same words.

Different order.

Different meaning.

#### Time Series
```
100
105
110
115
```

Order contains information.

#### Speech
```
Audio Frame 1
Audio Frame 2
Audio Frame 3
```

The sequence matters.

### How Feedforward Networks Work
Recall MLPs:
```
Input
 ↓
Hidden Layer
 ↓
Output
```

**Example**

Features:
```
Age
Salary
Experience
```

Input:
```
[25, 50000, 3]
```

Order doesn't matter much.

### Why MLPs Work for Tabular Data
Each feature has a fixed meaning.
```
Position 1 → Age

Position 2 → Salary

Position 3 → Experience
```

Everything is fixed.

### Text Example
Sentence:
```
I love NLP
```

Represented as:
```
[I, love, NLP]
```

Suppose we flatten it:
```
x1 x2 x3
```

and send it to an MLP.

Diagram:
```
Words
 ↓
Vector
 ↓
MLP
 ↓
Output
```

Seems okay initially.

#### Problem 1: Fixed Input Length
Suppose training uses:
```
3 Words
```

Example:
```
I love NLP
```

Now input:
```
I really love NLP
```

contains:
```
4 Words
```

Problem:
```
Input Size Changed
```

MLPs expect:
```
Fixed Input Dimensions
```

##### Visual Diagram
```
Training

Word1
Word2
Word3

      ↓

Fixed Architecture
```

Later
```
Word1
Word2
Word3
Word4
```

Network cannot naturally handle it.

#### Problem 2: No Memory
Consider:
```
The movie was not good
```

The word:
```
not
```

changes meaning.

Feedforward network processes:
```
Input
 ↓
Output
```

without remembering previous words.

Diagram
```
Word 1

Word 2

Word 3

Word 4

No Memory
```

### Why Is Memory Important?

Consider:
```
I grew up in France.
I speak fluent French.
```

To understand:
```
French
```

the model may need information from earlier words.

MLPs lack:
```
Context Memory
```

#### Problem 3: Context Dependency
Sentence:
```
The bank approved my loan.
```

Meaning:
```
Financial Bank
```

Sentence:
```
I sat near the river bank.
```

Meaning:
```
River Bank
```

Same word:
```
bank
```

Different meanings.

Context determines interpretation.

Feedforward networks struggle to model such dependencies.

#### Problem 4: Order Information
Consider:
```
Dog bites man
```

versus
```
Man bites dog
```

Same words.

Completely different meaning.

MLPs are not naturally designed for:
```
Sequential Relationships
```

#### Problem 5: Shared Patterns
Suppose to learn:
```
I love NLP
```

and
```
I love Deep Learning
```

Notice:
```
I love ...
```

appears repeatedly.

Ideally we want:
```
Reuse Knowledge
Across Positions
```

MLPs use:
```
Different Weights
For Different Inputs
```

Not ideal for sequences.

### What is Needed
A model that:
- Handles Variable Length
- Remembers Previous Inputs
- Uses Context
- Understands Order
- Shares Knowledge Across Time

### Enter RNNs
Recurrent Neural Networks solve this by introducing: **Memory**



### RNN Idea
Instead of:
```
Word1
Word2
Word3
```

being processed independently,

RNN processes:
```
Word1
  ↓
Memory

Word2
  ↓
Memory

Word3
  ↓
Memory
```

Diagram
```
Word1
  ↓
RNN
  ↓
Memory

Word2
  ↓
RNN
  ↓
Memory

Word3
  ↓
RNN
```

### Comparison
#### Feedforward Network
```
Input
 ↓
Output
```

No memory.

#### RNN
```
Input
 ↓
Memory
 ↓
Next Input
 ↓
Memory
```

Sequence-aware.
