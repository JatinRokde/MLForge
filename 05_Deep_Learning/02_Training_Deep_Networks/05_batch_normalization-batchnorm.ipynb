### Batch Normalization (BatchNorm)
#### Definition
Batch Normalization is a technique that normalizes the activations of a layer during training to make learning more stable and faster.

Introduced to solve:
```
Deep Networks
      ↓
Unstable Activations
      ↓
Slow / Difficult Training
```

### Why Was BatchNorm Needed?
As data passes through multiple layers:
```
Input
  ↓
Layer 1
  ↓
Layer 2
  ↓
Layer 3
  ↓
Output
```

the distribution of activations keeps changing.

**Example:**
```
Layer 1 Output:
Mean ≈ 0

Layer 2 Output:
Mean ≈ 10

Layer 3 Output:
Mean ≈ 100
```

This makes training unstable.

### Intuition
Think of students taking an exam.

Without normalization:
```
Student A Score = 20/100
Student B Score = 90/100
Student C Score = 50/100
```

Scores have very different scales.

Normalize:
```
Mean = 0
Std = 1
```

Now everyone is on a comparable scale.

BatchNorm does the same for neuron activations.

### How It Works
For each mini-batch:
```
Layer Output
      ↓
Normalize
      ↓
Mean ≈ 0
Std ≈ 1
      ↓
Pass to Next Layer
```

**Flow:**
```
Input
  ↓
Linear Layer
  ↓
BatchNorm
  ↓
Activation
  ↓
Next Layer
```

**Modern architecture:**
```
Linear
   ↓
BatchNorm
   ↓
ReLU
```

### What Does It Normalize?
Suppose a batch contains:
```
[10, 20, 30, 40, 50]
```

Mean:
```
30
```

Standard Deviation:
```
14.14
```

Normalize:
```
(value - mean) / std
```

Result:
```
[-1.41, -0.71, 0, 0.71, 1.41]
```

Now:
```
Mean ≈ 0
Std ≈ 1
```

### Benefits of BatchNorm
#### 1. Faster Training

Without BatchNorm:
```
100 Epochs
```

With BatchNorm:
```
30-50 Epochs
```
(often, though not always)

#### 2. More Stable Training
Before:
```
Loss

100
 80
 60
 40
 20
```

Loss may fluctuate.

After BatchNorm:
```
Loss

100
 70
 50
 30
 15
```

Typically smoother convergence.

#### 3. Reduces Vanishing Gradients
Because activations remain in a healthy range.
```
Healthy Activations
        ↓
Healthier Gradients
```

#### 4. Allows Higher Learning Rates
Without BatchNorm:
```
lr = 0.01
```
may diverge.

With BatchNorm:
```
lr = 0.1
```
might still train successfully.

#### 5. Slight Regularization Effect
Because each mini-batch has different statistics:
```
Batch 1
Batch 2
Batch 3
```

the model sees slight noise.

This can reduce overfitting somewhat.

### Where Is BatchNorm Used?
Very common in:
* CNNs
* Deep Feedforward Networks
* ResNet

Historically one of the biggest breakthroughs in Computer Vision.

### Does BatchNorm Solve Everything?
No.

It helps with:
* Training Stability
* Convergence Speed
* Gradient Flow

But does not eliminate:
* Overfitting
* Poor Data
* Bad Architecture

### BatchNorm vs No BatchNorm
**Without BatchNorm**
```
Layer 1 → Normal

Layer 2 → Large Values

Layer 3 → Huge Values

Layer 4 → Unstable
```

**With BatchNorm**
```
Layer 1 → Controlled

Layer 2 → Controlled

Layer 3 → Controlled

Layer 4 → Controlled
```

### PyTorch Example
```python 
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(128, 64),
    nn.BatchNorm1d(64),
    nn.ReLU()
)
```

For CNNs:

```python 
nn.BatchNorm2d(...)
```

is commonly used.
