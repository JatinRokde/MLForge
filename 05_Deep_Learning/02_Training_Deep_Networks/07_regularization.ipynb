### Regularization
#### Definition
Regularization refers to techniques used to prevent a model from becoming overly complex and overfitting the training data.

Goal:
```
Good Training Performance
          +
Good Test Performance
```

### Why Do We Need Regularization?
Deep Neural Networks have:
```
Thousands
Millions
Billions
```
of parameters.

This gives them huge learning capacity.

Unfortunately:
```
High Capacity
      ↓
Memorization
      ↓
Overfitting
```

### Overfitting Reminder
```
Training Accuracy = 99%

Test Accuracy = 80%
```

Model memorizes training data instead of learning general patterns.

### Types of Regularization
1. L1 Regularization
2. L2 Regularization (Weight Decay)
3. Early Stopping
4. Dropout 


#### 1. L2 Regularization (Weight Decay)
##### Core Idea
Penalize large weights.

Without L2:
```
Weight = 100
Weight = 200
Weight = 500
```

The network may become overly dependent on a few features.

With L2:
```
Large Weights
      ↓
Penalty
      ↓
Smaller Weights
```

##### Intuition
Imagine a student solving a problem.

Without L2:
```
Uses One Shortcut
For Everything
```

With L2:
```
Uses Multiple Clues
```

More balanced learning.

##### Effect
Before:
```
Feature A Weight = 500
Feature B Weight = 2
Feature C Weight = 1
```

After L2:
```
Feature A Weight = 20
Feature B Weight = 10
Feature C Weight = 8
```

Model becomes more robust.

##### Why Called Weight Decay?
During training:
```
Weights
   ↓
Gradually Shrink
```

Hence:
```
Weight Decay
```

This is the most common regularization technique today.

Used in:
* CNNs
* Transformers
* BERT
* GPT
* LLaMA

Modern LLM training heavily relies on:
```
AdamW Optimizer
```

where:
```
W = Weight Decay
```

#### 2. L1 Regularization
##### Core Idea
Also penalizes weights.

But differently.

L1 pushes many weights to:
```
Exactly Zero
```

##### Before
```
[10, 8, 5, 2, 1]
```

##### After L1
```
[10, 7, 0, 0, 0]
```

##### Effect
Automatically performs:
```
Feature Selection
```

because unimportant weights become zero.

##### When Used?
Less common in Deep Learning.

More common in traditional ML.

##### L1 vs L2
| Property             | L1  | L2  |
| -------------------- | --- | --- |
| Shrinks Weights      | Yes | Yes |
| Produces Exact Zeros | Yes | No  |
| Feature Selection    | Yes | No  |
| Most Common in DL    | No  | Yes |

#### 3. Early Stopping
One of the simplest and most effective techniques.

##### Idea
Stop training before overfitting begins.

Training:
```
Epoch 1
Epoch 2
Epoch 3
...
Epoch 100
```

Do not automatically assume:
```
More Epochs = Better
```

##### Example
Validation Loss:
```
Epoch 10 → 0.40
Epoch 20 → 0.25
Epoch 30 → 0.18
Epoch 40 → 0.15
Epoch 50 → 0.14
Epoch 60 → 0.15
Epoch 70 → 0.18
```

Best model:
```
Epoch 50
```

Stop there.

##### Why It Works
After a certain point:
```
Model Stops Learning Patterns
         ↓
Starts Memorizing Noise
```

Early stopping prevents this.

#### 4. Dropout 
##### Idea
```
Randomly Disable Neurons
```

##### Diagram
```
● X ● ●
● ● X ●
X ● ● ●
```

##### Purpose
```
Reduce Overfitting
```

### Which Regularization Is Most Important?
For modern Deep Learning:
* L2 (Weight Decay)
* Early Stopping
* Dropout
