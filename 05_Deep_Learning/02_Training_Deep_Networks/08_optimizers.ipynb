### Optimizers
#### Definition
An Optimizer is an algorithm that uses gradients computed by Backpropagation to update model weights efficiently.

Recall:
```
Backpropagation
      ↓
Gradients
      ↓
Optimizer
      ↓
Updated Weights
```

The optimizer decides:
* How large the update should be
* How fast learning should happen
* How stable training should be

### Why Do We Need Optimizers?
Basic Gradient Descent:
```
w = w - lr × gradient
```

works, but has limitations:
* Slow Convergence
* Sensitive To Learning Rate
* Can Get Stuck
* Can Oscillate

Modern optimizers improve this.

### 1. SGD (Stochastic Gradient Descent)
#### Idea
Use gradients from mini-batches to update weights.

Update rule:
```
w = w - lr × gradient
```

Same formula as basic Gradient Descent.

Path:
```
Zig-Zag
```

because each mini-batch gives slightly different gradients.

#### Pros
* Simple
* Memory Efficient

#### Cons
* Slow
* Noisy Updates

### 2. Momentum
#### Problem With SGD
SGD may oscillate.

#### Solution
Add momentum.

Think of a rolling ball.

The ball builds speed in the correct direction.

#### Intuition

Instead of looking only at:
```
Current Gradient
```

also consider:
```
Previous Updates
```

Result:
```
Smoother
Faster
```
training.

#### Pros
* Faster Than SGD
* Less Oscillation

### 3. RMSProp
#### Problem

Different weights may need different learning speeds.

#### Example
```
Weight A → Large Updates
Weight B → Small Updates
```
Using one learning rate for all weights may not be ideal.

#### Solution
RMSProp adapts the learning rate individually.

#### Diagram
```
Weight 1 → lr = small
Weight 2 → lr = large
```

#### Benefits
* Adaptive Learning Rate
* Faster Convergence

Historically important for:

* RNNs
* Sequence Models

### 4. Adam
### Most Important Optimizer
Adam combines:
```
Momentum
      +
RMSProp
```

##### Diagram
```
Momentum
     +
Adaptive LR
     ↓
Adam
```

#### Why It Became Popular
Provides:
* Fast Training
* Stable Training
* Minimal Tuning

Works well in most situations.


If unsure:
```
Use Adam
```

It is often the first optimizer practitioners try.

### 5. AdamW
#### Modern Improvement Over Adam
Problem:
```
Adam + L2
```

was not implemented optimally.

#### Solution
```
AdamW
```

where:
```
W = Weight Decay
```

is handled correctly.

#### Why AdamW Matters
Modern architectures frequently use:

* Transformers
* BERT
* GPT
* LLaMA

and most are trained using:
```
AdamW
```

### Optimizer Evolution
```
Gradient Descent
       ↓
SGD
       ↓
Momentum
       ↓
RMSProp
       ↓
Adam
       ↓
AdamW
```

### Quick Comparison

| Optimizer | Main Idea                  |
| --------- | -------------------------- |
| SGD       | Basic updates              |
| Momentum  | Uses past direction        |
| RMSProp   | Adaptive learning rates    |
| Adam      | Momentum + RMSProp         |
| AdamW     | Adam + Proper Weight Decay |


### PyTorch Examples
#### SGD
```python 
optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.01
)
```

#### Momentum SGD
```python
optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.01,
    momentum=0.9
)
```

#### Adam
```python
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)
```

#### AdamW
```python
optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=0.001,
    weight_decay=0.01
)
```
