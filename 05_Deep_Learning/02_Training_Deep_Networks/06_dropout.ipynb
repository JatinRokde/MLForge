### Dropout
#### Definition
Dropout is a regularization technique where a random subset of neurons is temporarily disabled during training.

Goal:
```
Reduce Overfitting
      ↓
Improve Generalization
```

### Why Do We Need Dropout?
Deep neural networks have a huge number of parameters.

This gives them the ability to:
```
Learn Patterns
```

but also:
```
Memorize Training Data
```

This leads to: **Overfitting**


#### Overfitting Example
```
Training Accuracy = 99%

Test Accuracy = 75%
```

The model performs well on training data but poorly on unseen data.

**Diagram:**
```
Training Data
      ↓
Model Memorizes
      ↓
Poor Generalization
```

### Core Idea of Dropout
During training:
```
Randomly Turn Off Some Neurons
```

**Diagram:**



Where:
```
X = Temporarily Disabled
```

### Important Point
The neurons are removed:
```
Only During Training
```

Not during inference.

### Intuition
Imagine preparing for an exam.

Without Dropout:
```
Student Always Depends
On Same Friend
```

If the friend is absent:
```
Performance Drops
```

With Dropout:
```
Different Friends Missing
Each Day
```

The student learns independently.

Similarly:
```
Neurons Cannot Depend
Too Much On Other Neurons
```

### How Dropout Works
Suppose:
```
Dropout Rate = 0.5
```

Meaning:
```
50% of neurons
are randomly disabled
during each training step
```

#### Diagram
**Original**
```
● ● ● ● ● ●
```

**After Dropout**
```
● X ● X X ●
```

**Next batch**
```
● ● X ● X X
```

The pattern changes every iteration.

### Why Does It Help?
Without Dropout:
```
Neuron A
      ↓
Strongly Depends
On Neuron B
```

Network becomes fragile.

With Dropout:
```
Neuron B Might Disappear
```

So Neuron A learns more robust features.

Result:
```
Better Generalization
```

### Typical Dropout Rates
| Layer Type     | Common Rate |
| -------------- | ----------- |
| Dense Layers   | 0.2 – 0.5   |
| Large Networks | 0.5         |
| Small Networks | 0.1 – 0.3   |

Example:
```
Dropout(0.5)
```

means:
```
50% neurons dropped
```

during training.

### Training vs Inference
#### Training
Random Neurons Removed

**Diagram**
```
● X ● ●
● ● X ●
```

#### Inference
All Neurons Active

**Diagram**
```
● ● ● ●
● ● ● ●
```

This ensures maximum predictive power.

### Benefits
#### 1. Reduces Overfitting
Main reason it exists.

#### 2. Improves Generalization
Model performs better on unseen data.

#### 3. Acts Like Ensemble Learning
Every batch effectively trains a slightly different network.

**Diagram**
```
Network 1
Network 2
Network 3
Network 4
```

Dropout approximates training many networks and averaging them.

### Drawbacks
#### Slower Training
Because learning becomes harder.

#### Too Much Dropout
Example:
```
Dropout = 0.9
```

90% neurons removed.

Result:
```
Network Cannot Learn Properly
```

### Dropout vs BatchNorm
| BatchNorm              | Dropout              |
| ---------------------- | -------------------- |
| Stabilizes training    | Reduces overfitting  |
| Normalizes activations | Removes neurons      |
| Helps optimization     | Helps generalization |


### PyTorch Example
```python 
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(128, 64),
    nn.ReLU(),
    nn.Dropout(0.5)
)
```

Meaning:
```
50% neurons dropped
during training
```
