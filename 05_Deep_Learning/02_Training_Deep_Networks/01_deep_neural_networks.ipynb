### Multi-Layer Neural Networks (Deep Neural Networks)
#### Definition
A neural network containing multiple hidden layers.

**Single-layer network:**
```
Input
  ↓
Hidden
  ↓
Output
```

**Deep Neural Network:**
```
Input
  ↓
Hidden
  ↓
Hidden
  ↓
Hidden
  ↓
Output
```

### Why Do We Need More Layers?
A single neuron can learn:
```
OR
AND
Simple Linear Patterns
```
But real-world problems are much m
* Language understanding
* Image recognition
* Speech processing
* Reasoning

One hidden layer often lacks sufficient representation power.

### Core Idea
Each layer learns progressively more abstract features.

**Example: Image Classification**
```
Image
  ↓
Edges
  ↓
Shapes
  ↓
Object Parts
  ↓
Object
  ↓
Prediction
```

**Example: NLP**
```
Words
  ↓
Phrases
  ↓
Sentence Meaning
  ↓
Context
  ↓
Prediction
```

This is called: **Hierarchical Feature Learning**

and is one of the main reasons Deep Learning works.

### Width vs Depth
#### Width

More neurons in a layer.
```
● ● ● ● ● ● ● ●
```

#### Depth
More hidden layers.
```
●
↓
●
↓
●
↓
●
```

Deep Learning focuses primarily on:
```
Increasing Depth
```

rather than simply increasing width.

### Representation Learning
Each hidden layer creates a new representation of the data.

**Example:**
```
Raw Text
```
Layer 1:
```
Words
```
Layer 2:
```
Grammar
```
Layer 3:
```
Meaning
```
Layer 4:
```
Context
```
Output:
```
Prediction
```

This is the foundation of:
* Transformers
* BERT
* GPT
* LLaMA

### Problem Introduced by Deep Networks
Adding more layers increases power.

But also introduces new problems:
```
Deep Network
      ↓
Harder Training
      ↓
Gradient Problems
```

Specifically:
* Vanishing Gradients
* Exploding Gradients

These became major obstacles in early Deep Learning.
