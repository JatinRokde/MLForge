### Hyperparameter Tuning
#### Definition
Hyperparameter Tuning is the process of finding the best configuration of parameters that control the training process.

Unlike model weights:
```
Weights
Biases
```

which are learned automatically,

hyperparameters are chosen by the developer.

### Parameters vs Hyperparameters
#### Parameters
Learned during training.

Examples:
```
Weights
Biases
```

#### Hyperparameters
Set before training.

Examples:
```
Learning Rate
Batch Size
Number of Layers
Dropout Rate
Optimizer
Weight Decay
Epochs
```

### Why Is It Important?
Two models can have:
```
Same Dataset
Same Architecture
```

but perform very differently because of hyperparameters.

Example:
```
Model A Accuracy = 82%

Model B Accuracy = 92%
```

Difference:
```
Better Hyperparameters
```

### Most Important Hyperparameters
For practical Deep Learning, focus on:

#### 1. Learning Rate
Most important hyperparameter.

Too small:
```
Training Very Slow
```

Too large:
```
Training Unstable
```

Typical starting values:
```
Adam/AdamW → 1e-3

Fine Tuning → 1e-5 to 5e-5
```

#### 2. Batch Size
Number of samples processed before updating weights.

Example:
```
Batch Size = 32
```

Diagram:
```
Dataset
 ↓

[32 Samples]
      ↓
Update

[32 Samples]
      ↓
Update
```

##### Small Batch Size
```
16
32
64
```

Pros:
```
Less Memory
Better Generalization
```

Cons:
```
Noisier Training
```

##### Large Batch Size
```
512
1024
4096
```

Pros:
```
Faster GPU Utilization
```

Cons:
```
High Memory Usage
```

##### Rule of Thumb
Start with:
```
32
64
128
```

#### 3. Number of Layers
Shallow:
```
Input
 ↓
Hidden
 ↓
Output
```

Deep:
```
Input
 ↓
Hidden
 ↓
Hidden
 ↓
Hidden
 ↓
Output
```

More layers:
```
More Capacity
```

but also:
```
Harder Training
```

#### 4. Number of Neurons
Example:
```
64
128
256
512
```

More neurons:
```
More Learning Capacity
```

but:
```
More Overfitting Risk
```

#### 5. Dropout Rate
Typical values:
```
0.2
0.3
0.5
```

Too small:
```
Little Regularization
```

Too large:
```
Model Struggles To Learn
```

#### 6. Weight Decay
L2 Regularization strength.

Common values:
```
0.01
0.001
0.0001
```

Too much:
```
Underfitting
```

Too little:
```
Overfitting
```

#### 7. Optimizer
Most common choices:
```
SGD
Adam
AdamW
```

For modern projects:
```
AdamW
```

is usually the default.

### Hyperparameter Search Methods
#### 1. Manual Search
Most common in practice.

Example:
```
lr = 0.001
```

Try:
```
0.0005
0.0001
0.005
```

and compare.

#### 2. Grid Search
Try every combination.

Example:
```
Learning Rate
[0.1, 0.01]

Batch Size
[32, 64]
```

Combinations:
```
0.1, 32
0.1, 64
0.01, 32
0.01, 64
```

Problem:
```
Becomes expensive quickly.
```

#### 3. Random Search
Instead of checking all combinations:
```
Randomly Sample
```

Example:
```
lr = 0.0031
batch_size = 48
dropout = 0.27
```
Surprisingly effective.

#### 4. Bayesian Optimization
Advanced approach.

Idea:
```
Use Previous Results
To Choose Better Trials
```

Popular tools:
* Optuna
* Hyperopt

For interviews:
```
Just know the concept.
```

### Practical Tuning Priority
When performance is poor:

Tune in this order:
```
1. Learning Rate
2. Batch Size
3. Weight Decay
4. Dropout
5. Architecture
```

Learning rate often gives the biggest improvement.


### Common Starting Point
For a typical DL project:
```
Optimizer    = AdamW

Learning Rate = 1e-3

Batch Size    = 32 or 64

Dropout       = 0.3

Weight Decay  = 0.01
```

Then tune from there.

### PyTorch Example
```python
optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=0.001,
    weight_decay=0.01
)

batch_size = 64

dropout = 0.3
```

These are hyperparameters.
