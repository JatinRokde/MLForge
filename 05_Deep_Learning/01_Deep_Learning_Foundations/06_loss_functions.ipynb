In the previous topic:
```
Input
   ↓
Neural Network
   ↓
Prediction
```

A critical question remains:
> How does the network know whether its prediction is correct or incorrect?

The answer is: **Loss Function**

A loss function measures the difference between:
```
Actual Value
     vs
Predicted Value
```

The objective of training is simple:
```
Minimize Loss
```

### What Is a Loss Function?
Definition:
> A loss function is a mathematical function that measures how wrong a machine learning model's predictions are compared to the actual values.

**Flow:**
```
Actual Value
    ↓
Prediction
    ↓
Compare
    ↓
Loss
```

**Example:**

| Actual | Predicted |
| ------ | --------- |
| 100    | 90        |

Loss:
```
Error = 10
```

Small loss:
```
Good Prediction
```

Large loss:
```
Bad Prediction
```

### Why Loss Functions Are Needed
Imagine learning to play cricket.

After every shot:
```
Shot Played
      ↓
Feedback
      ↓
Improvement
```

Without feedback:
```
No Improvement
```

Similarly:
```
Prediction
      ↓
Loss
      ↓
Feedback
      ↓
Weight Update
```

Loss acts as feedback for the neural network.

#### Neural Network Training Loop
This entire roadmap eventually leads to understanding this loop:
```
Input Data
     ↓
Forward Propagation
     ↓
Prediction
     ↓
Loss Calculation
     ↓
Backpropagation
     ↓
Weight Update
     ↓
Better Prediction
```

Loss is the bridge between:
```
Prediction
      ↓
Learning
```

### Loss vs Cost Function
These terms are often used interchangeably.

#### Loss
Error for one sample.

Example:
```
Customer 1
```

Prediction:
```
90
```

Actual:
```
100
```

Loss:
```
10
```

#### Cost
Average loss over many samples.

Concept:
```
Sample 1 Loss
Sample 2 Loss
Sample 3 Loss
Sample 4 Loss
       ↓
Average
       ↓
Cost
```

In practice:
```
Loss ≈ Cost
```
for most discussions.

### Part A: Regression Losses
#### Mean Squared Error (MSE)
Most common regression loss.

**Formula:**
$$
\text{MSE} = \frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2
$$

Where:
```
y  = Actual Value
ŷ  = Predicted Value
```

**Example**

Actual:
```
100
```

Prediction:
```
90
```

Error:
```
100 - 90 = 10
```

Square:
```
10² = 100
```

Loss:
```
100
```

**Why Square?**

Without squaring:
```
+10
-10
```
would cancel.

Squaring ensures:
```
Positive Error
Negative Error
```

both contribute positively.

##### MSE Characteristics
**Advantages:**
- Smooth
- Differentiable
- Strongly penalizes large mistakes

**Disadvantages:**
- Sensitive to outliers

**Example:**

Errors:
```
2
3
4
100
```

Squared:
```
4
9
16
10000
```
The outlier dominates.

#### Mean Absolute Error (MAE)
**Formula:**
$$
\text{MAE} = \frac{1}{n}\sum_{i=1}^{n}\left|y_i-\hat{y}_i\right|
$$

**Example**

Errors:
```
10
-10
20
```

Absolute:
```
10
10
20
```

Average:
```
13.33
```


**MSE vs MAE**
| Property          | MSE    | MAE         |
| ----------------- | ------ | ----------- |
| Squares Error     | Yes    | No          |
| Outlier Sensitive | High   | Low         |
| Smooth Gradient   | Better | Less Smooth |
| Most Common in DL | Yes    | Less        |

### Part B: Classification Losses
##### Why MSE Is Not Ideal For Classification
Suppose:
```
Spam Probability
```

Prediction:
```
0.95
```

Actual:
```
1
```

MSE can work.

But probability-based learning requires a better loss.

This led to: **Cross Entropy**


#### Binary Cross Entropy (BCE)
Used for:
```
Two Classes
```

Examples:
* Spam / Not Spam
* Fraud / Not Fraud
* Yes / No

**Output Layer**

Usually:
```
Sigmoid
```

Output:
```
0 → 1
```
Probability.


#### Categorical Cross Entropy (CCE)
Used for:
```
Multiple Classes
```

Examples:
* Cat
* Dog
* Horse

or
* Positive
* Neutral
* Negative

## Output Layer:
```
Softmax
```

#### Softmax
Softmax converts outputs into probabilities.

**Example:**

Before Softmax:
```
[2.5, 4.1, 1.2]
```

After Softmax:
```
[0.15, 0.80, 0.05]
```

Properties:
* All probabilities between 0 and 1
* Sum = 1
