In the previous topic:

Gradient Descent was learned.

Weight Update
```
w = w - lr × gradient
```

A crucial question remains:
> Where does the gradient come from?

The answer is: **Backpropagation**

Backpropagation is the algorithm used to compute gradients for every weight in a neural network.

Without Backpropagation:
```
No Gradients
      ↓
No Weight Updates
      ↓
No Learning
```

### Big Picture
A neural network learns using two phases:

#### Phase 1: Forward Propagation
```
Input
  ↓
Prediction
  ↓
Loss
```

#### Phase 2: Backpropagation
```
Loss
  ↑
Gradients
  ↑
Weights
```

**Complete Flow**
```
Input
  ↓
Forward Pass
  ↓
Prediction
  ↓
Loss
  ↓
Backpropagation
  ↓
Gradients
  ↓
Weight Update
```
This cycle repeats thousands of times.

### The Chain Rule
Backpropagation is built on one concept: **Chain Rule**

Suppose:
```
A affects B

B affects C

C affects D
```

**Diagram**
```
A → B → C → D
```

Question:
```
How much does A affect D?
```

Answer:
```
Multiply all intermediate effects.
```
This is the Chain Rule.

#### Neural Network Version
```
Input
  ↓
Neuron
  ↓
Activation
  ↓
Loss
```

Question:
```
How much does a weight affect loss?
```
Chain Rule computes exactly that.

### Simplified Example
Suppose:
```
y = 2x
```

and
```
z = y²
```

Flow
```
x
↓
y = 2x
↓
z = y²
```

Question:
```
How much does x affect z?
```

Chain Rule:
```
dz/dx = dz/dy × dy/dx
```
This is the fundamental idea behind Backpropagation.

### What Backpropagation Produces
The output is:
```
Gradient For Every Weight
```

**Example**
| Weight | Gradient |
| ------ | -------- |
| w1     | 0.8      |
| w2     | -0.3     |
| w3     | 1.2      |
| w4     | -0.7     |

Interpretation:

**Positive Gradient**
```
Weight should decrease
```

**Negative Gradient**
```
Weight should increase
```

Gradient Descent then uses:
```
w = w - lr × gradient
```

### Why Deep Networks Become Difficult
Consider:
```
Input
 ↓
Layer 1
 ↓
Layer 2
 ↓
Layer 3
 ↓
Layer 4
 ↓
Output
```

Backpropagation must travel through:
```
Output
 ↑
Layer 4
 ↑
Layer 3
 ↑
Layer 2
 ↑
Layer 1
```

As networks become deeper:
```
More Layers
      ↓
Longer Gradient Path
```

This creates problems:
* Vanishing Gradients
* Exploding Gradients

### PyTorch Example
The beauty of PyTorch:
```text
No manual Backpropagation.
```

````python
import torch

x = torch.tensor(
    2.0,
    requires_grad=True
)

y = x ** 2

y.backward()

print(x.grad)
````

Because:

```text
y = x²

dy/dx = 2x

2 × 2 = 4
```
PyTorch computed the gradient automatically.
