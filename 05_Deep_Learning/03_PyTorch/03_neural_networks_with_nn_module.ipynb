### What is `nn.Module`?
#### Definition
`nn.Module` is the base class for all neural networks in PyTorch.

Think of it as:
```
Blueprint
   ↓
Neural Network
```

Almost every PyTorch model inherits from:
```python
torch.nn.Module
```

Examples:
* MLP
* CNN
* RNN
* LSTM
* Transformer
* BERT
* GPT

all inherit from `nn.Module`.

### Basic Structure
Every PyTorch model typically has:
```python
class MyModel(nn.Module):

    def __init__(self):
        pass

    def forward(self, x):
        pass
```

#### Components
##### `__init__()`
Defines layers.

* Linear Layers
* Activations
* Dropout
* BatchNorm

##### `forward()`
Defines how data flows through the network.
```
Input
 ↓
Layer 1
 ↓
Layer 2
 ↓
Output
```

### Our First Neural Network
**Architecture**
```
Input (2)
     ↓
Hidden (4)
     ↓
Output (1)
```

Diagram:
```
    x1   x2
    \   /
     \ /
 [4 Neurons]
      |
      |
 [1 Output]
```

### Step 1: Import Libraries


````python
import torch
import torch.nn as nn
````

### Step 2: Create Model

````python
class SimpleNN(nn.Module):

    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(
            2,
            4
        )
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(
            4,
            1
        )

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)

        return x
````

#### Understanding Each Layer
##### First Layer
```python
self.fc1 = nn.Linear(2,4)
```

Meaning:
```
2 Inputs
   ↓
4 Outputs
```

Produces 4 neurons.



##### ReLU
```python
self.relu = nn.ReLU()
```

Applies:
```
max(0,x)
```

Diagram:
```
Negative → 0

Positive → Keep
```

Adds non-linearity.

##### Output Layer
```python
self.fc2 = nn.Linear(4,1)
```

Meaning:
```
4 Inputs
   ↓
1 Output
```

### Instantiate Model

````python
model = SimpleNN()

print(model)
````

Output:
```python
SimpleNN(
  (fc1): Linear(2,4)
  (relu): ReLU()
  (fc2): Linear(4,1)
)
```

### Trainable Parameters
Every layer contains weights and biases.

Example:
```python
print(model.fc1.weight)
```

Output:
```
Random Weights
```

These are automatically initialized.

### Count Parameters
```python
total_params = sum(
    p.numel()
    for p in model.parameters()
)

print(total_params)
```


#### Manual Calculation

```

```

Second layer:
```

```


### Why Use `nn.Module`?
Without it:
```
Manually Create Weights
Manually Track Parameters
Manually Handle Layers
```
Very difficult.

With it:
```
Automatic Parameter Tracking
Easy Layer Composition
Built-in Utilities
```


### Sequential API
For simple models, PyTorch provides:

```python
nn.Sequential()
```

Example:
```python
model = nn.Sequential(

    nn.Linear(2,4),

    nn.ReLU(),

    nn.Linear(4,1)

)
```

Equivalent to the previous network.

### When To Use Sequential?
Good for:
```
Simple Feedforward Networks
```
Not ideal for:
```
Transformers
Custom Architectures
Complex Logic
```
For those:

```python
class MyModel(nn.Module)
```
is preferred.


````python
import torch
import torch.nn as nn

class SimpleNN(nn.Module):

    def __init__(self):

        super().__init__()

        self.fc1 = nn.Linear(2,4)

        self.relu = nn.ReLU()

        self.fc2 = nn.Linear(4,1)

    def forward(self, x):

        x = self.fc1(x)

        x = self.relu(x)

        x = self.fc2(x)

        return x

model = SimpleNN()

x = torch.tensor(
    [[1.0, 2.0]]
)

output = model(x)

print(output)
````
