### Big Picture
A training loop is simply:
```
Get Batch
    ↓
Forward Pass
    ↓
Compute Loss
    ↓
Backward Pass
    ↓
Update Weights
    ↓
Repeat
```

This is the core of almost every PyTorch training pipeline.

### The Core Four Lines
Every training loop revolves around:

```python
optimizer.zero_grad()

loss.backward()

optimizer.step()
```

plus:
```python
loss = criterion(...)
```

These are the most important lines in PyTorch training.

# Complete Training Loop
```python
epochs = 10

for epoch in range(epochs):

    for X_batch, y_batch in loader:

        # Clear gradients
        optimizer.zero_grad()

        # Forward pass
        predictions = model(
            X_batch
        )

        # Compute loss
        loss = criterion(
            predictions,
            y_batch
        )

        # Backpropagation
        loss.backward()

        # Update weights
        optimizer.step()

    print(
        f"Epoch {epoch+1}, "
        f"Loss: {loss.item():.4f}"
    )
```

### Understanding Each Step
#### Step 1
```python
optimizer.zero_grad()
```

**Purpose**

Clear old gradients.

**Why?**

PyTorch accumulates gradients.

Without clearing:
```
Old Gradients
      +
New Gradients
```

Training becomes incorrect.

#### Step 2
```python
predictions = model(X_batch)
```

This is: **Forward Pass**

Diagram:
```
Input
 ↓
Network
 ↓
Prediction
```

#### Step 3
```python
loss = criterion(
    predictions,
    y_batch
)
```

This measures: **Prediction Error**

Diagram:
```
Prediction
      ↓
Loss Function
      ↓
Loss Value
```

#### Step 4
```python
loss.backward()
```

Triggers: **Backpropagation**

Diagram:
```
Loss
 ↑
Gradients
 ↑
Weights
```

Autograd computes gradients automatically.

#### Step 5
```python
optimizer.step()
```

Uses gradients to update weights.

Diagram:
```
Gradients
      ↓
Optimizer
      ↓
Updated Weights
```

### Average Epoch Loss
In practice average loss is tracked instead of only the last batch loss.
```python
for epoch in range(epochs):

    total_loss = 0

    for X_batch, y_batch in loader:

        optimizer.zero_grad()

        predictions = model(
            X_batch
        )

        loss = criterion(
            predictions,
            y_batch
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    avg_loss = (
        total_loss /
        len(loader)
    )

    print(avg_loss)
```

This gives a more reliable metric.


### Production Training Flow
```
Load Dataset
      ↓
Create DataLoader
      ↓
Create Model
      ↓
Create Loss Function
      ↓
Create Optimizer
      ↓

For Each Epoch

    For Each Batch

        Forward Pass

        Compute Loss

        Backward Pass

        Update Weights
```

This is the standard Deep Learning workflow.

````python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

# Dataset
X = torch.randn(1000, 10)
y = torch.randn(1000, 1)

dataset = TensorDataset(X, y)

loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True
)

# Model
model = nn.Sequential(
    nn.Linear(10, 32),
    nn.ReLU(),
    nn.Linear(32, 1)
)

# Loss Function
criterion = nn.MSELoss()

# Optimizer
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)

# Training Loop
epochs = 10

for epoch in range(epochs):

    total_loss = 0

    for X_batch, y_batch in loader:

        optimizer.zero_grad()

        predictions = model(X_batch)

        loss = criterion(
            predictions,
            y_batch
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    avg_loss = total_loss / len(loader)

    print(
        f"Epoch [{epoch+1}/{epochs}] "
        f"Loss: {avg_loss:.4f}"
    )
````
