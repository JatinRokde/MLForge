Until now:
```python

```

where:

`X = Entire Dataset`

This works for tiny datasets.

But real-world datasets may contain:
- 10,000 samples
- 100,000 samples
- 1,000,000+ samples

Loading everything at once is inefficient.

PyTorch solves this using:
```
Dataset
   ↓
DataLoader
```

### Why Dataset & DataLoader is Needed?
Without them:
```
Entire Dataset
      ↓
Memory Issues
Slow Training
No Batching
```

With them:
```
Dataset
   ↓
DataLoader
   ↓
Mini-Batches
   ↓
Efficient Training
```

### What is a Dataset?
#### Definition
A Dataset is a PyTorch class that defines:

- How To Access Data
- How To Access Labels

Think of it as:
```
Dataset
     ↓
Data Source
```

### Dataset Structure
Every custom dataset implements:
- `__len__()`
- `__getitem__()`


#### `__len__()`
Returns:
```
Number Of Samples
```

Example:
```python
def __len__(self):
    return len(self.X)
```

#### `__getitem__()`
Returns:
```
One Sample
+
Its Label
```

Example:
```python
def __getitem__(self, idx):
    return self.X[idx], self.y[idx]
```

### Custom Dataset Example

````python
from torch.utils.data import Dataset

class MyDataset(Dataset):

    def __init__(self, X, y):

        self.X = X
        self.y = y

    def __len__(self):

        return len(self.X)

    def __getitem__(self, idx):

        return self.X[idx], self.y[idx]
````

### Create Dataset
```python
dataset = MyDataset(
    X,
    y
)
```

Diagram:
```
Raw Data
    ↓
Dataset Object
```

### What is a DataLoader?
#### Definition
A DataLoader loads data from a Dataset and creates mini-batches automatically.

Diagram:
```
Dataset
    ↓
DataLoader
    ↓
Batch 1

Batch 2

Batch 3
```

### Why Is It Important?
Instead of:
```
100,000 Samples At Once
```

train using:
```
32 Samples
      ↓
Update

32 Samples
      ↓
Update
```

This is called: **Mini-Batch Training**

### Creating a DataLoader
```python
from torch.utils.data import DataLoader

loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True
)
```

### Understanding Parameters
#### batch_size
```python
batch_size=32
```

Means:
```
32 Samples
      ↓
One Training Step
```

#### shuffle
```python
shuffle=True
```

Means:
```
Randomize Data Order
Every Epoch
```

This improves learning.

### Training Loop With DataLoader
Earlier:
```python
prediction = model(X)
```

Now:
```python
for X_batch, y_batch in loader:

    prediction = model(
        X_batch
    )
```


### Complete Training Pattern
```python
for epoch in range(epochs):

    for X_batch, y_batch in loader:

        optimizer.zero_grad()

        prediction = model(
            X_batch
        )

        loss = criterion(
            prediction,
            y_batch
        )

        loss.backward()

        optimizer.step()
```

This is the standard PyTorch training loop.

### Epoch vs Batch
A very common interview question.

#### Batch
One mini-group of samples.

Example:
```
32 Samples
```

#### Epoch
One complete pass through the dataset.

Example:
```
Dataset = 1000 Samples

Batch Size = 100
```

Then:
```
1000 / 100

= 10 Batches
```

One epoch:
```
Batch 1
Batch 2
Batch 3
...
Batch 10
```

### Diagram
```
Epoch

 ├── Batch 1
 ├── Batch 2
 ├── Batch 3
 ├── Batch 4
 ├── Batch 5
```
After all batches:
```
1 Epoch Complete
```

### PyTorch Example

````python
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

X = torch.randn(100, 10)

y = torch.randn(100, 1)

class MyDataset(Dataset):

    def __init__(self, X, y):

        self.X = X
        self.y = y

    def __len__(self):

        return len(self.X)

    def __getitem__(self, idx):

        return self.X[idx], self.y[idx]

dataset = MyDataset(X, y)

loader = DataLoader(
    dataset,
    batch_size=16,
    shuffle=True
)

for X_batch, y_batch in loader:

    print(
        X_batch.shape,
        y_batch.shape
    )

    break
````
