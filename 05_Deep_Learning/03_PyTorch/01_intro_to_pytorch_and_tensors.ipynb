### What is PyTorch?
#### Definition
PyTorch is an open-source Deep Learning framework developed by Meta.

It provides:
* Tensors
* Automatic Differentiation
* Neural Network APIs
* GPU Support
* Model Deployment


### NumPy vs PyTorch
Think of PyTorch as:
```
NumPy
  +
Automatic Gradients
  +
GPU Support
  +
Deep Learning Utilities
```

#### NumPy

````python
import numpy as np

arr = np.array([1, 2, 3])

print(arr)
````

#### PyTorch

````python
import torch

tensor = torch.tensor([1, 2, 3])

print(tensor)
````

Very similar.

### What is a Tensor?
#### Definition
A Tensor is the fundamental data structure in PyTorch.

Think:
```
Scalar → Vector → Matrix → Tensor
```

### Tensor Hierarchy
#### 0D Tensor (Scalar)



````python
x = torch.tensor(5)
````

Shape:

````python
torch.Size([])
````

#### 1D Tensor (Vector)

````python
x = torch.tensor([1, 2, 3])
print(x)
````

Shape:

````python
torch.Size([3])
````

#### 2D Tensor (Matrix)

````python
x = torch.tensor([
    [1, 2],
    [3, 4]
])

print(x)
````

Shape:

````python
torch.Size([2,2])
````

#### 3D Tensor

````python
x = torch.rand(3, 4, 5)
print(x)
````

Shape:
```
3 × 4 × 5
```
Often used for batches of data.

### Creating Tensors
#### From Python List


````python
import torch

x = torch.tensor([1, 2, 3])

print(x)
````

#### Zeros

````python
x = torch.zeros(3, 4)

print(x)
````

#### Ones

````python
x = torch.ones(2, 3)

print(x)
````

#### Random Values

````python
x = torch.rand(2, 3)

print(x)
````

### Tensor Shapes

````python
x = torch.rand(4, 5)

print(x.shape)
````

Meaning:
```
4 Rows
5 Columns
```

### Why Shapes Matter
Neural Networks perform:
```
Matrix Multiplication
```

which requires compatible shapes.

Example:
```
(32,100)
      ×
(100,50)
      ↓
(32,50)
```

Shape debugging is a daily Deep Learning task.

### Basic Tensor Operations
#### Addition

````python
a = torch.tensor([1,2,3])

b = torch.tensor([4,5,6])

print(a + b)
````

#### Multiplication

````python
print(a * b)
````

#### Matrix Multiplication

````python
A = torch.rand(2,3)

B = torch.rand(3,4)

C = torch.matmul(A, B)

print(C.shape)
````

### Tensor Reshaping
Often needed during model building.

#### Reshape

````python
x = torch.arange(12)

x = x.reshape(3,4)

print(x)
````

#### Broadcasting
One of PyTorch's most useful features.

````python
A = torch.tensor([
    [1,2],
    [3,4]
])

B = torch.tensor([10,20])

print(A + B)
````

PyTorch automatically expands:
```
[10,20]

to

[10,20]
[10,20]
```

This is called: **Broadcasting**
