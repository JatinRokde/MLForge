### What is a Loss Function?
#### Definition
A Loss Function measures:
```
Prediction Error
```

between:
```
Model Output
```
and
```
Actual Target
```

#### Example
Prediction:
```
8
```

Actual:
```
10
```

Loss:
```
Error = 2
```

Higher loss:
```
Bad Model
```

Lower loss:
```
Better Model
```

### Common Loss Functions
#### 1. MSE Loss
##### Used For
Regression Problems

Examples:
* House Price Prediction
* Sales Forecasting
* Temperature Prediction

##### Formula
```
(prediction - target)²
```

##### PyTorch
```python
import torch.nn as nn

criterion = nn.MSELoss()
```

##### Example




````python
import torch
import torch.nn as nn

prediction = torch.tensor([8.0])
target = torch.tensor([10.0])

criterion = nn.MSELoss()

loss = criterion(prediction, target)

print(loss)
````

Because:
```
(8 - 10)² = 4
```

#### 2. Binary Cross Entropy (BCE)
##### Used For
Binary Classification

Examples:
* Spam vs Not Spam
* Fraud vs Not Fraud
* Disease vs No Disease

Output:
```
0 or 1
```

##### PyTorch
```python
criterion = nn.BCELoss()
```

More commonly:
```python
nn.BCEWithLogitsLoss()
```

which is numerically more stable.

#### 3. Cross Entropy Loss
##### Used For
Multi-Class Classification

Examples:
* Cat
* Dog
* Horse
* Bird

Only one correct class.


##### PyTorch
```python
criterion = nn.CrossEntropyLoss()
```

This is one of the most frequently used loss functions.


### Quick Summary
| Problem Type               | Loss Function     |
| -------------------------- | ----------------- |
| Regression                 | MSELoss           |
| Binary Classification      | BCEWithLogitsLoss |
| Multi-Class Classification | CrossEntropyLoss  |


### What is an Optimizer?
#### Definition
An Optimizer updates model parameters using gradients.

Recall:
```
Backpropagation
       ↓
Gradients
       ↓
Optimizer
       ↓
New Weights
```


### Why Needed?
Suppose:
```
Prediction = Wrong
```

Loss tells us:
```
How Wrong
```

Backpropagation tells us:
```
Which Direction To Move
```

Optimizer performs:
```
Actual Weight Update
```

### Clear Old Gradients
```python
optimizer.zero_grad()
```

Important because PyTorch accumulates gradients.

### The Four Most Important Lines
Almost every training loop contains:
```python
optimizer.zero_grad()

loss = criterion(
    prediction,
    target
)

loss.backward()

optimizer.step()
```

Memorize this sequence.

### Full Example

````python
import torch
import torch.nn as nn

model = nn.Linear(2,1)

criterion = nn.MSELoss()

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.01
)

X = torch.tensor([
    [1.0,2.0]
])

y = torch.tensor([
    [5.0]
])

prediction = model(X)

loss = criterion(
    prediction,
    y
)

optimizer.zero_grad()

loss.backward()

optimizer.step()

print(loss.item())
````

### Production Training Loop
In practice:
```python
for epoch in range(epochs):

    optimizer.zero_grad()

    prediction = model(X)

    loss = criterion(
        prediction,
        y
    )

    loss.backward()

    optimizer.step()
```

This is the heart of Deep Learning training.
