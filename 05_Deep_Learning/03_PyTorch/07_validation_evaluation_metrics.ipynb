### Why Evaluation Matters
Suppose:
```
Training Accuracy = 99%
```

Looks amazing.

But:
```
Test Accuracy = 70%
```

The model has learned the training data but fails on new data.

This is: **Overfitting**

### Dataset Split
A dataset is usually divided into:
```
Training Set
Validation Set
Test Set
```

Diagram:
```
Entire Dataset
       ↓

 ├── Train (70-80%)
 ├── Validation (10-15%)
 └── Test (10-15%)
```

#### Training Set
Used to:
```
Learn Weights
```

#### Validation Set
Used to:
```
Tune Hyperparameters
Monitor Training
Select Best Model
```

### Test Set
Used once at the end.

Purpose:
```
Estimate Real Performance
```

### Typical Workflow
```
Train
 ↓
Validate
 ↓
Tune Model
 ↓
Train Again
 ↓
Final Test
```

### Training Mode vs Evaluation Mode
#### Training Mode
```python
model.train()
```

Used during training.

Enables:
- Dropout
- BatchNorm Updates

#### Evaluation Mode
```python
model.eval()
```

Used during validation and testing.

Disables:
- Dropout
- BatchNorm Updates

### Why Is This Important?
Recall Dropout:

Training:
```
● X ● ●
● ● X ●
```
Neurons randomly removed.

Evaluation:
```
● ● ● ●
● ● ● ●
```
All neurons active.

Without:
```python
model.eval()
```

predictions become inconsistent.

### Disable Gradients During Evaluation
```python
with torch.no_grad():

    predictions = model(X)
```

Benefits:
- Faster Execution
- Less Memory Usage

because gradients are unnecessary during evaluation.

### Evaluation Flow
```
model.eval()
      ↓
torch.no_grad()
      ↓
Predictions
      ↓
Metrics
```

### Common Metrics
#### 1. Accuracy
**Definition**

Percentage of correct predictions.

Formula:
$$
\frac{Correct \ Predictions}{Total \ Predictions}
$$

**Example**
```
100 Predictions

90 Correct
```

Accuracy:
```
90%
```

**PyTorch / Scikit-Learn**
```python
from sklearn.metrics import accuracy_score

accuracy_score(
    y_true,
    y_pred
)
```

**Problem With Accuracy**
Suppose:
```
1000 Transactions

990 Legit
10 Fraud
```

Model predicts:
```
Everything Legit
```

Accuracy:
```
99%
```

Looks good.

But:
```
Fraud Detection = Useless
```

Hence:
- Precision
- Recall

are needed.

##### Confusion Matrix
Foundation of many evaluation metrics.

Diagram:
$$
\begin{array}{c|cc}
 & \text{Predicted Positive} & \text{Predicted Negative} \\
\hline
\text{Actual Positive} & TP & FN \\
\text{Actual Negative} & FP & TN \\
\end{array}
$$

**Meaning**

**TP — True Positive**

Correct positive prediction.

Example:
```
Fraud predicted
Fraud actual
```

**TN — True Negative**

Correct negative prediction.

Example:
```
Legitimate predicted
Legitimate actual
```

**FP — False Positive**

Predicted positive but actually negative.

Example:
```
Legitimate transaction
Flagged as fraud
```

**FN — False Negative**

Predicted negative but actually positive.

Example:
```
Fraud transaction
Missed by model
```

#### 2. Precision
**Definition**

Of all predicted positives:
```
How Many Were Correct?
```

Formula:
$$
\frac{TP}{TP + FP}
$$

**Example**
```
Predicted Fraud = 100

Actually Fraud = 90
```

Precision:
```
90%
```

**Use When**

False positives are expensive.

Examples:
- Spam Detection
- Fraud Alerts
- Recommendation Systems

#### 3. Recall
**Definition**

Of all actual positives:
```
How Many Did We Find?
```

Formula:
$$
\frac{TP}{TP + FN}
$$

**Example**
```
100 Actual Fraud Cases

Detected 80
```

Recall:
```
80%
```

**Use When**

Missing positives is costly.

Examples:
- Disease Detection
- Fraud Detection
- Intrusion Detection
- Security Systems

##### Precision vs Recall
Think:
```
Precision
     ↓
How Correct?

Recall
     ↓
How Complete?
```

**Example**

Suppose:
```
Actual Fraud Cases = 100
```

**Model A**
```
Detected 20
All 20 Correct
```

Precision:
```
100%
```

Recall:
```
20%
```

**Model B**
```
Detected 90
80 Correct
```

Precision:
```
89%
```

Recall:
```
80%
```

Model B is often preferred because it catches more fraud.


#### 4. F1 Score
**Definition**

Combines:
- Precision
- Recall

into one metric.

**Why?**

Suppose:
```
Precision = 95%
Recall = 20%
```

Not ideal.

High precision but terrible recall.

F1 balances both.

**Intuition**
```
High Precision
+
High Recall
=
High F1
```

### Quick Comparison
| Metric | Question Answered |
|----------|----------|
| Accuracy | Overall correctness? |
| Precision | How many predicted positives were correct? |
| Recall | How many actual positives were found? |
| F1 Score | Balance between Precision and Recall? |
