### Why Save Models?
Training may take:
- Minutes
- Hours
- Days
- Weeks

Examples:

| Model | Training Time |
|---------|---------|
| Small MLP | Minutes |
| CNN | Hours |
| BERT | Days |
| GPT | Weeks / Months |

Re-training every time is impractical.

### Basic Workflow
```
Train Model
      ↓
Save Model
      ↓
Close Program
      ↓
Load Model Later
      ↓
Inference
```

### What Gets Saved?
The most important thing:
```
Learned Weights
```

Remember:
```
Weights
Biases
```

contain all the learned knowledge.


#### state_dict()
**Definition**

A `state_dict` is a Python dictionary containing:

- Weights
- Biases
- Trainable Parameters

of a model.

**Example**

Suppose:
```python
import torch.nn as nn

model = nn.Linear(2, 1)
```

View parameters:
```python
print(model.state_dict())
```

Output (simplified):

```python
{
    'weight': tensor(...),
    'bias': tensor(...)
}
```

### Why state_dict()?
Instead of saving:
```
Entire Python Object
```

PyTorch usually saves:
```
Only Parameters
```

This is the recommended approach.

Benefits:
- Smaller files
- More portable
- Safer across PyTorch versions
- Production standard

### Saving a Model
#### Standard Method
```python
import torch

torch.save(
    model.state_dict(),
    "model.pth"
)
```

Diagram:
```
Model
  ↓
state_dict()
  ↓
model.pth
```

### File Extensions
Common extensions:
```
.pth
.pt
```

Both are widely used.

Examples:
```
best_model.pth
classifier.pt
gpt_checkpoint.pth
```

### Loading a Model
Suppose the model exits:
```
model.pth
```

First create the architecture:
```python
model = SimpleNN()
```

Then load the weights:
```python
model.load_state_dict(
    torch.load("model.pth")
)
```

**Important**
```text
Create Architecture
      ↓
Load Weights
```

because weights alone do not define:
- Layer structure
- Number of neurons
- Connections

The architecture must already exist.


### Complete Example
#### Save
```python
torch.save(
    model.state_dict(),
    "model.pth"
)
```

#### Load
```python
model = SimpleNN()

model.load_state_dict(
    torch.load("model.pth")
)
```

### Switch To Evaluation Mode
After loading:
```python
model.eval()
```

Diagram:
```
Load Model
      ↓
model.eval()
      ↓
Inference
```

Very important when using:
- Dropout
- BatchNorm

### Typical Inference Flow
```python
model.eval()

with torch.no_grad():

    prediction = model(x)
```

Diagram:
```
Input
 ↓
Loaded Model
 ↓
Prediction
```

### Why torch.no_grad()?
During inference:
```
No Training
No Backpropagation
```

Benefits:
- Faster execution
- Lower memory usage
- Better deployment efficiency

### Saving Checkpoints

In production, often save more than weights.

Instead of:
```
Model Only
```

Save:
- Model State
- Optimizer State
- Epoch Number
- Loss Value

### Why Save Checkpoints?
Suppose:
```
Training Duration = 3 Days
```

Training crashes after:
```
Epoch 95
```

Without checkpoint:
```
Restart From Epoch 0
```

With checkpoint:
```
Resume From Epoch 95
```

Huge time savings.
