### What Makes Language Difficult?
Consider:
```
Cat
Dog
Tiger
Lion
```

Humans know:
- Cat and Dog are related.
- Cat and Lion are related.
- Cat and Car are unrelated.

Computers initially do not know this.

For a machine:
```
Cat
Dog
Car
```

are simply strings.

#### Goal of NLP
Convert text into numerical representations that preserve meaning.

Diagram
```
Cat
Dog
Lion

↓

Numbers

↓

Semantic Relationship Preserved
```

### Example of Meaning Preservation
Suppose if represented:
```
Cat → 1
Dog → 2
Car → 3
```

Problem:
```
Dog is close to Cat

Dog is also close to Car
```

Difference = 1

Meaning is lost.


Better representation:
```
Cat → [0.9, 0.8]

Dog → [0.8, 0.9]

Car → [0.1, 0.2]
```

Now:
```
Cat ≈ Dog

Cat ≠ Car
```

Meaning preserved.

This is the idea behind **embeddings**.

### Evolution of Text Representation
NLP evolved through multiple stages.


#### Stage 1 — Integer Encoding
Vocabulary:
```
I
Love
NLP
```

Mapping:

```python
{
 "I": 1,
 "Love": 2,
 "NLP": 3
}
```

Sentence:
```
I Love NLP
```

↓
```
[1, 2, 3]
```

**Problem**
```
3 > 2 > 1
```

But:
```
NLP
```

is not "greater" than:
```
Love
```

No semantic meaning.

#### Stage 2 — One-Hot Encoding
```
I      → [1,0,0]

Love   → [0,1,0]

NLP    → [0,0,1]
```

##### Advantages
- Unique representation

##### Problems
- No similarity information
- Sparse vectors
- Huge memory usage

#### Stage 3 — Bag of Words (BoW)
Count occurrences of words.

#### Stage 4 — TF-IDF
Importance-based text representation.


#### Stage 5 — Word Embeddings
Examples:
- Word2Vec
- GloVe
- FastText

Meaning captured.

#### Stage 6 — Contextual Embeddings
Examples:
- BERT
- GPT
- Llama

Meaning depends on context.


### Evolution Diagram
```
Integer Encoding
       ↓
One-Hot Encoding
       ↓
Bag of Words
       ↓
TF-IDF
       ↓
Word2Vec
       ↓
GloVe
       ↓
FastText
       ↓
BERT
       ↓
GPT
```
