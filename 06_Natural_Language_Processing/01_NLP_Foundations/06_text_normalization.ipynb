### What is Text Normalization?
#### Definition
Text Normalization is the process of converting text into a standard, consistent format before processing.

##### Example
```
HELLO
Hello
hello
```

Humans know these mean the same thing.

A computer may treat them as:
```
HELLO ≠ Hello ≠ hello
```

Normalization converts them into:
```
hello
```

#### Diagram
```
HELLO
Hello
hello

   ↓

hello
```

### Why Is Normalization Needed?
Imagine a sentiment dataset:
```
Good
GOOD
good
GoOd
```

Without normalization:
```
4 different tokens
```

With normalization:
```
good
```

Only one token.

#### Benefits
- Smaller Vocabulary
- Reduced Noise
- Better Consistency
- Better Learning

### Classical NLP vs Modern LLMs
#### Classical NLP

Heavy normalization was common.

Examples:
- Lowercasing
- Punctuation Removal
- Stopword Removal
- Stemming
- Lemmatization

Because:
```
BoW
TF-IDF
Word2Vec
```

needed cleaner inputs.

#### Modern LLMs
Much lighter normalization.

Because Transformers can learn many patterns automatically.

For example:
```
Hello!
```

and
```
Hello.
```

may carry different meanings.

Removing punctuation can actually lose information.

```
Classical NLP:
Heavy normalization

Modern LLMs:
Minimal normalization
```

### Lowercasing
Most common normalization technique.

#### Example
```
APPLE
Apple
apple
```
↓
```
apple
```

#### Python
```python
text = "APPLE"

print(text.lower())
```

Output:
```
apple
```

#### Benefits
- Smaller Vocabulary
- Reduced Sparsity

##### Example
Without Lowercasing:
```
Apple
APPLE
apple
```

Vocabulary Size:
```
3
```

With Lowercasing:
```
apple
```

Vocabulary Size:
```
1
```

#### Potential Problem
Sometimes capitalization matters.

Example:
```
US
```

vs
```
us
```

Different meanings.

Therefore:
```
Classical NLP → Often lowercase

Modern LLMs → Usually preserve case
```

### Whitespace Normalization
Text often contains extra spaces.

##### Example
```
I     love      NLP
```
↓
```
I love NLP
```

##### Python

````python
text = "I     love      NLP"

normalized = " ".join(text.split())

print(normalized)
````

#### Why?
Extra spaces usually carry no meaning.

### Punctuation Handling
##### Example
```
Hello!!!
```
↓
```
Hello
```

#### Python

````python
import string

text = "Hello!!!"

cleaned = text.translate(
    str.maketrans(
        '',
        '',
        string.punctuation
    )
)

print(cleaned)
````

#### Classical NLP
Very common.

#### Modern LLMs
Usually preserve punctuation.

Because punctuation carries information.

Example:
```
Let's eat, Grandma.
```

vs

```
Let's eat Grandma.
```

Completely different meaning.


### Special Character Removal
##### Example
```
Hello@#$%^
```
↓
```
Hello
```

##### Python

````python
import re

text = "Hello@#$%^"

cleaned = re.sub(
    r"[^a-zA-Z0-9\s]",
    "",
    text
)

print(cleaned)
````

#### Useful For
- Spam Detection
- Review Analysis
- Classical NLP Pipelines

### Unicode Normalization
Real-world text contains:
```
café
naïve
résumé
```

Different Unicode representations can exist.

Normalization ensures consistency.

##### Example
```
café
```

may internally have multiple Unicode forms.

````python
import unicodedata

text = "café"

normalized = unicodedata.normalize(
    "NFKD",
    text
)

print(normalized)
````

```
Unicode normalization ensures
consistent representation of text.
```

No need for deeper Unicode theory.

```

Number normalization is task dependent.
```
