### What is a Document?
A document is a single unit of text.

Think of it as one complete piece of content.

Examples:
- One Email
- One Review
- One News Article
- One PDF
- One Chat Message
- One Web Page

All of these can be treated as documents.

##### Example
Document:
```
I love NLP. Transformers are amazing.
```

Entire text = one document.

#### More Examples
##### Review
```
This movie was fantastic.
```

One document.

##### Email
```
Dear Team,
Please find the attached report.
```

One document.

### Why Documents Matter
Many NLP tasks operate at the document level.

##### Example: Sentiment Analysis
Document:
```
This phone is amazing.
```

Prediction:
```
Positive
```

##### Example: Spam Detection
Document:
```
Congratulations!
You won ₹10,00,000.
```

Prediction:
```
Spam
```

### What is a Sentence?
A sentence is a meaningful linguistic unit within a document.

Document:
```
I love NLP.
Transformers are amazing.
GPT changed AI.
```

Contains:
```
Sentence 1
I love NLP.

Sentence 2
Transformers are amazing.

Sentence 3
GPT changed AI.
```

**Diagram**
```
Document
    │
    ├── Sentence 1
    │
    ├── Sentence 2
    │
    └── Sentence 3
```

### Why Sentences Matter
Many NLP tasks operate at the sentence level.

Examples:
- Translate this sentence
- Summarize this sentence
- Classify this sentence


#### Machine Translation
Input:
```
How are you?
```

Sentence is the processing unit.

### What is a Token?
A token is the smallest meaningful unit processed by NLP systems.

For now, think:
```
Token ≈ Word
```

##### Example
Sentence:
```
I love NLP
```

Tokens:

```python
[
 "I",
 "love",
 "NLP"
]
```


##### Diagram
```
Sentence

I love NLP

      ↓

Tokens

[I]
[love]
[NLP]
```

### Hierarchy Example
Document:
```
I love NLP.

Hierarchy:
```
Document
│
├── Sentence 1

│       ├── I

```
Corpus

```

This hierarchy appears throughout NLP.

### Tokens Are the Real Input to Models
Humans think:
```
Sentence
```

LLMs think:
```
Tokens
```

##### Example
Input:
```
I love NLP
```

Model sees:
```
[I]
[love]
[NLP]
```

Then later:
```
Token IDs
```

Example:
```python
[101, 2050, 4789]
```

Then:
```
Embeddings
```

Then:
```
Transformer
```

### Complete Pipeline
```
Text
 ↓

Tokens
 ↓

Token IDs
 ↓

Embeddings
 ↓

Transformer
 ↓

Output
```

This is exactly how GPT works.

### Are Tokens Always Words?
No.

This is extremely important.

Modern LLMs do **not** operate purely on words.

##### Example
Word:
```
unhappiness
```

May become:
```python
["un", "happi", "ness"]
```


##### Example
Word:
```
ChatGPT
```

May become:

```python
["Chat", "G", "PT"]
```

(or something similar depending on the tokenizer)

This is called:
```
Subword Tokenization
```

```
Token ≠ Always Word
```

### Why Not Use Characters?
Sentence:
```
Transformers
```

Character Tokens:
```python
[
'T',
'r',
'a',
'n',
's',
'f',
'o',
'r',
'm',
'e',
'r',
's'
]
```

Problems:

- Too many steps
- Long sequences
- Harder learning

Modern NLP prefers:

```text
Words
or
Subwords
```

### Why Not Use Entire Sentences?
Sentence:
```
I love NLP
```

Another sentence:
```
I love Deep Learning
```

If each sentence became one token:
```
Infinite Vocabulary
```

Impossible.

Therefore:
```
Word/Subword
```

### Document-Level vs Sentence-Level vs Token-Level Tasks
#### Document-Level Tasks
Input:
```
Entire Review
```

Tasks:
- Sentiment Analysis
- Spam Detection
- Topic Classification

Diagram:
```
Document
     ↓
 Prediction
```

#### Sentence-Level Tasks
Input:
```
Single Sentence
```

Tasks:
- Translation
- Summarization
- Classification

Diagram:
```
Sentence
      ↓
 Prediction
```

### Token-Level Tasks
Input:
```
Individual Tokens
```

Tasks:
- Named Entity Recognition (NER)
- POS Tagging

Example:
```
John lives in Mumbai
```

Output:
```
John   → PERSON

Mumbai → LOCATION
```

Diagram:

```text
Token
     ↓
 Label
```
