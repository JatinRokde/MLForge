### Motivation
Consider these words:
```
play
playing
played
plays
```

Humans immediately understand:
```
All refer to the same core concept:
play
```

But a machine may treat them as:
```
play    ≠ playing
play    ≠ played
play    ≠ plays
```

Result:
* Larger Vocabulary
* More Features
* More Sparsity

#### The Problem
Sentence 1:
```
I play cricket
```

Sentence 2:
```
I played cricket
```

Sentence 3:
```
I am playing cricket
```

Without stemming:

Vocabulary:
```
play
played
playing
```

Three different tokens.

Ideally:
```
play
```

should represent all of them.

This is where stemming comes in.

### What is Stemming?
#### Definition
Stemming is the process of reducing words to their stem (base form) by removing prefixes or suffixes.

##### Example
```
playing
```
↓
```
play
```

##### Example
```
played
```
↓
```
play
```

##### Example
```
plays
```
↓
```
play
```

##### Diagram
```
playing
played
plays

    ↓

  play
```

### Why Do We Need Stemming?
Main goals:
* Reduce Vocabulary Size
* Reduce Sparsity
* Improve Matching
* Improve Retrieval

##### Example
Corpus:
```
playing football

played football

plays football
```

Without stemming:

Vocabulary:
```
playing
played
plays
football
```

Size:
```
4
```

With stemming:
```
play
football
```

Size:
```
2
```

Benefit:
```
Smaller Feature Space
```

especially useful for:
* BoW
* TF-IDF
* Information Retrieval

### How Stemming Works
Most stemmers use simple rules.

##### Example
```
playing
```

Ends with:
```
ing
```

Remove it:
```
play
```

##### Example
```
played
```

Ends with:
```
ed
```

Remove it:
```
play
```

##### Example
```
cars
```

Ends with:
```
s
```

Remove it:
```
car
```

##### Diagram
```
playing

    ↓

Remove "ing"

    ↓

play
```

##### Example

### Overstemming Problem
Stemming can become too aggressive.

##### Example
```
universe
```
↓
```
univers
```

##### Example
```
university
```
↓
```
univers
```

Now:
```
universe
university
```
become identical.

Meaning lost.

##### Diagram
```
universe
university

     ↓

   univers

Meaning Collapsed
```

This is called: **Overstemming**

### Understemming Problem
Opposite issue.

Related words fail to merge.

##### Example
```
analysis
analyze
```

may remain different.

Result:
```
Related concepts
not grouped together
```

This is called: **Understemming**

### Popular Stemming Algorithms
#### Porter Stemmer
Most famous.

Introduced by:
```
Martin Porter
```

Characteristics:
* Rule-Based
* Fast
* Widely Used


Example:
```
playing
```
↓
```
play
```

#### Snowball Stemmer
Improved version of Porter.

Advantages:
* More Languages
* Slightly Better Quality

Interview Answer:
```
Snowball = Improved Porter Stemmer
```

### Python Example (Porter Stemmer)




````python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = [
    "playing",
    "played",
    "plays"
]

for word in words:
    print(
        word,
        "->",
        stemmer.stem(word)
    )
````

````python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

print(stemmer.stem("studies"))
````

### Why Stemming Is Less Common in Modern LLMs
#### Old NLP Pipeline
```
Text
 ↓
Stemming
 ↓
BoW
 ↓
Model
```

#### Modern Pipeline
```
Text
 ↓
Tokenizer
 ↓
Transformer
```

Reason:

Transformers learn context directly.

Example:
```
play
played
playing
```

Embeddings can learn:
```
These words are related
```

without stemming.

Therefore:
```
BERT
GPT
Llama
Gemma
```

usually do NOT apply stemming.

### When Is Stemming Still Useful?
#### Useful
* Classical NLP
* BoW
* TF-IDF
* Search Engines
* Information Retrieval
* Keyword Matching

#### Rarely Used
* BERT
* GPT
* Llama
* Modern Embedding Models
* RAG Pipelines
