### What is an NLP Dataset?
#### Definition
An NLP dataset is a structured collection of text data used to train, validate, or evaluate NLP models.

##### Example
Sentiment Analysis Dataset:
| Text                   | Sentiment |
| ---------------------- | --------- |
| This movie is great    | Positive  |
| This movie is terrible | Negative  |
| Amazing acting         | Positive  |

Here:
```
Text      → Input
Sentiment → Label
```

This is a dataset.

### Corpus vs Dataset
#### Corpus
Raw text collection.

Example:
```
I love NLP.

Transformers are powerful.

GPT changed AI.
```

No labels.

#### Dataset
Text + additional information.

Example:
| Text       | Label    |
| ---------- | -------- |
| I love NLP | Positive |
| Bad movie  | Negative |


### Why Is Datasets Needed?
Imagine building a spam detector.

Without data:
```
Model cannot learn
```

Need examples.
```
"Win money now"      → Spam

"Meeting tomorrow"   → Not Spam

"Claim your prize"   → Spam
```

Model learns patterns from these examples.

#### Diagram
```
Dataset
    ↓
Training
    ↓
Model
    ↓
Predictions
```

### Types of NLP Datasets
Broadly:
* Labeled Datasets
* Unlabeled Datasets

### Labeled Dataset
Contains:
* Input Text
* Target Label

#### Example
Sentiment Analysis
| Text           | Sentiment |
| -------------- | --------- |
| Great movie    | Positive  |
| Terrible movie | Negative  |

Used for:

* Classification
* NER
* Question Answering
* Translation
* Supervised Learning

### Unlabeled Dataset
Contains only text.

Examples:
* Wikipedia Articles
* Books
* Web Pages
* Research Papers

No labels.

Used for:
* Word2Vec
* BERT Pretraining
* GPT Pretraining
* Llama Pretraining

### Supervised NLP
Dataset:
| Email        | Spam? |
| ------------ | ----- |
| Win money    | Yes   |
| Team meeting | No    |

Model learns:
```
Input → Output
```

Applications:
* Spam Detection
* Sentiment Analysis
* NER
* Text Classification

### Unsupervised NLP
No labels.

Dataset:
```
Millions of articles
```

Goal:
```
Learn language patterns
```

Examples:
* Word2Vec
* GloVe
* BERT Pretraining
* GPT Pretraining



**How is GPT trained without labels?**

**Answer**

GPT creates its own labels.

Example:
```
The sky is ___
```

Target:
```
blue
```
Generated automatically from text.

### Train, Validation, Test Split

Exactly like Machine Learning.

Dataset:
```
10000 Documents
```

Split:
```
Train       70–80%

Validation  10–15%

Test        10–15%
```

#### Training Set
Used to learn.

Example:
```
8000 samples
```
Model updates weights.

#### Validation Set
Used during training.

Purpose:
* Hyperparameter Tuning
* Model Selection
* Early Stopping

Example:
```
1000 samples
```

#### Test Set
Used only at the end.

Purpose:
```
Final Evaluation
```

Example:
```
1000 samples
```

### NLP Dataset Formats
#### CSV
```csv
Text,Label
Great movie,Positive
Bad movie,Negative
```

#### JSON
```json
{
  "text": "Great movie",
  "label": "Positive"
}
```

#### JSONL
```json
{"text":"Great movie","label":"Positive"}
{"text":"Bad movie","label":"Negative"}
```

Very common in:
* LLM Fine-Tuning
* Instruction Tuning
* Chat Datasets
