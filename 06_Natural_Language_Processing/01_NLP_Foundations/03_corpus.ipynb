### Intuition
Imagine teaching a child English.

You give the child:
- Books
- Newspapers
- Stories
- Articles
- Conversations

After reading enough text, the child learns:
- Words
- Grammar
- Meaning
- Context
- Language Patterns

LLMs learn similarly.

Instead of books, they are trained on:
- Web Pages
- Wikipedia
- Books
- Research Papers
- Forums
- Documentation
- Code

This huge collection of text is called a:
```
Corpus
```

#### Formal Definition
A corpus is:

> A structured collection of text documents used for NLP tasks, training, evaluation, or analysis.


### Why Does a Corpus Needed?
Suppose, user want to train a model to understand English.

It can be trained on:
```
I love NLP
```

alone?

**No.**

The model needs thousands, millions, or billions of examples.

Example:
```
I love NLP

Machine learning is interesting

Transformers changed NLP

Deep learning is powerful

RAG improves accuracy
```

This collection becomes a corpus.

**Diagram**
```
Document 1
Document 2
Document 3
Document 4
Document 5
      │
      ▼
    Corpus
```

### Corpus vs Document
#### Document
A single piece of text.

Example:
```
I love NLP.
```

This is one document.

Another example:
```
Transformers are powerful.
```

This is another document.

#### Corpus

Collection of documents.
```
Doc 1 → I love NLP.

Doc 2 → Transformers are powerful.

Doc 3 → RAG improves retrieval.

Doc 4 → LLMs generate text.
```

Together:
```
Corpus
```

**Diagram**
```
Document
      │
      ▼
Document
      │
      ▼
Document
      │
      ▼
Corpus
```

### Corpus → Document → Sentence → Token
This hierarchy appears everywhere in NLP.

##### Example Corpus
```text
Document 1:
I love NLP.

Document 2:
Transformers are amazing.
```

##### Hierarchy
```
Corpus
  │
  ├── Document 1
  │       │
  │       └── Sentence
  │               │
  │               └── Tokens
  │
  └── Document 2
          │
          └── Sentence
                  │
                  └── Tokens
```

##### Real Example
```
Corpus
 │
 ├── Document
 │     │
 │     └── I love NLP.
 │
 └── Document
       │
       └── Transformers are amazing.
```

##### Token Level
```
Transformers are amazing.
```

↓

```python
[
 "Transformers",
 "are",
 "amazing"
]
```

### Corpus in Classical NLP

### Corpus in Word2Vec
Word2Vec learns relationships from a corpus.

Corpus:
```
King is a ruler

Queen is a ruler

King and Queen lead kingdoms
```

Model learns:
```
King ≈ Queen
```

because they appear in similar contexts.

Without corpus:
```
No Learning
```


### Corpus Quality Matters
Imagine the corpus contains:
- Incorrect information
- Spam
- Duplicates
- Noise

The model learns bad patterns.


### Principle
```
Garbage In
      ↓
Garbage Out
```

### Good Corpus Characteristics
A high-quality corpus should be:
- Clean
- Diverse
- Accurate
- Representative

### Python Example
```python
corpus = [
    "I love NLP",
    "Transformers are powerful",
    "RAG improves retrieval"
]

print("Documents:", len(corpus))
```

Output:
```
Documents: 3
```

### Access Documents

```python
for doc in corpus:
    print(doc)
```

Output:
```
I love NLP

Transformers are powerful

RAG improves retrieval
```
