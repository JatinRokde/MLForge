### What is Tokenization?
#### Definition
Tokenization is the process of breaking text into smaller units called tokens.

##### Example
Input:
```
I love NLP
```

Output:
```python
[
 "I",
 "love",
 "NLP"
]
```

Each element is a token.

**Diagram**
```
Text

I love NLP

      ↓

Tokens

[I]
[love]
[NLP]
```

### Why Do We Need Tokenization?
Computers cannot process raw text directly.

Consider:
```
Transformers are amazing.
```

For a machine this is:
```
One long string
```

Models need smaller units.

Without Tokenization
```
"Transformers are amazing."
```
Single object.

With Tokenization
```python
[
 "Transformers",
 "are",
 "amazing"
]
```

Now the model can process each unit separately.

Pipeline
```
Raw Text
    ↓
Tokenization
    ↓
Tokens
    ↓
Numerical IDs
    ↓
Embeddings
    ↓
Model
```

### What Exactly Is a Token?
Many thinks:
```
Token = Word
```

This is only partially true.

In Classical NLP
```
Token ≈ Word
```

Example:
```
I love NLP
```

↓

```python
[
 "I",
 "love",
 "NLP"
]
```

In Modern LLMs
```
Token ≠ Word
```

A token can be:
- Word
- Part of a word
- Character sequence
- Punctuation

Example:
```
unhappiness
```

↓

```python
[
 "un",
 "happiness"
]
```

or

```python
[
 "un",
 "happy",
 "ness"
]
```
depending on tokenizer.

### Why Not Use Entire Sentences?
Sentence:
```
I love NLP
```

Another sentence:
```
I love Deep Learning
```

If every sentence became one token:
```
Vocabulary explodes
```

Millions or billions of unique sentences.

Impossible to manage.

### Why Not Use Individual Characters?
Sentence:
```
Transformers
```

Character tokens:

```python
[
 'T',
 'r',
 'a',
 'n',
 's',
 'f',
 'o',
 'r',
 'm',
 'e',
 'r',
 's'
]
```

Problems: **Very Long Sequences**

One word:
```
Transformers
```

becomes:
```
12 tokens
```

One paragraph:
```
Hundreds of tokens
```

Learning becomes difficult.

### Types of Tokenization
There are three major types:
- Character Tokenization
- Word Tokenization
- Subword Tokenization

### Character Tokenization
Break text into characters.

##### Example
```
CAT
```

↓

```python
[
 'C',
 'A',
 'T'
]
```

Diagram
```
CAT

 ↓

[C]
[A]
[T]
```

##### Advantages
- Tiny vocabulary
- No unknown words

##### Disadvantages
- Very long sequences
- Poor efficiency

```
Rarely used in modern LLMs.
```

### Word Tokenization
Break text into words.

##### Example
```
I love NLP
```

↓

```python
[
 "I",
 "love",
 "NLP"
]
```

Diagram
```
I love NLP

 ↓

[I]
[love]
[NLP]
```

##### Advantages
- Easy to understand
- Works for classical NLP

##### Problems
Consider:
```
chatgptization
```

Never seen before.

Word tokenizer says:
```
Unknown Word
```

Vocabulary becomes huge.

##### Problem with Word Tokenization?
Answer:
```
Huge vocabulary

Unknown words

Memory expensive
```

### Subword Tokenization (Most Important)
Modern LLMs use this.

#### Core Idea
Instead of:
```
Character
```
or
```
Entire Word
```

Use:
```
Word Pieces
```

##### Example
```
unhappiness
```

↓

```python
[
 "un",
 "happy",
 "ness"
]
```

##### Example
```
playing
```

↓

```python
[
 "play",
 "ing"
]
```



#### Advantages
- Smaller vocabulary
- Handles unseen words
- Captures meaning
- Efficient

This is why:
- GPT
- Llama
- Gemma
- Qwen
- Claude

all use subword tokenization.

```

```

Example:

### Context Window Connection
When GPT says:
```
128K Context Window
```

it means:
```
128K Tokens
```

NOT
```
128K Words
```

Example:
```
1000 words
```

might become:
```
1200–1500 tokens
```

depending on language.

This becomes important later in:
- RAG
- Prompt Engineering
- LLMs

### Python Example
#### Basic Tokenization

````python
text = "I love NLP"

tokens = text.split()

print(tokens)
````

````python
import nltk

nltk.download('punkt_tab')
````

````python
from nltk.tokenize import word_tokenize

text = "I love NLP."

tokens = word_tokenize(text)

print(tokens)
````

Notice:
```
Punctuation becomes token.
```
