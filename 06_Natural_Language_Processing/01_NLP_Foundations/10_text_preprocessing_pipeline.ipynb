### What is a Text Preprocessing Pipeline?
#### Definition
A text preprocessing pipeline is a sequence of transformations applied to raw text before it is used by an NLP model.

Raw text is usually messy.

Example:
```
HELLO!!!   I am learning NLP 😊
```

Before using it in a model, may need to:
* Clean it
* Normalize it
* Tokenize it
* Transform it into numbers

#### Diagram
```
Raw Text
    ↓
Preprocessing
    ↓
Clean Text
    ↓
Features / Embeddings
    ↓
Model
```

### Why Preprocessing Is Needed?
Imagine a sentiment analysis dataset.
```
Amazing Movie!!!

amazing movie

AMAZING MOVIE
```

Humans see:
```
Same meaning
```

A machine may see:
```
Three different patterns
```

Preprocessing helps:
* Reduce Noise
* Improve Consistency
* Reduce Vocabulary Size
* Improve Learning

### Classical NLP Pipeline
This is the pipeline used with:
* BoW
* TF-IDF
* Naive Bayes
* Logistic Regression
* SVM

#### Diagram
```
Raw Text
     ↓
Normalization
     ↓
Tokenization
     ↓
Stopword Removal
     ↓
Stemming/Lemmatization
     ↓
Vectorization
     ↓
ML Model
```

#### Step 1 — Raw Text
Example:
```
HELLO!!! I am learning NLP 😊
```

#### Step 2 — Normalization
Apply:
* Lowercasing
* Whitespace Cleanup
* Unicode Cleanup
* Special Character Handling

Result:
```
hello!!! i am learning nlp 😊
```

#### Step 3 — Tokenization
Split into tokens.
```python
[
 "hello",
 "i",
 "am",
 "learning",
 "nlp"
]
```

#### Step 4 — Stopword Removal
Remove common words.

Example:
```
i
am
```

Result:
```python
[
 "hello",
 "learning",
 "nlp"
]
```

#### Step 5 — Stemming/Lemmatization
Example:
```
learning
```
↓
Lemma:
```
learn
```

Result:
```python
[
 "hello",
 "learn",
 "nlp"
]
```

#### Step 6 — Vectorization
Convert text into numbers.

Example:
```
hello learn nlp
```
↓
BoW / TF-IDF Vector
```python
[
 1,
 0,
 3,
 1
]
```

#### Step 7 — Model
Input:
```python
[
 1,
 0,
 3,
 1
]
```

↓

Prediction:
```
Positive
```
or
```
Spam
```

### Modern Transformer Pipeline
Modern NLP is different.

#### Pipeline
```
Raw Text
     ↓
Light Cleaning
     ↓
Tokenizer
     ↓
Token IDs
     ↓
Embeddings
     ↓
Transformer
```

Notice:
```
No Stopword Removal

No Stemming

No Lemmatization
```
in most cases.

**Why?**

Transformers learn language directly.

Example:
```
play
playing
played
```

The model learns their relationships through embeddings and context.

No need to manually convert them to:
```
play
```

### Classical NLP vs Transformer NLP
| Step          | Classical NLP | Transformers |
| ------------- | ------------- | ------------ |
| Lowercasing   | Often         | Sometimes    |
| Tokenization  | Yes           | Yes          |
| Stopwords     | Often Remove  | Usually Keep |
| Stemming      | Often         | Rare         |
| Lemmatization | Often         | Rare         |
| Vectorization | BoW / TF-IDF  | Embeddings   |
| Model         | ML            | Transformer  |

### What Should Be Removed?
#### Remove (Often)
* Extra Whitespace
* Broken Unicode
* Noise
* Malformed Characters

#### Depends on Task
* Stopwords
* Numbers
* URLs
* Emails
* Emojis

#### Usually Keep for LLMs
* Grammar
* Punctuation
* Stopwords
* Word Variations
