### Intuition
Humans communicate using language.

Example:
```
Hello
How are you?
I love machine learning.
India won the match.
```

Humans understand these naturally.

Computers do not.

For a computer:
```
"I love NLP"
```

is simply:
```
Sequence of characters

or

Bytes
```

Computers don't inherently understand:
- Meaning
- Emotion
- Intent
- Context
- Sarcasm
- Language

NLP exists to bridge this gap.


#### Simple Definition
Natural Language Processing (NLP) is a branch of Artificial Intelligence that enables computers to:
- Read Human Language
- Understand Language
- Interpret Meaning
- Generate Language

### What Does NLP Actually Do?
NLP attempts to answer questions like:
```
What words are present?
What do they mean?
How are they related?
What is the user's intention?
What should be generated next?
```

**Example**

Sentence:
```
I love machine learning.
```

NLP tries to understand:
```
Subject   → I

Action    → love

Object    → machine learning

Sentiment → Positive
```

### Natural Language vs Programming Language
#### Programming Language
Example:
```python
if age > 18:
    print("Adult")
```

Characteristics:
- Strict
- Unambiguous
- Rule Based

#### Natural Language
Example:
```
It's cold here.
```

Possible meanings:
- Temperature is low.
- Close the window.
- Turn on the heater.

Meaning depends on context.

This makes NLP difficult.

### Why NLP Is Hard
Humans effortlessly understand language.

Machines struggle because language contains:

#### Ambiguity
Sentence:
```
I saw a man with a telescope.
```

Question:
```
Who has the telescope?
```

Possibilities:
```
I have telescope

OR

Man has telescope
```

#### Synonyms
```
Car
Automobile
Vehicle
```

Same meaning.

#### Polysemy
One word → Multiple meanings.

Example:
```
Bank
```

Could mean:
```
River bank
Financial bank
```


#### Context Dependence
Sentence:
```
Apple is amazing.
```

Meaning?
```
Fruit?
Company?
```

Depends on context.

#### Sarcasm
Sentence:
```
Great! My laptop crashed again.
```

Literal meaning:
```
Positive
```

Actual meaning:
```
Negative
```

### Major NLP Tasks
#### Task 1 — Text Classification
Input:
```
This movie is fantastic.
```

Output:
```
Positive
```

Applications:
- Sentiment Analysis
- Spam Detection
- Topic Classification

Diagram
```
Text
 ↓
NLP Model
 ↓
Class Label
```

#### Task 2 — Named Entity Recognition (NER)
Sentence:
```
Elon Musk founded Tesla.
```

Output:
```
Elon Musk → Person

Tesla → Organization
```

Diagram
```
Sentence
     ↓
Entity Detection
     ↓
Person / Location / Organization
```

#### Task 3 — Machine Translation
Input:
```
How are you?
```

Output:
```
¿Cómo estás?
```

Diagram
```
English
    ↓
Translator
    ↓
Spanish
```

#### Task 4 — Question Answering
Input:
```
Who invented Python?
```

Output:
```
Guido van Rossum
```

This later becomes:
```
RAG
LLMs
ChatGPT
```

#### Task 5 — Text Generation
Input:
```
Write an email requesting leave.
```

Output:
```
Dear Manager,
...
```

This is the foundation of:
- GPT
- Claude
- Gemini
- Llama

#### Task 6 — Summarization
Input:
```
10-page document
```

Output:
```
1-page summary
```

#### Task 7 — Information Extraction
Input:
```
John joined TCS in 2025.
```

Output:
```
Name    = John
Company = TCS
Year    = 2025
```

Used heavily in enterprise systems.

```
Tokenization
       ↓
BoW
       ↓
TF-IDF
       ↓
Word2Vec
       ↓
RNN
       ↓
LSTM
       ↓
Attention
       ↓
Transformer
       ↓
GPT
       ↓
ChatGPT
```

Everything in Generative AI is built on NLP foundations.
