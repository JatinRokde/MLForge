### What Are Stopwords?
#### Definition
Stopwords are very common words that occur frequently in a language and often carry little meaning by themselves.

Examples:
```
the
a
an
is
am
are
was
were
of
in
on
at
to
for
and
or
but
```

##### Example Sentence
```
The movie is very good.
```

Stopwords:
```
The
is
```

Content words:
```
movie
very
good
```

### Why Were Stopwords Important?
```
The movie is good

The movie is amazing

The movie is excellent
```

Word Frequencies
```
the       → 3
movie     → 3
is        → 3
good      → 1
amazing   → 1
excellent → 1
```

Observation:
```
the
is
```

appear everywhere.

They don't help much in distinguishing documents.

Therefore early NLP researchers said:
```
Remove them.
```

### Stopwords in TF-IDF

This is where stopwords become especially important.

Remember:
```
TF-IDF
=
Term Frequency
×
Inverse Document Frequency
```

Stopword:
```
the
```

appears in almost every document.

Therefore:
```
IDF ≈ Very Low
```

Result:
```
Very small TF-IDF score
```

Interesting fact:

Even if the user don't explicitly remove stopwords,

TF-IDF naturally downweights them.

Note
```
TF-IDF automatically assigns low importance
to extremely common words.
```

### Stopword Removal Example


### Python Example




````python
import nltk

nltk.download("stopwords")
````

````python
from nltk.corpus import stopwords

stop_words = set(
    stopwords.words("english")
)

words = [
    "the",
    "cat",
    "is",
    "sitting"
]

filtered = [
    w
    for w in words
    if w not in stop_words
]

print(filtered)
````
