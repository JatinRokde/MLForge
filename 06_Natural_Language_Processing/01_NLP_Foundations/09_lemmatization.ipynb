### Motivation
Consider:
```

```

Humans know:


Ideally, NLP systems should also know this.

```

```

Not a valid word.

```
studying → study

```


### What is Lemmatization?
#### Definition
Lemmatization is the process of converting a word to its lemma (dictionary base form) using linguistic knowledge and vocabulary rules.

The resulting word:
```
Must be a valid dictionary word.
```

##### Example
```
running
```
↓
```
run
```

##### Example
```
studies
```
↓
```
study
```

##### Example
```
better
```
↓
```
good
```
(Depending on the lemmatizer and part of speech.)

### What is a Lemma?
A lemma is the canonical dictionary form of a word.

##### Examples
| Word    | Lemma |
| ------- | ----- |
| playing | play  |
| played  | play  |
| plays   | play  |
| running | run   |
| studies | study |
| better  | good  |
| cars    | car   |

Think of a dictionary.

User won't find:
```
playing
played
plays
```

as separate root entries.

Rather, this will be available:
```
play
```

This root dictionary form is the lemma.

### Why Do We Need Lemmatization?
Same motivation as stemming.

#### Without Lemmatization
Vocabulary:
```
play
playing
played
plays
```

Four separate entries.

#### With Lemmatization
Vocabulary:
```
play
```

One entry.

Benefits:
* Reduced Vocabulary
* Reduced Sparsity
* Better Semantic Meaning
* Better Search Quality

* Dictionary
* Grammar Rules

Returns:

#### Comparison
| Feature              | Stemming            | Lemmatization            |
| -------------------- | ------------------- | ------------------------ |
| Method               | Rule-based chopping | Dictionary + Linguistics |

| Context Aware        | No                  | Sometimes                |
| Vocabulary Reduction | Good                | Good                     |
| Semantic Quality     | Lower               | Better                   |

### Why POS Matters
This is where lemmatization becomes intelligent.

Consider:
```

```

#### Sentence 1
```

```

POS:
```

```

Lemma:
```

```

#### Sentence 2
```

```

POS:
```

```

Lemma:
```
meeting
```

Same word.

Different lemmas depending on grammar.

#### Diagram

This is impossible with stemming.

### Python Example (NLTK)

````python
import nltk

nltk.download("wordnet")
````

````python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(
    lemmatizer.lemmatize("cars")
)
````

````python
print(
    lemmatizer.lemmatize(
        "running",
        pos="v"
    )
)
````

Notice:
```
POS provided
↓
Better result
```

### Example Comparison

````python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

word = "studies"

print(stemmer.stem(word))
print(lemmatizer.lemmatize(word))
````

```
Lemmatization
=
Cleaner Output
```

### Why Modern LLMs Rarely Use Lemmatization
#### Old NLP
```
Text
 ↓
Lemmatization
 ↓
Feature Extraction
```

#### Modern NLP
```
Text
 ↓
Tokenizer
 ↓
Embeddings
 ↓
Transformer
```

Reason:

Transformers learn relationships naturally.

Example:
```
play
playing
played
```

Embedding space learns:
```
These words are related
```

No need to convert everything to:
```
play
```
before training.

Therefore:
```
BERT
GPT
Llama
Gemma
Claude
```
typically do not use lemmatization.
