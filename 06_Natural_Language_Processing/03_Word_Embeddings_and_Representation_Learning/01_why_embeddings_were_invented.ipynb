### The Core Problem
Classical NLP techniques:
* One-Hot Encoding
* Bag of Words
* TF-IDF
* Co-occurrence Matrix
* LSA

all represented text numerically.

But they had one fundamental weakness:
```
They did not truly understand meaning.
```

##### Example
Words:
```
king

queen
```

Humans know:
```
Very Similar
```

Classical NLP sees:
```
Different Tokens

Nothing more.
```

This is the problem embeddings were invented to solve.

### Limitation of One-Hot Encoding
Vocabulary:
```
cat

dog

airplane
```

One-Hot Representations:
```python
cat       = [1,0,0]

dog       = [0,1,0]

airplane  = [0,0,1]
```

Distance:
```
cat ↔ dog

equals

cat ↔ airplane
```

**Diagram**
```
cat       [1,0,0]

dog       [0,1,0]

airplane  [0,0,1]

All equally distant.
```

Problem:
```
No Semantic Similarity
```

Humans know:
```
cat ≈ dog
```
One-Hot Encoding does not.

### Limitation of Bag of Words
Document A
```
I love NLP
```

Document B
```
NLP love I
```

BoW:
```python
[1,1,1]
```
for both.

Problem:
```
No Word Order
```

Also:
```
good

excellent
```

are unrelated.

No semantic understanding.

### Limitation of TF-IDF
TF-IDF was a huge improvement.

It measures:
```
Importance
```

instead of:
```
Raw Counts
```

##### Example
```
deep learning
```

may receive a high score.

But:
```
deep learning

neural networks
```
are still treated as different terms.

Problem:
```
No Semantic Relationships
```

### Limitation of LSA
LSA introduced latent semantics.

Huge improvement.

However:
```
bank
```

still gets:
```
One Vector
```

Meaning:
```
Financial Bank

River Bank
```

cannot be separated.

Problem:
```
Static Meaning
```

### The Dream
Researchers wanted:
```
Words with Similar Meaning
        ↓
Similar Vectors
```

Desired Behavior
```
king
queen
↓
Close Together

cat
dog
↓
Close Together

cat
airplane
↓
Far Apart
```

**Diagram**
```
                 queen

                   ●

          king ●


cat ●

dog ●




                          airplane ●
```

This became the goal of word embeddings.

### What is a Word Embedding?
#### Definition
A word embedding is a dense numerical vector that captures semantic meaning and relationships between words.

##### Example
Instead of:
```python
cat = [1,0,0,0,0]
```

use:
```python
cat = [0.82, 0.14, -0.27]

dog = [0.79, 0.12, -0.24]
```

These vectors are:
* Dense
* Compact
* Meaningful

### Sparse vs Dense
#### One-Hot
```
[0,0,0,0,1,0,0,0,0]
```

#### Embedding
```
[0.52, -0.17, 0.89]
```

**Diagram**
```
One-Hot

0 0 0 0 1 0 0 0 0


Embedding

0.52 -0.17 0.89
```

This is called:
```
Dense Representation
```

### The Distributional Hypothesis Revisited
Remember:
```
Words appearing in similar contexts tend to have similar meanings.
```

##### Example
Corpus:
```
The cat drinks milk

The dog drinks milk

The cat eats fish

The dog eats meat
```

Observe:
```
cat

dog
```

appear in similar contexts.

Embedding model learns:
```
cat ≈ dog
```

Automatically.

This is the foundation of Word2Vec.

### What Makes Embeddings Special?
Embeddings learn:
```
Relationships
```

instead of:
```
Counts
```

Classical NLP
```
Word
      ↓
Count
```

Embeddings
```
Word
      ↓
Meaning
```

Huge difference.

### Semantic Similarity
Desired behavior:

Words
```
doctor

physician
```

Embedding vectors:
```
Very Similar
```

Words
```
car

automobile
```

Embedding vectors:
```
Very Similar
```

Words
```
cat

banana
```

Embedding vectors:
```
Very Different
```

This was revolutionary.

### The Famous King–Queen Example
One of the most famous NLP examples.

Embeddings learn relationships such as:
```
King - Man + Woman ≈ Queen
```

**Diagram**
```
man -------- king

 |             |

 |             |

woman ----- queen
```

This demonstrates:
```
Embeddings capture relationships.
```
Not just similarity.

Vocabulary:
```
100,000 words
```

One-Hot:
```
100,000 dimensions
```

per word.

Embedding:
```
100 dimensions
```

or
```
300 dimensions
```
Much smaller.

Example
```
100000
     ↓
300
```
Massive reduction.

### Better Generalization
Suppose training data contains:
```
cat

dog
```

Model learns:
```
animals
```

Now sees:
```
puppy
```

Embedding similarity helps the model generalize.

TF-IDF cannot do this effectively.

### Why Embeddings Changed NLP Forever
Before embeddings:
```
Feature Engineering
```

dominated.

Engineers manually created:
* BoW
* TF-IDF
* N-Grams

After embeddings:
```
Representation Learning
```

became dominant.

Instead of manually designing features:

This was a paradigm shift.
