### What is an Embedding Space?
Recall:

Word2Vec converts words into vectors.

Example:
```python
cat = [0.82, 0.15]

dog = [0.79, 0.17]

airplane = [-0.63, 0.91]
```

Instead of viewing them as numbers, imagine each word as a point in space.

#### Diagram
```
        dog ●

             ● cat


---------------------------->

                            ● airplane
```

This space is called: **Embedding Space**

#### Definition
An embedding space is a high-dimensional vector space where words are represented as points.

Words with similar meanings tend to be located near each other.

### Why Similar Words Become Close
Consider corpus:
```
The cat drinks milk

The dog drinks milk

The cat eats fish

The dog eats meat
```

Word2Vec repeatedly observes:
```
cat

dog
```

in similar contexts.

Therefore:
```
cat

dog
```

learn similar embeddings.

#### Diagram
```
Animal Region
dog ●
cat ●
puppy ●
kitten ●



Vehicle Region
car ●
truck ●
bus ●
```

Meaning emerges from context.

### High-Dimensional Reality
Often embeddings are drawn in:
```
2 Dimensions

or

3 Dimensions
```
for visualization.

Real embeddings usually have:
```
50
100
200
300
768
1024
```

dimensions.

Example:
```python
cat = [
 0.24,
-0.81,
 0.55,
 ...
 0.72
]
```

Humans cannot visualize these spaces directly.

But mathematics still works.

```
car
truck
bus
bike
```

cluster together.

**Countries**
```
India
France
Italy
Germany
```

cluster together.

#### Diagram
```
Animals

cat ●
dog ●
lion ●


Vehicles

car ●
truck ●
bus ●


Countries

India ●
France ●
Italy ●
```

No human manually created these groups.

### Measuring Similarity
Suppose:
```python
cat = [0.82, 0.15]
dog = [0.79, 0.17]
```

Cosine Similarity:
```
Very High
```

Suppose:

```python
cat = [0.82, 0.15]
airplane = [-0.63, 0.91]
```

Cosine Similarity:
```
Very Low
```

This is why vector search works.


### The Surprising Discovery
Researchers expected:
```
Similar Words
      ↓
Similar Vectors
```

They did NOT expect:
```
Relationships
      ↓
Vector Directions
```

This was the breakthrough.


### Understanding Vector Offsets
Suppose:
```
King

and

Queen
```

have similar relationships to:
```
Man

Woman
```

Word2Vec learns:
```
King → Queen

and

Man → Woman
```

as similar directional movements.

#### Diagram
```
man ---------------- king

 |                     |

 |                     |

woman ------------- queen
```

Notice:
```
man → woman

and

king → queen
```

have similar directions.


### The Famous Analogy
Most famous result:
```
king - man + woman ≈ queen
```

#### Step 1
```
king - man
```

removes:
```
Male Component
```

#### Step 2
```
+ woman
```

adds:
```
Female Component
```

#### Result
```
queen
```

#### Diagram
```
king - man + woman
     ↓
queen
```

This shocked the NLP community.

### Another Example
Countries and Capitals.

Relationship:
```
France → Paris
```

Relationship:
```
Italy → Rome
```

Word2Vec learns:
```
Paris - France + Italy ≈ Rome
```

#### Diagram
```
France -------- Paris

   |               |

   |               |

 Italy -------- Rome
```

Again:
```
Same Vector Direction
```

### Why Does This Happen?
Because words appearing in similar contexts learn similar relationships.

Examples:
```
king

queen
```

appear in:
```
royalty

throne

monarchy
```

Similarly:
```
man

woman
```

appear in:
```
human

person

adult
```

These contextual patterns create consistent vector directions.


### What Embeddings Actually Learn
Many people think embeddings learn:
```
Dictionary Definitions
```

Wrong.

Embeddings learn:
```
Statistical Relationships
```

from context.

Example:

Word2Vec never sees:
```
cat = animal
```

It only sees:
```
cat drinks milk

cat eats fish

cat is a pet
```

Meaning emerges from patterns.

### Common Semantic Relationships
Word2Vec learns many relationship types.

#### Gender
```
king → queen

man → woman

actor → actress
```

#### Country–Capital
```
France → Paris

Italy → Rome

India → Delhi
```

#### Verb Tense
```
walk → walked

play → played

jump → jumped
```

#### Singular–Plural
```
car → cars

dog → dogs

cat → cats
```

#### Comparative
```
good → better

small → smaller
```

These relationships emerge automatically.


### Word2Vec Limitations
#### Limitation 1 — Static Embeddings
Example:
```
bank
```

has one vector.

But:
```
river bank

financial bank
```

are different meanings.

Word2Vec cannot distinguish.

#### Limitation 2 — No Sentence Context
Sentence:
```
I deposited money in a bank
```

Sentence:
```
I sat near the river bank
```

Same embedding.

#### Limitation 3 — Rare Words
Very rare words may receive poor embeddings.

These limitations led to:
```
FastText

ELMo

BERT

GPT
```
