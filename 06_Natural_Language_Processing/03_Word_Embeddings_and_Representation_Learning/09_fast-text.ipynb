### The Problem with Word2Vec & GloVe
Recall:

**Word2Vec**
```
cat → vector
dog → vector
car → vector
```

**GloVe**
```
cat → vector
dog → vector
car → vector
```

Both models assume:
```
One Word
    ↓
One Embedding
```

This assumption causes several practical problems.


#### Problem 1 - Rare Word
Suppose a corpus contains:
```
cat
```

100,000 times.

But:
```
microservice
```

appears only:
```
5 times
```

Word2Vec learns:
```
cat
```

very well.

But:
```
microservice
```

receives very little training.

Result:
```
Poor Embedding Quality
```

#### Problem 2 - Out-of-Vocabulary (OOV)
Training vocabulary:
```
cat
dog
car
```

During inference:
```
GenAIOps
```

appears.

Word2Vec:
```
No Embedding Available
```

Why?

Because:
```
GenAIOps
```
was never seen during training.

This is called:
```
Out Of Vocabulary (OOV)
```

#### Problem 3 - Misspelling
Example:

Training word:
```
transformer
```

User types:
```
transfomer
```

(missing the letter **r**)

Word2Vec treats it as:
```
Completely Different Word
```

Result:
```
No Semantic Relationship
```

#### Problem 4 - Morphologically Rich Language
Consider:
```
play
playing
played
player
```

Word2Vec treats them as:
```
Four Independent Words
```

However, humans know they share the same root:
```
play
```

Important linguistic information is lost.

### The FastText Insight
Researchers at Meta (Facebook) asked:
```
Why learn embeddings
for whole words only?
```

What if embeddings were learned for:
```
Parts of Words?
```

This idea became:
```
FastText
```

### Core Idea of FastText
Instead of:
```
Word
   ↓
Embedding
```

FastText uses:
```
Word
   ↓
Character Subwords
   ↓
Embedding
```

This is the key innovation behind FastText.

### What Are Character N-Grams
Suppose the word is:
```
apple
```

FastText first adds boundary symbols:
```
<apple>
```

Now generate character n-grams.

For n = 3:
```
<ap
app
ppl
ple
le>
```

For n = 4:
```
<app
appl
pple
ple>
```

These subword units become learnable features.


### Word Representation in FastText
Word2Vec:
```
apple
   ↓
One Vector
```

FastText:
```
apple

   ↓

Character N-Grams

   ↓

<ap
app
ppl
ple
le>

   ↓

Individual Vectors

   ↓

Average / Sum

   ↓

Final Word Embedding
```


#### Visual Intuition
```
apple

  ↓

Character N-Grams

  ↓

<ap
app
ppl
ple
le>

  ↓

Average / Sum

  ↓

Word Embedding
```

Instead of learning only a word vector, FastText learns vectors for the word's components.


### Why FastText Solves Rare Words
Suppose:
```
microservice
```

appears only twice.

Even if the full word is rare, many subwords are common:
```
micro
service
cro
erv
vice
```

These subwords may appear elsewhere in the corpus.

FastText leverages this shared information.

Result:
```
Better Rare Word Embeddings
```

### Why FastText Solves OOV Words
Suppose the model never saw:
```
GenAIOps
```

during training.

Word2Vec:
```
Unknown Word
```

FastText:

Break the word into subwords:
```
Gen
AI
Ops
GenAI
AIOps
```

Many of these fragments may already have learned embeddings.

Therefore:
```
Generate Embedding Anyway
```

This was one of FastText's biggest breakthroughs.


### Example
Training words:
```
machine
learning
deep
network
```

New word:
```
deepnetwork
```

Word2Vec:
```
Unknown Word
```

FastText:
```
deep
network
```

Subwords already exist.

Therefore:
```
Meaningful Embedding Generated
```

### Morphological Understanding
Consider:
```
play
playing
played
player
```

Shared subword:
```
play
```

appears in all forms.

FastText learns:
```
play
```

is important.

Result:
```
playing
played
player
```

receive related embeddings.

### Visual Intuition of Morphology
Word2Vec:
```
play

playing

played

↓

Separate Vectors
```

FastText:
```
play
playing
played
player

↓

Shared Subwords

↓

Related Vectors
```

Diagram:
```
        play

      /  |  \

 played playing player

      ↓

Shared Subword Features
```

### FastText Architecture
FastText is essentially:
```
Word2Vec
+
Character N-Grams
```

CBOW Version:
```
Subwords
    ↓
CBOW
    ↓
Embedding
```

Skip-Gram Version:
```
Subwords
    ↓
Skip-Gram
    ↓
Embedding
```

FastText extends Word2Vec rather than replacing it completely.

### Example Workflow
Word:
```
transformer
```

Generate character n-grams:
```
tra
ran
ans
nsf
sfo
for
orm
rme
mer
```

Each n-gram has:
```
Its Own Vector
```

Combine:
```
Sum / Average
```

Result:
```
transformer embedding
```

### Advantages of FastText
#### Better Rare Word Handling
Uses subword information to learn meaningful representations.

#### OOV Support
Can generate embeddings for unseen words.

#### Robust to Misspellings
Example:
```
transformer
transfomer
```

share many subwords.

Therefore:
```
Embeddings Remain Similar
```

#### Better Morphological Understanding
Especially useful for:
- German
- Turkish
- Finnish
- Hindi
- Arabic

and other morphologically rich languages.

### Limitations of FastText
#### Static Embeddings
Word:
```
bank
```

always gets:
```
One Vector
```

Cannot distinguish:
```
river bank
financial bank
```

#### No Sentence Context
Sentence A:
```
I deposited money in a bank
```

Sentence B:
```
I sat near the river bank
```

FastText produces:
```
Same Embedding
```

#### Still Word-Level
FastText is not truly contextual.

This limitation later motivated:
- ELMo
- BERT
- GPT


### Word2Vec vs GloVe vs FastText
| Feature | Word2Vec | GloVe | FastText |
|----------|----------|--------|----------|
| Dense Embeddings | Yes | Yes | Yes |
| Uses Context | Yes | Indirectly | Yes |
| Uses Subwords | No | No | Yes |
| Handles Rare Words | Poorly | Poorly | Well |
| Handles OOV Words | No | No | Yes |
| Robust to Misspellings | No | No | Yes |
| Static Embeddings | Yes | Yes | Yes |
