How can a machine learn embeddings automatically?

The answer came from an idea that changed NLP forever:
```
Language Modeling
```

### The Historical Problem
Before Word2Vec, NLP mostly relied on:
* BoW
* TF-IDF
* N-Grams
* LSA

These methods:
- Required Feature Engineering
- Were Sparse
- Had Limited Semantics


Researchers wanted:
```
Automatic Feature Learning
```

Instead of:
```
Humans Designing Features
```

### The Big Question
Suppose, a sentence:
```
I love NLP
```

Humans know:
```
love is related to I

NLP is related to love
```

But how can a machine discover this?

Researchers asked:
> What if we force a model to predict missing words?

During prediction, maybe it will learn word relationships automatically.

This idea became:
```
Language Modeling
```

### What is a Language Model?
#### Definition
A Language Model (LM) is a model that learns the probability of word sequences in a language.

Simple intuition:

A language model answers:
```
What word is likely to come next?
```

##### Example
```
I drink hot _____
```

Likely answers:
* tea
* coffee

Unlikely answer:
* airplane

The model learns:
```
Language Patterns
```
from large amounts of text.

### Classical Language Models (N-Gram Models)
Before neural networks, language models used N-Grams.

##### Example Corpus
```
I love NLP

I love AI

I love Python
```

Input:
```
I love
```

Prediction:
```
NLP

AI

Python
```
Based on frequency counts.

**Diagram**
```
Previous Words
       ↓
Frequency Counts
       ↓
Next Word Prediction
```

Problems:
* Huge Tables
* Sparse Data
* Poor Generalization

### The Neural Network Idea
Researchers proposed:

Instead of storing counts:
```
Learn Patterns Using
A Neural Network
```

New idea:
```
Words
   ↓
Neural Network
   ↓
Predict Next Word
```

If prediction becomes accurate:
```
The network must learn
something about language.
```

This insight led to:
```
Neural Language Models
```

### The Famous 2003 Breakthrough
One of the most important papers in NLP history:
```
A Neural Probabilistic Language Model
```

Authors included:
```
Yoshua Bengio
```

Core idea:
```
Learn Word Representations
While Learning
Next Word Prediction
```

This was revolutionary.

Before this:
```
Representations
were manually designed.
```

After this:
```
Representations
could be learned.
```

### Neural Language Model Architecture
Simplified architecture:
```
Input Words
      ↓
Embeddings
      ↓
Neural Network
      ↓
Prediction
```

**Diagram**
```
Words

 I      love

  ↓       ↓

Embeddings

  ↓

Neural Network

  ↓

Predict

 NLP
```

Notice something important:
```
Embeddings
appear inside the model.
```

This is the birth of learned embeddings.

### Where Do Embeddings Come From?
This is the key insight.

Initially:
```
Embeddings
=
Random Numbers
```

Example:
```python
I     = [0.12, -0.45]

love  = [0.81, 0.22]

NLP   = [-0.17, 0.39]
```

Random initialization.

During training:
```
Prediction Error
      ↓
Backpropagation
      ↓
Update Embeddings
```

Eventually:
```
Words used similarly
develop similar vectors.
```

This was a huge discovery.

### Why Prediction Creates Meaning
Suppose corpus contains:
```
The cat drinks milk

The dog drinks milk

The cat eats fish

The dog eats meat
```

To predict correctly:
```
cat

dog
```
must behave similarly.

The neural network gradually learns:
```
cat ≈ dog
```

because they appear in similar contexts.

This automatically produces:
```
Semantic Similarity
```

No human rules required.

### The Learning Process
Training loop:
```
Sentence
    ↓
Create Prediction Task
    ↓
Predict Word
    ↓
Calculate Error
    ↓
Backpropagation
    ↓
Update Embeddings
    ↓
Repeat Millions of Times
```

Over time:
```
Embeddings improve
```
and
```
Predictions improve
```

### A Concrete Example
Suppose vocabulary:
```
I
love
NLP
AI
```

Sentence:
```
I love NLP
```

Training Sample:
```
Input  → I love

Target → NLP
```

Initially:
```
Prediction:

AI
```

Wrong.

Error:
```
High
```

Backpropagation updates:
```
Embedding(I)

Embedding(love)

Neural Network Weights
```

After thousands of updates:
```
Prediction:

NLP
```

Correct.

While learning prediction:
```
The model learns meaningful embeddings.
```

### Why Similar Words Become Similar Vectors
Imagine the model repeatedly sees:
```
The cat drinks milk

The dog drinks milk
```

Prediction tasks become almost identical.

The network discovers:
```
cat

dog
```

can often be substituted.

Eventually:
```python
cat = [0.82, 0.15]

dog = [0.79, 0.17]
```

Very similar vectors.

This is not programmed manually.

The model discovers it automatically.

### Representation Learning
This is one of the biggest ideas in modern AI.

Old NLP:
```
Human Engineer
        ↓
Design Features
        ↓
Model
```

Neural NLP:
```
Data
 ↓
Neural Network
 ↓
Learn Features
 ↓
Prediction
```

The network learns its own representation.

This is called:
```
Representation Learning
```

### Why This Was Revolutionary
Before neural language models:
```
Words
      ↓
Manual Features
      ↓
Prediction
```

After neural language models:
```
Words
      ↓
Learned Embeddings
      ↓
Prediction
```

For the first time:
```
Meaning emerged automatically
from data.
```

### Connection to Word2Vec
Neural language models proved something incredible:
```
Next Word Prediction
      ↓
Creates Useful Embeddings
```

Researchers then asked:
```
Can we simplify this idea
and learn embeddings faster?
```

This led directly to:
```
Word2Vec
```
