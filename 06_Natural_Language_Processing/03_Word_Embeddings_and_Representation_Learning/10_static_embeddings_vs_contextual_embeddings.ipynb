### The Journey So Far
So far:
```
Word2Vec
    ↓
Semantic Embeddings

GloVe
    ↓
Global Semantic Embeddings

FastText
    ↓
Subword-Aware Embeddings
```

These were major breakthroughs.

However, they all suffer from the same core limitation.


### The Core Problem
Consider the word:
```
bank
```

Sentence 1:
```
I deposited money in the bank.
```

Meaning:
```
Financial Institution
```

Sentence 2:
```
I sat near the river bank.
```

Meaning:
```
Side of a River
```

Humans immediately understand the difference.

But Word2Vec, GloVe, and FastText do not.

They see:
```
bank
    ↓
One Vector
```

### The Static Embedding Problem
Word2Vec:
```
bank

↓

[0.71, -0.28, 0.44, ...]
```

GloVe:
```
bank

↓

[0.63, -0.15, 0.38, ...]
```

FastText:
```
bank

↓

[0.69, -0.22, 0.41, ...]
```

Regardless of context:
```
Same Word
     ↓
Same Vector
```

This is called: **Static Embedding**

#### Definition of Static Embeddings
A **static embedding** assigns exactly one vector to a word regardless of where that word appears.

Formula:
```
Word
  ↓
Fixed Vector
```

Example:
```
bank

↓

[0.71, -0.28, 0.44]
```

always.


### Visualizing Static Embeddings
```
Financial Bank

      \
       \
        bank
       /
      /

River Bank

      ↓

 One Vector
```

Different meanings.

Same representation.

This is the biggest weakness of Word2Vec-style embeddings.


### Another Example: Apple
Sentence A:
```
I ate an apple.
```

Meaning:
```
Fruit
```

Sentence B:
```
Apple released a new iPhone.
```

Meaning:
```
Technology Company
```

Static embeddings:
```
apple

↓

One Vector
```

Unable to distinguish meanings.

### Polysemy
#### Definition
**Polysemy** occurs when a single word has multiple meanings.

Examples:
```
bank
apple
mouse
bat
crane
```


#### Examples of Polysemy
##### Mouse
Can mean:
```
Animal
```

or
```
Computer Device
```

##### Bat
Can mean:
```
Flying Mammal
```

or
```
Cricket Bat
```

##### Crane
Can mean:
```
Bird
```

or
```
Construction Machine
```

Static embeddings struggle with all of these.

### What Is Really Needed
Instead of:
```
One Word
     ↓
One Vector
```

We want:
```
Word + Context
        ↓
Meaning-Specific Vector
```

Example:
```
I deposited money in the bank.
```
↓
```
bank = Financial Vector
```

Example:
```
I sat near the river bank.
```
↓
```
bank = River Vector
```

Different meanings.

Different vectors.

### Contextual Embeddings
#### Definition
A **contextual embedding** generates a different vector for the same word depending on its surrounding context.

Instead of:
```
Word
   ↓
Fixed Lookup
```

We use:
```
Word + Context
        ↓
Embedding
```

#### Example of Contextual Embeddings
Word:
```
bank
```

Context:
```
money
loan
deposit
```
↓
```
Financial Vector
```

Context:
```
river
water
shore
```
↓
```
River Vector
```

Same word.

Different embeddings.


#### Visual Intuition
Sentence A:
```
money → bank
```
↓
```
Financial Embedding
```

Sentence B:
```
river → bank
```
↓
```
River Embedding
```

This was a revolutionary breakthrough.


### Static vs Contextual Embeddings
| Feature | Static Embeddings | Contextual Embeddings |
|----------|----------|----------|
| One Word = One Vector | Yes | No |
| Context Aware | No | Yes |
| Handles Polysemy | Poorly | Well |
| Meaning Changes Dynamically | No | Yes |
| Modern NLP Standard | No | Yes |


### Examples of Static Embeddings
```
Word2Vec
GloVe
FastText
```

All produce:
```
One Word
    ↓
One Vector
```

### Examples of Contextual Embeddings
```
ELMo
BERT
GPT
```

All produce:
```
Word + Context
        ↓
Context-Specific Vector
```

### Why Context Matters
Consider:

Sentence 1:
```
The bat flew away.
```

Meaning:
```
Animal
```

Sentence 2:
```
He bought a cricket bat.
```

Meaning:
```
Sports Equipment
```

Static embedding:
```
bat

↓

Same Vector
```

Contextual embedding:
```
bat

↓

Different Vectors
```

depending on context.


### The Key Insight
Meaning does not come from the word alone.

Meaning comes from:
```
Word
+
Surrounding Context
```

Example:
```
money
```

helps define:
```
bank
```

Example:
```
river
```

helps define:
```
bank
```

Therefore:
```
Meaning
=
Word + Context
```

This idea changed NLP forever.

### How Contextual Embeddings Are Learned
Static Embeddings:
```
Word
   ↓
Lookup Table
   ↓
Vector
```

Contextual Embeddings:
```
Sentence
     ↓
Analyze Context
     ↓
Generate Embedding
```

The embedding is created dynamically.

Not retrieved from a fixed table.

#### Real Example
Sentence:
```
The bank approved the loan.
```

Important context words:
```
approved
loan
```

Embedding generated:
```
Financial Meaning
```

Sentence:
```
The fisherman sat on the bank.
```

Important context words:
```
fisherman
sat
```

Embedding generated:
```
River Meaning
```

Different vectors for the same word.


### Why This Was a Huge Breakthrough
Before contextual embeddings:
```
Word Meaning
      ↓
Fixed
```

After contextual embeddings:
```
Word Meaning
      ↓
Dynamic
```

This dramatically improved:
- Machine Translation
- Question Answering
- Information Retrieval
- Search
- Chatbots
- Text Generation
- Summarization


### Why Transformers Need Contextual Embeddings
Modern Transformers process:
```
Entire Sentence
```

instead of:
```
Single Word
```

This allows them to generate:
```
Context-Aware Representations
```

for every token.

This is why:
- BERT understands word meaning based on context.
- GPT generates context-aware text.
- Modern LLMs outperform Word2Vec-style systems.
