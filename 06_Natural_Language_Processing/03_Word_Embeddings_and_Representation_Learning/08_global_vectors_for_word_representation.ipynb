After Word2Vec became successful, researchers noticed something interesting:
```
Word2Vec = Excellent Embeddings

But...

It only uses local context information.
```

Researchers wondered:

If combine the strengths of:
- Word2Vec (prediction-based learning)
- Co-occurrence matrices (global corpus statistics)

This idea led to: **GloVe (Global Vectors for Word Representation)**


### Why Was GloVe Created?
To understand GloVe, we must first understand the limitations of Word2Vec.

#### Word2Vec's Strength
Word2Vec learns from:
```
Local Context
```

Example:
```
The cat drinks milk
```

For:
```
cat
```

Word2Vec looks at nearby words:
```
drinks

milk
```

and learns embeddings.

This works very well.

#### But There Is A Problem
Word2Vec learns through:
```
Prediction Tasks
```

such as:
```
Predict Context

or

Predict Target
```

It does **not explicitly use**:
```
Global Corpus Statistics
```

Researchers believed:
```
Entire Corpus
```

contains useful semantic information.


### Local vs Global Information
#### Local Information
Looks only at nearby words.

Example:
```
The cat drinks milk
```

For:
```
cat
```

Local context:
```
drinks

milk
```

Word2Vec mainly learns from this.

#### Global Information
Looks at:
```
Entire Corpus
```

Example:

Suppose:
```
ice
```

appears with:
```
cold

snow

winter

frozen
```
throughout the corpus.

And:
```
steam
```

appears with:
```
hot

boiling

water
```
throughout the corpus.

These global patterns contain semantic information.


### The Core Idea Behind GloVe
GloVe starts with a:
```
Word Co-occurrence Matrix
```

Example:

#### Corpus
```
I like NLP

I like AI
```

#### Co-occurrence Matrix
| Word | I | like | NLP | AI |
|--------|---|------|------|----|
| I | 0 | 2 | 1 | 1 |
| like | 2 | 0 | 1 | 1 |
| NLP | 1 | 1 | 0 | 0 |
| AI | 1 | 1 | 0 | 0 |

This matrix contains:
```
Global Statistical Information
```
about the corpus.

GloVe uses this information directly.


### The Key Insight
The GloVe paper made a fascinating observation.

Consider:
```
ice

and

steam
```

Look at how frequently they occur with:
```
solid

gas
```
Co-occurrence counts reveal meaning.

Example:
```
P(solid | ice)
```
might be high.

But:
```
P(solid | steam)
```
might be low.

The ratio:
```
P(solid | ice)
────────────────
P(solid | steam)
```
contains semantic information.

This was the key insight behind GloVe.


### What Does GloVe Learn?
GloVe learns embeddings such that:
```
Vector Relationships ≈ Co-occurrence Relationships
```

Goal:
```
Words appearing in similar
global contexts
↓
Similar embeddings
```
Very similar to Word2Vec.

But learned differently.


### High-Level Pipeline
```
Corpus
   ↓
Co-occurrence Matrix
   ↓
Global Statistics
   ↓
Optimization
   ↓
Word Embeddings
```

Compare with Word2Vec:
```
Corpus
   ↓
Prediction Task
   ↓
Neural Network
   ↓
Word Embeddings
```

**How does GloVe differ from Word2Vec?**

- Word2Vec learns embeddings through prediction.
- GloVe learns embeddings through global co-occurrence statistics.


### Word2Vec vs GloVe Philosophy
#### Word2Vec
Learns by:
```
Prediction
```

Example:
```
love

↓

Predict:

I

NLP
```

Embeddings emerge.

#### GloVe
Learns by:
```
Matrix Factorization
+
Co-occurrence Statistics
```

Embeddings are optimized to preserve corpus-wide relationships.

### Visual Comparison
#### Word2Vec
```
Text
 ↓
Prediction
 ↓
Embeddings
```

#### GloVe
```
Text
 ↓
Co-occurrence Matrix
 ↓
Embeddings
```

### Why GloVe Became Popular
Researchers discovered:
```
Word2Vec

and

GloVe
```

produce similarly high-quality embeddings.

But GloVe offered:
```
Strong Theoretical Foundation
```

using global statistics.

For several years:
```
Word2Vec

GloVe
```
were the two dominant embedding methods.

### Embedding Space in GloVe
Like Word2Vec:
```
cat

dog
```

become nearby vectors.

#### Diagram
```
Animals
dog ●
cat ●
lion ●



Vehicles
car ●
truck ●
bus ●
```

Semantic clusters emerge.

### Vector Arithmetic Still Works
GloVe also learns:
```
king - man + woman ≈ queen
```

Example:
```
Paris - France + Italy ≈ Rome
```

#### Diagram
```
France -------- Paris

   |               |

   |               |

 Italy -------- Rome
```

Because semantic relationships are preserved.


### Why GloVe Was Important
It showed:
```
Global Statistics
```

can produce embeddings as powerful as:
```
Prediction-Based Learning
```

This was a major discovery.


### Advantages of GloVe
#### Uses Global Information
Looks at the entire corpus.

#### Efficient Training
Can leverage matrix operations efficiently.

#### Excellent Semantic Quality
Produces strong embeddings.

#### Good Interpretability
Based on explicit co-occurrence statistics.


### Limitations of GloVe
#### Limitation 1 — Static Embeddings
Word:
```
bank
```

gets:
```
One Vector
```

Problem:
```
Financial Bank

River Bank
```

share the same embedding.


#### Limitation 2 — No Sentence Context
Sentence A:
```
I deposited money in a bank
```

Sentence B:
```
I sat near the river bank
```

Same embedding.

#### Limitation 3 — Needs Co-occurrence Matrix
For huge corpora:
```
Storage Cost
```

can be large.

Building and storing co-occurrence matrices becomes expensive.


### Word2Vec vs GloVe
| Feature | Word2Vec | GloVe |
|----------|----------|--------|
| Learning Method | Prediction-Based | Count-Based |
| Uses Neural Network | Yes | No (Direct Optimization) |
| Uses Co-occurrence Matrix | No | Yes |
| Uses Global Statistics | Limited | Yes |
| Embedding Quality | Excellent | Excellent |
| Static Embeddings | Yes | Yes |
