### The Problem with Original Word2Vec
Revisit Skip-Gram.

Sentence:
```
I love NLP
```

Target word:
```
love
```

Context words:
```
I

NLP
```

Goal:
```
P(I | love)

P(NLP | love)
```
should be high.

How to compute these probabilities?

Using:
```
Softmax
```

### Full Softmax
Suppose vocabulary:
```
50,000 words
```

Target:
```
love
```

Output layer:
```
love → probability of word 1

love → probability of word 2

love → probability of word 3

...

love → probability of word 50,000
```

#### Diagram
```
            love

              ↓

     Output Layer

              ↓

  50,000 Probabilities
```

Every training step computes:
```
50,000 scores
```


### Why Is This Expensive?
Vocabulary:
```
50,000 words
```

Training Corpus:
```
100 million words
```

For every training example:
```
Compute 50,000 outputs
```

Complexity:
```
O(V)
```

Where:
```
V = Vocabulary Size
```

If:
```
V = 100,000
```
training becomes extremely expensive.

#### Problems
```
Too Slow
Too Much Memory
```

Researchers needed:
```
Faster Training
```

### Core Insight
Suppose target:
```
love
```

Actual context:
```
NLP
```

Do we really need to calculate probabilities for:
```
banana

airplane

elephant

transformer

coffee

...

all 100,000 words?
```
Not really.

Only need:
```
Positive Examples

and

A Few Negative Examples
```

This idea became:
```
Negative Sampling
```

### What is Negative Sampling?
#### Definition
Negative Sampling trains Word2Vec by learning to distinguish real context words from randomly sampled fake context words.

Instead of:
```
Target
      ↓
Predict Entire Vocabulary
```

we do:
```
Target
      ↓
Real Context?

or

Fake Context?
```

Much simpler.


### Positive and Negative Examples
Sentence:
```
I love NLP
```

Target:
```
love
```

Actual Context:
```
NLP
```

#### Positive Example
```
(love, NLP)
```

Label:
```
1
```

Meaning:
```
Real Context Pair
```

Now randomly sample:
```
banana

airplane

elephant
```

#### Negative Examples
```
(love, banana)

(love, airplane)

(love, elephant)
```

Labels:
```
0
```

Meaning:
```
Fake Context Pair
```

### Training Objective
The model learns:
```
(love, NLP)

↓

High Similarity
```

and
```
(love, banana)

↓

Low Similarity
```

#### Diagram
```
love + NLP
      ↓
 Real Pair
      ↓
 Score ↑


love + banana
      ↓
 Fake Pair
      ↓
 Score ↓
```

### Why Negative Sampling Works
Suppose corpus contains:
```
cat drinks milk

dog drinks milk
```

Positive pairs:
```
(cat, drinks)

(cat, milk)

(dog, drinks)

(dog, milk)
```

Negative pairs:
```
(cat, airplane)

(cat, banana)

(dog, transformer)
```

The model gradually learns:
```
cat ≈ dog
```
because they share positive contexts.

Semantic similarity emerges naturally.

### Computational Advantage
#### Original Softmax
Need:
```
100,000 outputs
```

#### Negative Sampling
```
1 Positive

+
5 Negative Samples
```

Only:
```
6 computations
```

instead of:
```
100,000 computations
```

Huge improvement.

### Complexity Comparison
#### Full Softmax
```
O(V)
```

Vocabulary:
```
V = 100,000
```

#### Negative Sampling
```
O(k)
```

Where:
```
k = number of negative samples
```

Typically:
```
5–20
```
Massive speedup.


### Visual Intuition
#### Full Softmax
```
love

 ↓

100,000 outputs
```

#### Negative Sampling
```
love

 ↓

NLP      ✓

banana   ✗

airplane ✗

coffee   ✗

dog      ✗
```

Only a few comparisons.

### Typical Negative Sampling Values
Common choices:
```
5 Negative Samples
```

for large datasets.

Sometimes:
```
10

15

20
```

### What Does the Model Actually Learn?
The model learns:
```
Dot Product
```

between word vectors.

For positive pair:
```
(cat, milk)
```

Dot Product:
```
High
```

For negative pair:
```
(cat, airplane)
```

Dot Product:
```
Low
```

Eventually:
```
Similar words cluster together.
```

### Why Negative Sampling Became Popular
Because it made Word2Vec:
```
Fast

Scalable

Practical
```

Without Negative Sampling:
```
Word2Vec

would not have achieved widespread adoption.
```


### Hierarchical Softmax
Another solution to the same problem.

Problem:
```
Full Softmax

is expensive.
```

Solution:
```
Hierarchical Softmax
```

Idea:
```
Don't compare against every word.

Use a Binary Tree instead.
```

### Binary Tree Representation
Suppose vocabulary:
```
cat

dog

apple

milk
```

Tree:
```
               ROOT

             /      \

            /        \

         cat        Internal

                   /       \

                dog      Internal

                         /      \

                     apple     milk
```

Each word becomes:
```
Path From Root
```

### Prediction Using Tree Traversal
Suppose target:
```
milk
```

Instead of:
```
50,000 comparisons
```

We only traverse:
```
ROOT
 ↓
Internal
 ↓
Internal
 ↓
milk
```

Much fewer operations.


### Complexity of Hierarchical Softmax
#### Full Softmax
```
O(V)
```

#### Hierarchical Softmax
```
O(log V)
```

Example:

Vocabulary:
```
100,000 words
```

Full Softmax:
```
100,000 computations
```

Hierarchical Softmax:
```
~17 computations
```

because:
```
log₂(100000) ≈ 17
```

Huge improvement.


### Visual Comparison
#### Full Softmax
```
ROOT

 ↓

Compare With

Word1
Word2
Word3
...
Word100000
```

#### Hierarchical Softmax
```
ROOT
 ↓
Left
 ↓
Right
 ↓
Left
 ↓
Target Word
```

Only follow a single path.

### Why Hierarchical Softmax Works
Instead of asking:
```
Which of 100,000 words is correct?
```

It asks a sequence of binary questions:
```
Left or Right?

Left or Right?

Left or Right?
```

until the target word is reached.

Much more efficient.


### Negative Sampling vs Hierarchical Softmax
| Feature        | Negative Sampling | Hierarchical Softmax |
| -------------- | ----------------- | -------------------- |
| Complexity     | O(k)              | O(log V)             |
| Speed          | Faster            | Slower               |
| Simplicity     | Simple            | More Complex         |
| Common Usage   | Very Common       | Less Common          |
| Large Corpora  | Excellent         | Good                 |
| Rare Words     | Good              | Often Better         |
| Implementation | Easy              | Harder               |
