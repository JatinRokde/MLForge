### What is CBOW?
#### Definition
CBOW predicts the target word using its surrounding context words.

In simple words:
```
Context Words
      ↓
Predict
      ↓
Target Word
```

##### Example
Sentence:
```
I love NLP
```

Target word:
```
love
```

Context words:
```
I

NLP
```

CBOW Training Example:
```
Input:
I, NLP

Output:
love
```

#### Diagram
```
I      NLP

 \     /

  \   /

 Context

    ↓

 Predict

  love
```

### Why Is It Called Continuous Bag of Words?
#### Bag of Words
Recall BoW:
```
Word Order Ignored
```

Example:
```
I NLP

NLP I
```
are treated the same.

CBOW does something similar.

It uses surrounding words but:
```
Does Not Care About Their Order
```

#### Continuous
The word:
```
Continuous
```

means:
```
Dense Vector Representation
```

instead of:
```
Sparse One-Hot Representation
```

Therefore:
```
Continuous Bag of Words = Dense Vector-Based BoW
```

### CBOW Architecture
```
Context Words
      ↓
 One-Hot Vectors
      ↓
Embedding Matrix
      ↓
Average Embeddings
      ↓
Output Layer
      ↓
Predicted Word
```

#### Simplified Architecture
```
Input Words

I      NLP

 ↓      ↓

Embedding Layer

 ↓      ↓

Average

     ↓

Hidden Vector

     ↓

Softmax

     ↓

love
```

### Embedding Matrix
This is the heart of Word2Vec.

Suppose:
```
Vocabulary Size = 5

Embedding Size = 3
```

#### Embedding Matrix
```python
W =
[
 [0.2, 0.8, 0.1],   # I
 [0.7, 0.3, 0.5],   # love
 [0.4, 0.9, 0.6],   # NLP
 [0.8, 0.2, 0.3],   # very
 [0.1, 0.4, 0.7]    # much
]
```

Shape:
```
Vocabulary Size × Embedding Dimension

5 × 3
```

#### Important Interview Point
```
The rows of this matrix become the final word embeddings.
```

### Embedding Lookup
Suppose input word:
```
I
```

One-Hot:
```
[1,0,0,0,0]
```

Matrix Multiplication:
```
OneHot × W
```

returns:
```
[0.2,0.8,0.1]
```

Similarly:
```
NLP
```

returns:
```
[0.4,0.9,0.6]
```

This process is called:
```
Embedding Lookup
```

### Averaging Context Embeddings
Context:
```
I

NLP
```

Embeddings:
```
I   = [0.2,0.8,0.1]

NLP = [0.4,0.9,0.6]
```

Average:
```
[
 (0.2+0.4)/2,
 (0.8+0.9)/2,
 (0.1+0.6)/2
]
```

Result:
```
[0.3,0.85,0.35]
```

This becomes:
```
Hidden Representation
```

#### Diagram
```
I Embedding

      \
       \
        Average
       /
      /

NLP Embedding

      ↓

Hidden Vector
```

### Output Layer
Now hidden vector:
```
[0.3,0.85,0.35]
```

must predict:
```
love
```

Output Layer computes scores for every vocabulary word.

Example:
```
I      → 1.2

love   → 3.8

NLP    → 0.9

very   → 0.7

much   → 0.5
```

Highest score:
```
love
```

Prediction is correct.

### Softmax
Scores are converted into probabilities.

#### Raw Scores
```
[1.2,3.8,0.9,0.7,0.5]
```

#### After Softmax
```
[
 0.05,
 0.83,
 0.04,
 0.04,
 0.04
]
```

Interpretation:
```
83% probability

love
```

Softmax ensures:
```
All Probabilities Sum To 1
```

### Loss Function
Suppose actual target:
```
love
```

Prediction:
```
0.83
```

Good prediction.
```
Small Loss
```

Suppose prediction:
```
0.05
```

Bad prediction.
```
Large Loss
```

CBOW uses:
```
Cross Entropy Loss
```

### Backpropagation
After computing loss:
```
Loss
  ↓
Backpropagation
  ↓
Update Weights
```

Most important update:
```
Embedding Matrix
```

gets updated.

This is how embeddings are learned.

#### Training Loop
```
Context
   ↓
Prediction
   ↓
Loss
   ↓
Backpropagation
   ↓
Update Embeddings
```

Repeated millions of times.

### Why Similar Words Get Similar Embeddings
Suppose corpus:
```
The cat drinks milk

The dog drinks milk

The cat eats fish

The dog eats meat
```

CBOW repeatedly sees:
```
cat

dog
```

in similar contexts.

To make good predictions:
```
cat

dog
```

gradually move closer in vector space.

Eventually:
```
cat = [0.82,0.15]

dog = [0.79,0.17]
```

Thus:
```
Semantic Similarity Emerges
```

automatically.

### Why CBOW Is Fast
CBOW combines context words:
```
Average

↓

One Hidden Vector

↓

One Prediction
```

Therefore:
```
Less Computation
```

than Skip-Gram.

This makes CBOW:
* Fast
* Memory Efficient

### Advantages of CBOW
#### Faster Training
Uses aggregated context.

#### Efficient
Less computation.

#### Good For Large Corpora
Works well on huge datasets.

#### Stable
Produces reasonable embeddings quickly.

### Limitations of CBOW
#### Ignores Context Order
```
I love NLP

NLP love I
```
treated similarly.

#### Context Averaging Loses Information
Different context words get blended.

#### Poorer For Rare Words
Compared to Skip-Gram.

#### Static Embeddings
```
bank
```
always gets one vector.

### CBOW vs Skip-Gram
| CBOW                      | Skip-Gram             |
| ------------------------- | --------------------- |
| Context → Target          | Target → Context      |
| Faster                    | Slower                |
| Less Memory               | More Memory           |
| Better For Frequent Words | Better For Rare Words |
| Simpler                   | More Accurate         |

```
CBOW is generally faster.

Skip-Gram generally produces
higher-quality embeddings.
```
