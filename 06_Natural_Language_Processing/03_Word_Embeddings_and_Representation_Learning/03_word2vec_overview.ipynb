### Current Topic in NLP History
Quick recap.

Classical NLP gave us:
* One-Hot Encoding
* Bag of Words
* TF-IDF
* LSA

These methods suffered from:
* Sparse Representations
* No True Semantics
* No Word Relationships
* Poor Generalization

Neural Language Models introduced:
```
Representation Learning
```

The key discovery was:
```
If a model learns to predict words,
it automatically learns meaningful word representations.
```

This idea led to:
```
Word2Vec
```

### What is Word2Vec?
#### Definition
Word2Vec is a shallow neural network that learns dense vector representations (embeddings) of words from their contexts.

Developed by researchers at Google in 2013.

Associated with:
```
Tomas Mikolov and colleagues.
```

The goal:
```
Word
  ↓
Dense Vector
```

##### Example
```python
cat = [0.82, 0.14, -0.27]

dog = [0.79, 0.12, -0.24]

airplane = [-0.63, 0.91, 0.11]
```

Notice:
```
cat ≈ dog

cat ≠ airplane
```

### Why Was Word2Vec Revolutionary?
Before Word2Vec:
```
Words
=
Symbols
```

After Word2Vec:
```
Words = Meaningful Vectors
```

For the first time:
* Semantic Similarity
* Relationships
* Analogies

could be learned automatically.

### The Core Idea
Word2Vec is based on:
```
Distributional Hypothesis
```

Recall:
> Words occurring in similar contexts tend to have similar meanings.

##### Example
```
The cat drinks milk

The dog drinks milk

The cat eats fish

The dog eats meat
```

Observe:
```
cat

dog
```
appear in similar contexts.

Word2Vec learns:
```
cat ≈ dog
```
without any human labeling.

### High-Level Architecture
Very simplified:
```
Context Words
      ↓
 Neural Network
      ↓
Target Word
```

or
```
Target Word
      ↓
 Neural Network
      ↓
Context Words
```
depending on the variant.

The network is intentionally tiny.

**Diagram**
```
Input Word(s)

      ↓

 Embedding Layer

      ↓

 Output Layer

      ↓

 Prediction
```

Important:
```
The embeddings learned in the hidden layer
become the final word vectors.
```

### The Two Variants of Word2Vec
Word2Vec has two training approaches:
1. CBOW
2. Skip-Gram

**Diagram**
```
           Word2Vec
               │
      ┌────────┴────────┐
      │                 │

    CBOW          Skip-Gram
```

These are among the most common NLP interview questions.

### CBOW (Continuous Bag of Words)
#### Core Idea
```
Context
   ↓
Predict Target Word
```

Sentence:
```
I love NLP
```

Training Example

Input:
```
I      NLP
```

Target:
```
love
```

**Diagram**
```
I      NLP

 \     /

  \   /

 Context

    ↓

 Predict

  love
```

Question being asked:
> Given neighboring words, what is the missing word?


### Skip-Gram
#### Core Idea
```
Target Word
     ↓
Predict Context Words
```

Sentence:
```
I love NLP
```

Input:
```
love
```

Targets:
```
I

NLP
```

**Diagram**
```
     love

       ↓

  ┌────┴────┐

  ↓         ↓

  I        NLP
```

Question being asked:
> Given a word, what surrounding words are likely?

### CBOW vs Skip-Gram Intuition
Suppose sentence:
```
Transformers revolutionized NLP
```

#### CBOW
```
Transformers     NLP

        ↓

 Predict

revolutionized
```

#### Skip-Gram
```
revolutionized

      ↓

Transformers

NLP
```

### Which One Is Better?
#### CBOW
Advantages:
* Faster
* Less Memory
* Works well on large datasets

Disadvantages:
* Less effective for rare words

#### Skip-Gram
Advantages:
* Better embeddings
* Handles rare words better
* Usually preferred

Disadvantages:
* Slower

#### Interview Answer
```
CBOW is faster.

Skip-Gram generally produces better embeddings,
especially for infrequent words.
```

### Context Window
Word2Vec uses a:
```
Context Window
```

Sentence:
```
I love NLP very much
```

Target:
```
NLP
```

Window Size = 2

Context:
```
I

love

very

much
```

**Diagram**
```
I   love   NLP   very   much

      ← 2 →

      ← 2 →
```

Important Hyperparameter:
```
Window Size
```

Larger Window:
```
More Context
```

Smaller Window:
```
More Local Meaning
```

### How Training Data Is Generated
Suppose:
```
I love NLP
```

Window Size = 1

#### CBOW Training Samples
```
Input      → Target

(I)        → love

(love)     → NLP

(NLP)      → love
```

#### Skip-Gram Training Samples
```
Input      → Target

love       → I

love       → NLP

NLP        → love
```

Millions of such training pairs are generated automatically.

### The Magic: Semantic Similarity Emerges
Nobody explicitly tells Word2Vec:
```
cat ≈ dog
```

Instead it repeatedly sees:
```
cat drinks milk

dog drinks milk

cat eats fish

dog eats meat
```

To predict correctly:
```
cat

dog
```

must behave similarly.

The network gradually learns:
```
cat ≈ dog
```

This is one of the most beautiful ideas in NLP.

### The Famous King–Queen Analogy
Word2Vec became famous because researchers discovered:
```
king - man + woman ≈ queen
```

**Diagram**
```
man -------- king

 |             |

 |             |

woman ----- queen
```

Embeddings learned:
* Gender Relationship
* Royalty Relationship
automatically.

### Inside the Word2Vec Network
Very simplified architecture:
```
Input Word
     ↓

 One-Hot Vector
     ↓

Embedding Matrix
     ↓

Dense Embedding
     ↓

Output Layer
     ↓

Prediction
```

Key Insight:
```
The embedding matrix itself
becomes the learned word embeddings.
```

After training:
```
We throw away the prediction layer
and keep the embedding matrix.
```

### Limitations of Word2Vec
#### Limitation 1 — Static Embeddings
One word:
```
bank
```

gets:
```
One Vector
```

Problem:
```
Financial Bank

River Bank
```

share the same embedding.

Called:
```
Static Embeddings
```

#### Limitation 2 — No Sentence-Level Context
Example:
```
Apple released a phone

I ate an apple
```

Same vector:
```
apple
```

Meaning differs.

Word2Vec cannot distinguish.

#### Limitation 3 — Out-of-Vocabulary (OOV) Words
Example:
```
GenAIX2026
```

Not seen during training.

No embedding available.

#### Limitation 4 — Entire Vocabulary Must Be Known
Vocabulary is fixed after training.

New words require retraining or updating.

### Why Word2Vec Still Matters
Modern LLMs do not use Word2Vec directly.

However:
```
Every Modern Embedding System
Descends From Word2Vec
```

Concepts introduced:
* Dense Vectors

remain fundamental.
