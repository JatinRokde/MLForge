BoW represents documents using word counts.

However, this raises an important question:
```
Are all words equally important?
```

Consider:

**Document A**
```
NLP NLP NLP NLP NLP
```

**Document B**
```
NLP
```

Intuitively:
```
"NLP" is more important in Document A

because it appears many times.
```

This idea leads to:
```
Term Frequency (TF)
```

### What is Term Frequency?
#### Definition
Term Frequency (TF) measures how often a term appears in a document.

Intuition:
```
More Occurrences
        ↓
Potentially More Important
```

##### Example
Document:
```
NLP is fun.
NLP is powerful.
I love NLP.
```

Word:
```
NLP
```

appears:
```
3 times
```

Therefore:
```
TF(NLP) = 3
```

### Why TF Is Needed?
Suppose:

#### Document 1
```
NLP NLP NLP NLP NLP
```

#### Document 2
```
NLP
```

BoW says:
```
NLP exists in both documents.
```

TF says:
```
Document 1 talks much more
about NLP.
```

This provides:
```
Importance
```

inside a document.

### Simplest TF Formula
Most basic version:

**Raw Count**
```
TF(term) = Number of occurrences
```

#### Example
Document:
```
I love NLP.
NLP is amazing.
```

Counts:
| Word    | Count |
| ------- | ----- |
| I       | 1     |
| love    | 1     |
| NLP     | 2     |
| is      | 1     |
| amazing | 1     |

Therefore:

```text
TF(NLP) = 2
```

### Example Calculation
Document:
```
AI AI NLP AI ML
```

Word Counts
| Word | Count |
| ---- | ----- |
| AI   | 3     |
| NLP  | 1     |
| ML   | 1     |

#### Term Frequencies
```
TF(AI)  = 3

TF(NLP) = 1

TF(ML)  = 1
```

Interpretation:
```
AI is most important
within this document.
```

### Problem with Raw Counts
Consider two documents.

#### Document A
```
NLP NLP NLP NLP NLP
```

Length:
```
5 words
```

#### Document B
```
NLP appears 10 times
inside a document of 1000 words.
```

Raw Counts:
```
Doc A = 5

Doc B = 10
```

Looks like:
```
Doc B is more about NLP
```

But actually:
```
Doc A is heavily focused on NLP.
```

Reason:
```
Document lengths differ.
```

### Normalized Term Frequency
To fix document length bias:
```
TF = Term Count / Total Words
```

#### Formula
```
TF(t,d) = Term Count(t, d) / Total Words(d)
```

#### Example
Document:
```
AI AI NLP ML
```

Total Words:
```
4
```

Count(AI):
```
2
```

Normalized TF:
```
2 / 4

= 0.5
```

Count(NLP):
```
1
```

TF:
```
1 / 4

= 0.25
```

### Full Example
Document:
```
AI AI NLP ML
```

Total Terms:
```
4
```

Table
| Word | Count | TF   |
| ---- | ----- | ---- |
| AI   | 2     | 0.50 |
| NLP  | 1     | 0.25 |
| ML   | 1     | 0.25 |

Interpretation:
```
AI dominates the document.
```

### TF Values
Normalized TF always lies between:
```
0 and 1
```

Examples:
#### Word Absent
```
TF = 0
```

#### Word Appears Everywhere
```
TF ≈ 1
```

#### Most Words
```
Small Fractions
```

### Why TF Alone Is Not Enough
Consider a corpus.

#### Document 1
```
the cat sat
```

#### Document 2
```
the dog ran
```

#### Document 3
```
the bird flew
```

TF values:
```
the
```

appear frequently.

TF says:
```
Very Important
```

Reality:
```
the
```

contains little information.

Problem:
TF only looks at:
```
Single Document
```

It ignores:
```
Entire Corpus
```

This leads to:
```
IDF
```

### Visual Intuition


```
AI   ████████

NLP  ██

```

TF emphasizes:
```
Frequently occurring words
```

inside a document.


### Python Example
#### Using Counter




````python
from collections import Counter

document = [
    "AI",
    "AI",
    "NLP",
    "ML"
]

counts = Counter(document)

print(counts)
````

#### Normalized TF

````python
total_terms = len(document)

for word, count in counts.items():
    tf = count / total_terms
    print(word, tf)
````

### Advantages of TF
* Simple to understand
* Easy to compute
* Captures term importance within a document
* Works well as a building block for TF-IDF
* Useful for document ranking and retrieval

### Limitations of TF
* Cannot identify globally common words
* Gives high importance to stopwords
* Considers only a single document
* Ignores semantics
* Ignores word order

Example:
```
the
```
may receive a high TF score even though it carries little meaning.


### TF in the Classical NLP Pipeline
```

    ↓

    ↓

```
