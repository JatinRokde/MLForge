### What is an N-Gram?
#### Definition
An N-Gram is a sequence of N consecutive tokens.

Examples:
```
N = 1 → Unigram

N = 2 → Bigram

N = 3 → Trigram

N = 4 → Four-Gram
```

Sentence:
```
I love NLP
```

#### Unigrams
```python
[
 "I",
 "love",
 "NLP"
]
```

#### Bigrams
```python
[
 "I love",
 "love NLP"
]
```

#### Trigram
```python
[
 "I love NLP"
]
```

#### Diagram
```
I ---- love ---- NLP

Unigrams:
[I] [love] [NLP]

Bigrams:
[I love]
[love NLP]

Trigram:
[I love NLP]
```

### Why N-Grams Is Needed?
Consider:

#### Sentence A
```
not good
```

#### Sentence B
```
good
```

BoW sees:
```
not
good
```

as separate words.

May fail to understand:
```
not good
```

is negative.

Bigram captures:
```
not good
```

as a single feature.

This is the key idea:
```
Capture Local Context
```

### General Formula
Sentence contains:
```
T Tokens
```

N-Grams generated:
```
T - N + 1
```

#### Example
Sentence:
```
I love NLP
```

Tokens:
```
3
```

Bigrams:
```
3 - 2 + 1 = 2
```

Correct:
```
I love

love NLP
```

### Advantages of N-Grams
#### Captures Local Word Order
BoW:
```
No Order
```

N-Grams:
```
Some Order Preserved
```

#### Captures Phrases
Examples:
```
New York

Machine Learning

Deep Learning

Artificial Intelligence
```

Useful for:
* Classification
* Search
* Information Retrieval

#### Improves Context
Example:
```
not good
```

becomes meaningful.


### Major Limitations
#### Limitation 1: Vocabulary Explosion
Vocabulary:
```
10000 words
```

Possible bigrams:
```
Millions
```

Possible trigrams:
```
Much Larger
```

Memory becomes huge.

#### Limitation 2: Sparsity
Many N-Grams appear only once.

Example:
```
transformers revolutionized modern NLP
```

Rare phrase.

Vector becomes mostly zeros.


#### Limitation 3: Long-Range Context Missing
Sentence:
```
The movie that I watched yesterday was amazing
```

Bigram sees:
```
The movie

movie that

that I

...
```

Cannot connect:
```
movie ↔ amazing
```

effectively.

This limitation eventually led to:
```
RNNs

LSTMs

Transformers
```

### Python Example




````python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love NLP"
]

vectorizer = CountVectorizer(
    ngram_range=(2,2)
)

X = vectorizer.fit_transform(
    documents
)

print(
    vectorizer.get_feature_names_out()
)
````

### N-Grams in Language Models
Before deep learning:

Language models used N-Grams.

Example:
```
I love ___
```

From training data:
```
I love NLP

I love AI

I love Python
```

Predict next word using:
```
Previous N-1 Words
```

These are called:
```
Statistical Language Models
```

### N-Grams vs Bag of Words
| Feature         | BoW     | N-Grams |
| --------------- | ------- | ------- |
| Word Counts     | Yes     | Yes     |
| Word Order      | No      | Partial |
| Context         | None    | Local   |
| Vocabulary Size | Smaller | Larger  |
| Sparsity        | High    | Higher  |
