### Motivation

Suppose, movie reviews:
```
Review 1
I love this movie

Review 2
This movie is terrible
```

A machine learning model cannot directly process these sentences.

Need:
```
Numerical Features
```
BoW converts documents into vectors.


### What is Bag of Words?
#### Definition
Bag of Words represents a document using the frequency (count) of words appearing in that document.

Core Idea
* Keep the words
* Ignore grammar
* Ignore word order
* Count occurrences

The term **"Bag"** means:
```
Words are treated like objects inside a bag.

Order does not matter.
```

##### Example
Sentence A:
```
I love NLP
```

Sentence B:
```
NLP love I
```

BoW sees both as:
```
Same document
```
because word counts are identical.

### Building a Bag of Words
#### Document 1
```
I love NLP
```

#### Document 2
```
I love AI
```

#### Step 1: Build Vocabulary
Collect unique words.
```python
[
 "I",
 "love",
 "NLP",
 "AI"
]
```

Vocabulary Size:
```
4
```

#### Step 2: Create Count Vectors
##### Document 1
```
I love NLP
```

Counts:
| Word | Count |
| ---- | ----- |
| I    | 1     |
| love | 1     |
| NLP  | 1     |
| AI   | 0     |

Vector:
```python
[
 1,
 1,
 1,
 0
]
```

##### Document 2
```
I love AI
```

Counts:
| Word | Count |
| ---- | ----- |
| I    | 1     |
| love | 1     |
| NLP  | 0     |
| AI   | 1     |

Vector:
```python
[
 1,
 1,
 0,
 1
]
```

### Final Representation
#### Vocabulary
```
[I, love, NLP, AI]
```

#### Document Matrix
| Document | I | love | NLP | AI |
| -------- | - | ---- | --- | -- |
| Doc1     | 1 | 1    | 1   | 0  |
| Doc2     | 1 | 1    | 0   | 1  |

This is called the: **Document-Term Matrix (DTM)**

### Binary Bag of Words


```python

```


Meaning:
```

```

### Advantages of Bag of Words
#### Simple
Very easy to understand.

#### Fast
Computationally cheap.

#### Works Well for Small Problems
Examples:
* Spam Detection
* Basic Sentiment Analysis
* Topic Classification

#### Interpretable
You can see exactly:
```
Which words contributed
```

### Major Limitations of BoW
#### Limitation 1: Word Order Is Lost
Sentence:
```
Dog bites man
```

Sentence:
```
Man bites dog
```

BoW:
```
Same Vector
```

But meanings are completely different.

#### Limitation 2: No Context
Word:
```
bank
```

could mean:
* River Bank
* Bank Account

BoW treats both identically.

No understanding of context.

#### Limitation 3: No Semantics
Vocabulary:
```
good

excellent
```

Humans know:
```
Similar Meaning
```

BoW:
```
Completely Different Words
```

No semantic understanding.

#### Limitation 4: Sparse Vectors
Vocabulary:
```
10000 words
```

Document:
```
Contains 50 words
```

Vector:
```
9950 zeros
```

This creates:
```
Sparse Matrices
```

#### Limitation 5: Vocabulary Explosion
Large corpus:
```
100000 words
```

BoW vector:
```
100000 dimensions
```

Huge memory usage.

### Python Example

````python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love NLP",
    "I love AI"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

print(
    vectorizer.get_feature_names_out()
)

print(X.toarray())
````

Notice:
```
Vocabulary created automatically.
```

### Why BoW Led to TF-IDF
Consider words:
```
the
is
am
are
```

They appear everywhere.

BoW gives them high importance.

Example:
```
the
```

appears:
```
1000 times
```

BoW says:
```
Very Important
```

Reality:
```
Not Informative
```

This problem led to:
```
TF-IDF
```

which weights words intelligently.


### BoW vs One-Hot Encoding

| Feature            | One-Hot         | BoW             |
| ------------------ | --------------- | --------------- |
| Represents         | Single Word     | Entire Document |
| Dimension          | Vocabulary Size | Vocabulary Size |
| Values             | One 1, Rest 0   | Counts          |
| Captures Frequency | No              | Yes             |
| Captures Order     | No              | No              |
