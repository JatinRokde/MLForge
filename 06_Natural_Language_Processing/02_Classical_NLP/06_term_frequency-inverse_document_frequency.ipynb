TF-IDF solves this problem.

### What is TF-IDF?
#### Definition
TF-IDF measures how important a word is in a document relative to an entire corpus.

It combines:
```
TF = Importance inside document

IDF = Rarity across corpus
```

Formula:
```
TF-IDF = TF × IDF
```

**Diagram**
```
Word
  ↓

TF
(Local Importance)

  +

IDF
(Global Rarity)

  ↓

TF-IDF Score
```

### Intuition
A good keyword should satisfy:

#### Condition 1
Appear frequently in the document.
```
High TF
```

#### Condition 2
Appear rarely in other documents.
```
High IDF
```

Result:
```
High TF-IDF
```

##### Example
Document about NLP:
```
NLP NLP NLP NLP

transformers

embeddings
```

Important words:
```
NLP
transformers
embeddings
```

Less important:
```
the
is
am
are
```

TF-IDF automatically learns this.

### TF-IDF Formula
Recall:
```
Term Frequency (TF) = Count(term) / Total Terms
```

```
Inverse Document Frequency (IDF) = log(N / DF)
```

#### Combined Formula
```
TF-IDF = TF × IDF
```

### Complete Example
Corpus:
```
Doc1:
cat cat dog

Doc2:
cat bird

Doc3:
cat fish
```

Total Documents:
```
N = 3
```

#### Step 1 — Calculate TF
For Doc1:
```
cat cat dog
```

Total Words:
```
3
```

Count(cat):
```
2
```

TF(cat):
```
2 / 3

=
0.67
```

Count(dog):
```
1
```

TF(dog):
```
1 / 3 = 0.33
```

#### Step 2 — Calculate IDF
DF(cat):
```
3
```

IDF(cat):
```
log(3/3) = 0
```

DF(dog):
```
1
```

IDF(dog):
```
log(3/1) = 1.10
```

#### Step 3 — Calculate TF-IDF
For cat:
```
0.67 × 0 = 0
```

For dog:
```
0.33 × 1.10 ≈ 0.36
```

Result:
| Word | TF-IDF |
| ---- | ------ |
| cat  | 0      |
| dog  | 0.36   |

#### Interpretation
```
dog
```

is more informative than:
```
cat
```

because:
```
cat
```

appears everywhere.

### Why TF-IDF Is Better Than BoW
BoW:
```
Word Count
```

TF-IDF:
```
Importance Score
```

Example:

Word:
```
the
```

BoW:
```
100
```

Very important.

TF-IDF:
```
Near Zero
```

Not important.

#### Comparison
| Feature                 | BoW | TF-IDF |
| ----------------------- | --- | ------ |
| Counts Words            | Yes | Yes    |
| Uses Corpus Information | No  | Yes    |
| Reduces Common Words    | No  | Yes    |
| Better Search Quality   | No  | Yes    |

### Why TF-IDF Became Popular
Before embeddings existed:
```
TF-IDF
```

was the dominant representation for:
* Search
* Document Ranking
* Classification
* Information Retrieval

Reason:

It captures:
```
Importance
```

rather than:
```
Simple Counts
```

### Python Example




````python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "cat cat dog",
    "cat bird",
    "cat fish"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(documents)

print(
    vectorizer.get_feature_names_out()
)

print(
    X.toarray()
)
````

Observe:
```
cat
```

receives a smaller weight than:
```
dog
bird
fish
```
because it appears everywhere.

### Limitations of TF-IDF
#### Limitation 1 — No Semantics
Words:
```
good
excellent
```

Humans:
```
Similar Meaning
```

TF-IDF:
```
Completely Different Words
```

#### Limitation 2 — No Context
Word:
```
bank
```

Could mean:
```
River Bank

Financial Bank
```

TF-IDF cannot distinguish.

#### Limitation 3 — No Word Order
Sentence:
```
Dog bites man
```

Sentence:
```
Man bites dog
```

Nearly identical TF-IDF representations.

#### Limitation 4 — Sparse Vectors
Vocabulary:
```
100,000 words
```

Document:
```
50 words
```

Vector:
```
Mostly zeros
```

#### Limitation 5 — Fixed Vocabulary
New words:
```
GenAI
```
need vocabulary updates.

### TF-IDF vs Embeddings
| Feature          | TF-IDF | Embeddings |
| ---------------- | ------ | ---------- |
| Sparse           | Yes    | No         |
| Dense            | No     | Yes        |
| Semantics        | No     | Yes        |
| Context          | No     | Yes        |
| Vocabulary-Based | Yes    | No         |
| Modern LLMs      | No     | Yes        |

Example:
```
king
queen
```

TF-IDF:
```
Unrelated
```

Embeddings:
```
Highly Similar
```

This is why Word2Vec became revolutionary.
