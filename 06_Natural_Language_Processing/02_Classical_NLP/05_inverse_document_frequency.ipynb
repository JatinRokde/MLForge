TF answers:
```
How important is a word inside a document?
```

But TF has a major problem.

Consider a corpus:
```
Doc1: the cat sat on the mat

Doc2: the dog ran in the park

Doc3: the bird flew over the tree
```

The word:
```
the
```

appears many times.

TF says:
```
Very Important
```

But:
```
the
```
carries very little information.

This problem leads to:
```
Inverse Document Frequency (IDF)
```

### Motivation
Consider these documents.
```
Doc1:
The cat sat

Doc2:
The dog ran

Doc3:
The bird flew
```

Word:
```
the
```

appears in:
```
All documents
```

Word:
```
cat
```

appears in:
```
Only one document
```

Intuition:
```
cat
```

is more informative than:
```
the
```
because it helps distinguish documents.

This idea is exactly what IDF measures.

### What is IDF?
#### Definition
Inverse Document Frequency (IDF) measures how rare or unique a term is across an entire corpus.

Intuition:
```
Appears in many documents
           ↓
Less Important

Appears in few documents
           ↓
More Important
```

#### Diagram
```
Many Documents
      ↓
 Low IDF

Few Documents
      ↓
High IDF
```

### Document Frequency (DF)
Before understanding IDF, understand:
```
Document Frequency (DF)
```

#### Definition
Document Frequency is the number of documents containing a term.

##### Example Corpus
```
Doc1:
cat dog

Doc2:
cat bird

Doc3:
cat fish
```

Word:
```
cat
```

appears in:
```
Doc1
Doc2
Doc3
```

Therefore:
```
DF(cat) = 3
```

Word:
```
dog
```

appears only in:
```
Doc1
```

Therefore:
```
DF(dog) = 1
```

##### Important
DF counts:
```
Documents
```
not occurrences.

Example:
```
Doc1:
cat cat cat cat
```

Still:
```
DF(cat) = 1
```
because it appears in only one document.

### What Does IDF Measure?
IDF is based on:
```
How many documents contain the term?
```

If a term appears everywhere:
```
Low Information
```

If a term appears rarely:
```
High Information
```

#### Example
Corpus:
```
1000 documents
```

Word:
```
the
```

appears in:
```
1000 documents
```

IDF:
```
Very Low
```

Word:
```
transformer
```

appears in:
```
10 documents
```

IDF:
```
High
```

### IDF Formula
The most common formula:
```
IDF(t) = log(N / DF(t))
```

Where:
```
N = Total Number of Documents

DF(t) = Document Frequency
```


### Understanding the Formula
Suppose:
```
N = 1000
```

Word:
```
the
```

appears in:
```
1000 documents
```

IDF:
```
log(1000/1000) = log(1) = 0
```

Result:
```
Very Low Importance
```

Now:
```
transformer
```

appears in:
```
10 documents
```

IDF:
```
log(1000/10) = log(100) ≈ 2
```

Result:
```
Much Higher Importance
```

### Example Calculation
Corpus:
```
Doc1:
cat dog

Doc2:
cat bird

Doc3:
cat fish
```

Total Documents:
```
N = 3
```

#### DF Values
| Word | DF |
| ---- | -- |
| cat  | 3  |
| dog  | 1  |
| bird | 1  |
| fish | 1  |

#### IDF(cat)
```
log(3/3) = 0
```

#### IDF(dog)
```
log(3/1) = 1.10
```

#### IDF(bird)
```
1.10
```

#### IDF(fish)
```
1.10
```

Interpretation:
```
cat

is common.

dog
bird
fish

are distinctive.
```

#### Rule
```text
Higher DF
      ↓
Lower IDF

Lower DF
      ↓
Higher IDF
```


### Why Use Logarithm?
Without log:
```
1000000 / 1 = 1000000
```
Huge value.

Using log:
```
log(1000000) ≈ 6
```
Much more manageable.

#### Benefits
* Stable Values
* Reduced Scale
* Better Numerical Behavior


### TF vs DF vs IDF
| Metric | Looks At        |
| ------ | --------------- |
| TF     | Single Document |
| DF     | Entire Corpus   |
| IDF    | Entire Corpus   |

#### Example
Document:
```
NLP NLP NLP
```

TF:
```
High
```

because NLP appears frequently.

Corpus:
```
Every document contains NLP
```

IDF:
```
Low
```

because NLP is not unique.


### Why IDF Alone Is Not Enough
Suppose:
```
transformer
```

appears in only one document.

IDF:
```
Very High
```

But suppose it appears:
```
Only once
```

inside that document.

Should it dominate the document?
```
Not necessarily.
```

Need both:
```
Local Importance
+
Global Rarity
```

This leads to:
```
TF-IDF
```

#### Diagram
```
TF = Importance Inside Document

IDF = Importance Across Corpus

TF + IDF = Best of Both
```


### Smoothed IDF
Real systems often use:
```
IDF
=
log(N+1 / DF+1) + 1
```

Reason:
```
Avoid:

Division By Zero
```

and improve numerical stability.


### Python Example


````python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "cat dog",
    "cat bird",
    "cat fish"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(documents)

print(vectorizer.idf_)
````

Output will show:
```
cat
```

has lower IDF than:
```
dog
bird
fish
```

### Why IDF Was Revolutionary
Before IDF:
```
Common Words = High Scores
```

After IDF:
```
Common Words = Downweighted

Rare Informative Words = Upweighted
```

This dramatically improved:
* Information Retrieval
* Search Engines
* Document Ranking


### Advantages of IDF
* Measures global importance of terms
* Downweights common words
* Highlights rare and informative words
* Improves search and retrieval quality
* Forms the foundation of TF-IDF

### Limitations of IDF
* Ignores frequency within a document
* Cannot capture semantics
* Cannot understand context
* Rare words may receive excessively high importance
* Requires corpus-wide statistics
