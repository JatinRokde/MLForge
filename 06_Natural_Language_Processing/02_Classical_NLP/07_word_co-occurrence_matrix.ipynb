Words that occur in similar contexts tend to have similar meanings.

This idea is called:
```
Distributional Hypothesis
```

### Motivation
Let's look at a corpus.
```
The cat drinks milk

The dog drinks milk

The cat eats fish

The dog eats meat
```

Observe:
```
cat

dog
```
appear in similar environments:

```
drinks

milk

eats
```

Intuitively:
```
cat ≈ dog
```

because they occur in similar contexts.

This observation led to one of the most important ideas in NLP.

### Distributional Hypothesis
A famous NLP principle:
> "You shall know a word by the company it keeps."

Often attributed to: John Rupert Firth

#### Meaning
If two words occur in similar contexts:
```
They probably have similar meanings.
```

##### Example
```
cat

dog
```

occur near:
```
pet

food

animal

milk
```

Therefore:
```
cat and dog
```

are semantically related.

This idea became the foundation of:
* Word2Vec
* GloVe
* FastText
* Modern Embeddings

### What is a Word Co-occurrence Matrix?
#### Definition
A Word Co-occurrence Matrix records how often words appear together within a specified context window.

Instead of counting:
```
Word → Document
```

like TF-IDF,

count:
```
Word → Neighboring Words
```

**Diagram**
```
TF-IDF
Word ↔ Document

Co-occurrence
Word ↔ Word
```

### What is a Context Window?
Before building the matrix, needed:
```
Context Window
```

Sentence:
```
I love NLP very much
```

Suppose:
```
Window Size = 1
```

For word:
```
NLP
```

Context words:
```
love

very
```

**Diagram**
```
I  love  NLP  very  much

        ↑

Window = 1

Context:
love
very
```

If:
```
Window Size = 2
```

Context becomes:
```
I

love

very

much
```

Context window:
> A fixed number of neighboring words around a target word.

### Building a Co-occurrence Matrix
Corpus:
```
I love NLP

I love AI
```

#### Step 1 — Build Vocabulary
```
I

love

NLP

AI
```

Assume:
```
Window Size = 1
```

#### Step 2 — Generate Co-occurrences
Sentence:
```
I love NLP
```

Target:
```
I
```

Context:
```
love
```

Count:
```
(I, love) += 1
```

Target:
```
love
```

Context:
```
I

NLP
```

Counts:
```
(love, I) += 1

(love, NLP) += 1
```
Continue similarly for all words.

### Final Co-occurrence Matrix
| Word | I | love | NLP | AI |
| ---- | - | ---- | --- | -- |
| I    | 0 | 2    | 0   | 0  |
| love | 2 | 0    | 1   | 1  |
| NLP  | 0 | 1    | 0   | 0  |
| AI   | 0 | 1    | 0   | 0  |

This is a:
```
Word × Word Matrix
```

### Interpretation
Look at rows:

#### Row for NLP
```
[0,1,0,0]
```

#### Row for AI
```
[0,1,0,0]
```

Notice:
```
NLP

AI
```

have identical contexts.

Therefore:
```
NLP ≈ AI
```

This is exactly the insight used by embeddings.

### Larger Example
Corpus:
```
The cat drinks milk

The dog drinks milk
```

Vocabulary:
```
the
cat
dog
drinks
milk
```

Partial Matrix:
| Word   | cat | dog | drinks | milk |
| ------ | --- | --- | ------ | ---- |
| cat    | 0   | 0   | 1      | 0    |
| dog    | 0   | 0   | 1      | 0    |
| drinks | 1   | 1   | 0      | 1    |
| milk   | 0   | 0   | 1      | 0    |

#### Observation
```
cat

dog
```

have similar context patterns.

Hence:
```
Similar Meaning
```

### Why Is This Important?
For the first time in NLP:
```
Meaning
```

starts emerging automatically.

Earlier approaches:
* Bag of Words
* TF-IDF

treated words independently.

Co-occurrence introduces:
```
Relationships Between Words
```

This was revolutionary.
### Matrix as Word Representation
Each row becomes a vecor.

Example:
```
cat
↓
[0,0,1,0]
```

```
dog
↓
[0,0,1,0]
```

These vectors represent words.

This was one of the earliest forms of:
```text
Word Embeddings
```
although still primitive.

### Similarity Using Co-occurrence
Suppose:
```
cat
↓
[0,0,5,4]
```

```
dog
↓
[0,0,4,5]
```

Very similar vectors.

Therefore:
```
cat ≈ dog
```

This idea later evolved into:
* Cosine Similarity
* Word2Vec
* GloVe

### Major Problems
#### Problem 1 — Huge Matrices
Vocabulary:
```
100,000 words
```

Matrix Size:
```
100,000 × 100,000
```

Entries:
```
10 Billion
```

Massive memory usage.

#### Problem 2 — Sparsity
Most word pairs never occur together.

Example:
```
transformer

banana
```

rarely co-occur.

Matrix becomes:
```
Mostly Zeros
```

A sparse representation.

#### Problem 3 — Limited Semantics
Context similarity helps but is limited.

Example:
```
king

queen
```

may not always appear in identical contexts.

Need more advanced methods.

### How NLP Solved These Problems
Researchers asked:
```
Can we compress this huge sparse matrix
into a small dense representation?
```

This led to:
* Latent Semantic Analysis (LSA)
* Singular Value Decomposition (SVD)
* Word2Vec
* GloVe

**Diagram**
```
Co-occurrence Matrix

Huge
Sparse

       ↓

Compression

       ↓

Dense Vectors

       ↓

Embeddings
```

### Relation to GloVe
GloVe stands for:
```
Global Vectors
```

Key Idea:

Use:
```
Global Co-occurrence Statistics
```

to learn embeddings.

In simple terms:
```
GloVe = Smarter Co-occurrence Matrix
```

### Relation to Word2Vec
Word2Vec also uses context.

But differently.

#### Co-occurrence Matrix
```
Explicitly stores counts
```

#### Word2Vec
```
Learns counts implicitly
through a neural network
```

### Comparison
| Method               | Uses Context |
| -------------------- | ------------ |
| Co-occurrence Matrix | Explicitly   |
| Word2Vec             | Implicitly   |
| GloVe                | Explicitly   |
| BERT                 | Contextually |
