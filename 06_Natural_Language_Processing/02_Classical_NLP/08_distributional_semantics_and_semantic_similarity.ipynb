### Motivation
Consider these words:
```
cat

dog
```

Humans immediately know:
```
Similar Meaning
```

Now consider:
```
cat

airplane
```

Clearly less similar.

Question:
```
How can a machine measure similarity?
```

This is the core problem of semantic similarity.

### What is Semantics?
#### Definition
Semantics refers to the meaning of words, phrases, or text.

##### Example
```
car

automobile
```

Different words.

Same meaning.

##### Example
```
doctor

physician
```

Very similar meaning.

##### Example
```
cat

banana
```

Very different meaning.

Humans understand this naturally.

Machines need mathematical representations.

### Distributional Semantics
This concept comes directly from the:
```
Distributional Hypothesis
```

The idea:
```
Words that occur in similar contexts
tend to have similar meanings.
```

##### Example
Corpus:
```
The cat drinks milk

The dog drinks milk

The cat eats fish

The dog eats meat
```

Observe:
```
cat

dog
```

appear near:
```
drinks

milk

eats
```

Therefore:
```
cat ≈ dog
```

This is called:
```
Distributional Semantics
```

### Vector Space Model
To compare words mathematically, it is represented as vectors.

##### Example
```
cat
↓
[2, 5, 1]
```

```
dog
↓
[2, 4, 1]
```

```
airplane
↓
[8, 0, 9]
```

Now words exist in a:
```
Vector Space
```

**Diagram**
```
          dog

           ●

      ● cat


-------------------->

                        ● airplane
```

Nearby words:
```
Similar Meaning
```

Far words:
```
Different Meaning
```

### What is Semantic Similarity?
#### Definition
Semantic similarity measures how similar the meanings of two words are.

##### Examples
###### High Similarity
```
car

automobile
```

###### Medium Similarity
```
car

truck
```

###### Low Similarity
```
car

banana
```

Goal:
```
Assign numerical similarity scores.
```

### Why Distance Matters
Suppose vectors:
```
cat
[2, 5]
```

```
dog
[2, 4]
```

```
airplane
[9, 0]
```

Intuition:
```
cat ↔ dog
Close
```

```
cat ↔ airplane
Far
```

Therefore:
```
Distance = Similarity Information
```

### Euclidean Distance
The most intuitive metric.

Measures:
```
Straight-Line Distance
between two points.
```

**Diagram**
```
cat ●

     \

      \

       ● dog
```

#### Interpretation
```
Smaller Distance = More Similar
```

#### Problem
```
Magnitude affects distance heavily.
```

### Cosine Similarity (Most Important)
Instead of distance:
```
Compare Angles
between vectors.
```

Smaller angle:
```
Higher Similarity
```

Larger angle:
```
Lower Similarity
```

#### Formula
$$
\cos(\theta) = \frac{\mathbf{A} \cdot \mathbf{B}}{|A| |B|}
$$

Don't memorize now.

Understand the intuition:
```
Same Direction = Similar Meaning
```

### Cosine Similarity Values
Range:
```
-1 to 1
```

#### 1
```
Exactly Same Direction

Perfect Similarity
```

#### 0
```
Orthogonal

No Similarity
```

#### -1
```
Opposite Directions
```

Rare in NLP.

Most NLP embeddings produce values between:
```
0 and 1
```

### Why Cosine Similarity Beats Euclidean Distance
Suppose:
```
cat
[1,1]
```

```
dog
[10,10]
```

Semantically:
```
Very Similar
```

Euclidean Distance:
```
Large
```

Cosine Similarity:
```
1
```

Perfect.

Because:
```
Direction matters

Magnitude doesn't
```

This is why:
```
Cosine Similarity
```

became the standard similarity metric in NLP.

### Similarity from Co-occurrence Matrix
Suppose:
```
cat
↓
[0,0,5,4]
```

```
dog
↓
[0,0,4,5]
```

Cosine Similarity:
```
Very High
```

Meaning:
```
cat ≈ dog
```

This is exactly how early NLP systems measured meaning.

```
cat and dog

Very Similar
```

### Nearest Neighbors
Given:
```
cat
```

Find:
```
Most Similar Words
```

Example:
```
cat

↓

dog

kitten

pet
```

This idea is heavily used in:
* Word Embeddings
* Vector Search
* RAG

### Why This Was Revolutionary
Before this:
```
Words = Symbols
```

After Distributional Semantics:
```
Words = Points in Space
```

Meaning became:
```
Mathematically Measurable
```

This was one of the biggest breakthroughs in NLP history.

### Connection to Word Embeddings
Co-occurrence Matrix:
```
Huge

Sparse
```

Researchers asked:
```
Can we learn smaller vectors
that preserve similarity?
```

This led to:
* Word2Vec
* GloVe
* FastText

Goal:
```
Similar Words
↓
Similar Vectors
```

### Distributional Semantics → Embeddings
```
Distributional Hypothesis
            ↓
Words in Similar Contexts
            ↓
Similar Co-occurrence Patterns
            ↓
Similar Vectors
            ↓
Word Embeddings
            ↓
Semantic Similarity
```
