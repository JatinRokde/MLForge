### 1. Standardization (Z-Score Scaling)
**Z-score scaling** (also called **standardization**) is a data normalization technique that transforms numerical values so they have:
- **Mean** = 0
- **Standard deviation** = 1

This makes features comparable even if they originally had different units or ranges.
****
#### Formula
**For each value *x***:
$$
z = \frac{(x - \mu)}{\sigma}
$$
Where:
- $\mu$ = mean of the feature
- $\sigma$ = standard deviation of the feature

**What the Z-score Represents**
- z = 0 -> value equals the mean
- z = 1 -> value is 1 standard deviation above the mean
- z = -1 -> value is 1 standard deviation below the mean

So the Z-score tells **how many standard deviations away** a value is from the average.

````python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled
````

**When to Use:**
- Linear models
- Logistic regression
- SVM
- PCA

### 2. Min-Max Normalization
**Min–max normalization** (also called **min–max scaling**) is a data preprocessing technique that rescales numerical features into a **fixed range**, most commonly **0 to 1**.

It preserves the original relationships between values while standardizing their scale.
***
#### Formula
**For each value *x***:
$$
x_{scaled} = \frac{x - x_{min}}{x_{max} - x_{min}}
$$
Where:
- $x_{\min}$ = minimum value in the feature
- $x_{\max}$ = maximum value in the feature

**What It Does:**
- The **smallest value becomes 0**
- The **largest value becomes 1**
- All other values fall between 0 and 1



````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_minmax = scaler.fit_transform(df)
df_minmax
````

**When to Use:**
- Neural Networks
- Image-like bounded features

**Important**: It is sensitive to outliers

### 3. Robust Scaling (Outlier Safe)
**Robust Scaling** is a data preprocessing technique used in machine learning and statistics to scale numerical features in a way that is **resistant to outliers**. It centers and scales data using **median** and **interquartile range (IQR)** instead of mean and standard deviation.
***
#### Formula
**For each feature:**
$$
x_{scaled} = \frac{x - median}{IQR}
$$

Where:
- **Median** = middle value of the feature
- **IQR (Interquartile Range)** = Q3 − Q1

This transformation:
- Centers the data around **0**
- Scales so that the middle 50% of values typically lie between **-1 and 1**


**Key Properties:**
- Resistant to extreme outliers
- Does **not** force values into a fixed range
- Preserves relative distances between typical values
- Good default choice when distributions are skewed


````python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df_robust = scaler.fit_transform(df)
df_robust
````

- Best for real-world numeric data
- Resistant to outliers

#### Comparison
| Method                    | Uses           | Sensitive to Outliers? |
| ------------------------- | -------------- | ---------------------- |
| Standardization (Z-score) | Mean & Std Dev | Yes                    |
| Min–Max Scaling           | Min & Max      | Yes                    |
| Robust Scaling            | Median & IQR   | No                     |


### Order
```
Clean -> Encode -> Scale -> Model
```
Scaling before encoding is meaningless

### Normalization vs Standardization
| Aspect         | Standardization | Normalization |
| -------------- | --------------- | ------------- |
| Output range   | (-∞, +∞)        | (0, 1)        |
| Outlier safe   | No              | No             |
| Distance-based | Yes               | Yes           |
| Gradient-based | Yes              | Yes           |

### Scaling in Pipelines
```python

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])
```
- Prevents leakage
- Cleaner code
- Production safe
